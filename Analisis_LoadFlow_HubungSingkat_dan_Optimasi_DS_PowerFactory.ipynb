```python
import powerfactory

# ==========================
# INPUT GI
# ==========================

SUBSTATION_NAME = "ALTA PRIMA5"

# Voltage limit
VMAX = 1.05
VMIN = 0.95

def evaluate_voltage(sub_name):

    app = powerfactory.GetApplication()

    if app is None:
        raise RuntimeError("Cannot connect to PowerFactory")

    # ==========================
    # RUN LOAD FLOW
    # ==========================

    lf = app.GetFromStudyCase("ComLdf")
    lf.iopt_net = 0
    lf.iopt_o4  = 0

    rc = lf.Execute()

    if rc != 0:
        raise RuntimeError("Load flow failed")

    print("✅ Load Flow Completed\n")

    # ==========================
    # GET ALL SUBSTATIONS
    # ==========================

    all_subs = app.GetCalcRelevantObjects("*.ElmSubstat")

    sub_map = {s.loc_name.strip(): s for s in all_subs}

    sub = sub_map.get(sub_name)

    if not sub:
        print(f"❌ Substation '{sub_name}' not found")
        return

    print(f"🔍 Substation Found : {sub_name}")

    # ==========================
    # GET BUS IN SUBSTATION
    # ==========================

    terms = sub.GetContents("*.ElmTerm",1)

    buses = [b for b in terms if b.GetAttribute("iUsage")==0]

    if not buses:
        print("⚠ No bus found in this GI")
        return

    print(f"Number of buses : {len(buses)}\n")

    # ==========================
    # PRINT BUS VOLTAGE
    # ==========================

    print("===== BUS VOLTAGE =====")

    PIv = 0

    for bus in buses:

        name = bus.loc_name
        vpu  = bus.GetAttribute("m:u")
        vkv  = bus.GetAttribute("m:Ul")

        print(f"{name:<25} {vpu:6.4f} pu   {vkv:7.2f} kV")

        # ==========================
        # PENALTY CALCULATION
        # ==========================

        if vpu > VMAX:

            penalty = ((vpu - VMAX)/VMAX)**2
            PIv += penalty

            print(f"   ⚠ Voltage HIGH violation → penalty = {penalty:.6f}")

        elif vpu < VMIN:

            penalty = ((VMIN - vpu)/VMIN)**2
            PIv += penalty

            print(f"   ⚠ Voltage LOW violation → penalty = {penalty:.6f}")

        else:

            print("   ✓ Voltage within limit")

    print("\n===== PERFORMANCE INDEX =====")

    print(f"PIv = {PIv:.6f}")


# ==========================
# MAIN
# ==========================

if __name__ == "__main__":

    evaluate_voltage(SUBSTATION_NAME)
```

```python
import powerfactory

SUBSTATION_NAME = "ALTA PRIMA5"

VMAX = 1.05
VMIN = 0.95


def evaluate_voltage(sub_name):
    app = powerfactory.GetApplication()
    if app is None:
        raise RuntimeError("Cannot connect to PowerFactory")

    # Run load flow
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("Load flow object 'ComLdf' not found in study case")

    lf.iopt_net = 0
    lf.iopt_o4 = 0

    rc = lf.Execute()
    if rc != 0:
        raise RuntimeError("Load flow failed")

    print("✅ Load Flow Completed\n")

    # Get all substations
    all_subs = app.GetCalcRelevantObjects("*.ElmSubstat")
    sub_map = {s.loc_name.strip(): s for s in all_subs}
    sub = sub_map.get(sub_name.strip())

    if not sub:
        print(f"❌ Substation '{sub_name}' not found")
        return

    print(f"🔍 Substation Found: {sub_name}")

    # Get buses in substation
    terms = sub.GetContents("*.ElmTerm", 1)
    buses = [b for b in terms if b.GetAttribute("iUsage") == 0]

    if not buses:
        print("⚠ No bus found in this GI")
        return

    print(f"Number of buses: {len(buses)}\n")
    print("===== BUS VOLTAGE =====")

    PIv = 0.0

    for bus in buses:
        name = bus.loc_name
        vpu = bus.GetAttribute("m:u")
        vkv = bus.GetAttribute("m:Ul")

        if vpu is None or vkv is None:
            print(f"{name:<25} Data not available")
            continue

        print(f"{name:<25} {vpu:6.4f} pu   {vkv:7.2f} kV")

        if vpu > VMAX:
            penalty = ((vpu - VMAX) / VMAX) ** 2
            PIv += penalty
            print(f"   ⚠ Voltage HIGH violation → penalty = {penalty:.6f}")

        elif vpu < VMIN:
            penalty = ((VMIN - vpu) / VMIN) ** 2
            PIv += penalty
            print(f"   ⚠ Voltage LOW violation → penalty = {penalty:.6f}")

        else:
            print("   ✓ Voltage within limit")

    print("\n===== PERFORMANCE INDEX =====")
    print(f"PIv = {PIv:.6f}")


if __name__ == "__main__":
    evaluate_voltage(SUBSTATION_NAME)
```

```python
import os
import time
from itertools import combinations

import powerfactory
import pandas as pd

# ── 0. USER CONFIGURATION ────────────────────────────────────────────
# Exact BC Coupler names to be tested
COUP_LIST = [
    "BC Waru",
    "BC Rungkut",
    "BC Sawahan",
    "BC Gunung Sari",
    "BC Tandes",
    "BC Altaprima",
]

# All substations whose buses will be included in PIv total
SUBSTAT_LIST = [
    "GRESIK BARU5", "KRIAN5", "GRESIK LAMA55", "KARANGPILANG5", "SAWAHAN5",
    "DARMO GRANDE5", "ALTA PRIMA5", "SEGOROMADU5", "SAMBIKEREP5", "WILMAR5",
    "BAMBE5", "GUNUNG SARI5", "PETRO KIMIA5", "WARU5", "RUNGKUT5",
     "BUDURAN5", "SURABAYA SELATAN5", "SUKOLILO5", "HANIL JAYA STEEL5",
    "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5",
    "SIDOARJO5", "NEWBUDURAN5", "SEDATI5", "SIMPANG5", "GILITIMUR5",
    "NWARU5", "TANDES5", "PERAK5", "UNDAAN5", "KUPANG5",
    "KREMBANGAN5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "KEDINDING5", "BANGKALAN5",
    "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5"
]

# Evaluate scenarios with exactly 1 BC open, then exactly 2 BC open
OPEN_COUNTS = [1, 2]

# PIv Mode A parameters
# PIv = Σ [ Wvi/(2n) * ( | |Vi| - |Vi_rated| | / ΔVlim )^(2n) ]
N_EXP = 1
DEFAULT_WEIGHT = 1.0
V_RATED_PU_DEFAULT = 1.0
DV_LIMIT_PU_DEFAULT = 0.05

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\Hasil Uji"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "BruteForce_BC_PIV_1BC_2BC.xlsx")


# ── 1. POWERFACTORY INIT ────────────────────────────────────────────
def get_app():
    app = powerfactory.GetApplication()
    if not app:
        raise RuntimeError("Run inside DIgSILENT PowerFactory")
    app.PrintPlain("=== BRUTE FORCE BC VOLTAGE START (1 BC THEN 2 BC) ===")
    return app


# ── 2. LOAD FLOW ────────────────────────────────────────────────────
def run_loadflow(app):
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("ComLdf not found in active study case")

    lf.iopt_net = 0
    lf.iopt_o4 = 0

    return lf.Execute() == 0


# ── 3. CACHE OBJECTS ────────────────────────────────────────────────
def cache_objects(app):
    # Cache couplers
    couplers = {}
    for c in app.GetCalcRelevantObjects("*.ElmCoup"):
        name = c.loc_name.strip()
        if name in COUP_LIST:
            couplers[name] = {
                "obj": c,
                "orig_open": bool(c.IsOpen())
            }

    ordered_coupler_names = [name for name in COUP_LIST if name in couplers]
    app.PrintPlain(f"Found {len(ordered_coupler_names)} couplers from COUP_LIST")

    # Cache substations
    subs = {
        s.loc_name.strip(): s
        for s in app.GetCalcRelevantObjects("*.ElmSubstat")
    }

    # Get all buses from all substations in SUBSTAT_LIST
    buses = []
    seen_keys = set()

    for sub_name in SUBSTAT_LIST:
        sub = subs.get(sub_name)
        if not sub:
            app.PrintPlain(f"Warning: Substation '{sub_name}' not found")
            continue

        for b in sub.GetContents("*.ElmTerm", 1):
            try:
                if b.GetAttribute("iUsage") != 0:
                    continue
            except Exception:
                continue

            obj_key = id(b)
            if obj_key in seen_keys:
                continue
            seen_keys.add(obj_key)

            buses.append({
                "substation": sub_name,
                "bus_name": b.loc_name.strip(),
                "obj": b
            })

    app.PrintPlain(f"Total monitored buses from SUBSTAT_LIST = {len(buses)}")
    return couplers, ordered_coupler_names, buses


# ── 4. RESTORE SYSTEM ───────────────────────────────────────────────
def restore_couplers(couplers):
    for info in couplers.values():
        obj = info["obj"]
        orig_open = info["orig_open"]

        try:
            current_open = bool(obj.IsOpen())
        except Exception:
            continue

        if current_open != orig_open:
            if orig_open:
                obj.Open()
            else:
                obj.Close()


# ── 5. APPLY OPEN SCENARIO ──────────────────────────────────────────
def apply_open_scenario(couplers, open_names):
    open_set = set(open_names)

    for name, info in couplers.items():
        obj = info["obj"]
        if name in open_set:
            obj.Open()
        else:
            obj.Close()


# ── 6. CALCULATE PIv MODE A ─────────────────────────────────────────
def calculate_piv_total_mode_a(buses, verbose=False):
    total_piv = 0.0
    details = []

    for item in buses:
        sub_name = item["substation"]
        bus_name = item["bus_name"]
        bus_obj = item["obj"]

        vpu = bus_obj.GetAttribute("m:u")
        vkv = bus_obj.GetAttribute("m:Ul")

        if vpu is None or vkv is None:
            details.append({
                "Substation": sub_name,
                "Bus": bus_name,
                "Vpu": None,
                "VkV": None,
                "Weight": DEFAULT_WEIGHT,
                "Vrated_pu": V_RATED_PU_DEFAULT,
                "DeltaVlim_pu": DV_LIMIT_PU_DEFAULT,
                "Deviation": None,
                "PIv_i": None,
                "Status": "NO_RESULT",
            })
            continue

        weight = DEFAULT_WEIGHT
        v_rated = V_RATED_PU_DEFAULT
        dv_limit = DV_LIMIT_PU_DEFAULT

        deviation = abs(abs(vpu) - abs(v_rated))
        piv_i = (weight / (2 * N_EXP)) * (deviation / dv_limit) ** (2 * N_EXP)
        total_piv += piv_i

        if vpu > (v_rated + dv_limit):
            status = "HIGH"
        elif vpu < (v_rated - dv_limit):
            status = "LOW"
        else:
            status = "NORMAL"

        details.append({
            "Substation": sub_name,
            "Bus": bus_name,
            "Vpu": round(vpu, 6),
            "VkV": round(vkv, 3),
            "Weight": weight,
            "Vrated_pu": v_rated,
            "DeltaVlim_pu": dv_limit,
            "Deviation": round(deviation, 8),
            "PIv_i": round(piv_i, 8),
            "Status": status,
        })

        if verbose:
            print(
                f"{sub_name:<18} {bus_name:<28} "
                f"{vpu:8.5f} pu   {vkv:8.3f} kV   "
                f"dev={deviation:.6f}   PIv_i={piv_i:.6f}"
            )

    return total_piv, details


# ── 7. BASELINE ─────────────────────────────────────────────────────
def evaluate_baseline(app, couplers, buses):
    restore_couplers(couplers)

    if not run_loadflow(app):
        raise RuntimeError("Baseline load flow failed")

    total_piv, details = calculate_piv_total_mode_a(buses, verbose=False)
    return total_piv, details


# ── 8. EVALUATE A GIVEN OPEN SET ────────────────────────────────────
def evaluate_open_set(app, couplers, buses, open_names):
    restore_couplers(couplers)

    try:
        apply_open_scenario(couplers, open_names)

        if not run_loadflow(app):
            return None, [], "LOADFLOW_FAILED"

        total_piv, details = calculate_piv_total_mode_a(buses, verbose=False)
        return total_piv, details, "OK"

    except Exception as e:
        return None, [], f"ERROR: {e}"

    finally:
        # Always restore after each scenario
        restore_couplers(couplers)


# ── 9. MAIN BRUTE FORCE ─────────────────────────────────────────────
def run_bruteforce():
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    app = get_app()

    couplers, coupler_names, buses = cache_objects(app)

    if not couplers:
        raise RuntimeError("No couplers found from COUP_LIST")

    if not coupler_names:
        raise RuntimeError("No ordered coupler names available")

    if not buses:
        raise RuntimeError("No buses found from SUBSTAT_LIST")

    results = []
    detail_rows = []

    baseline_piv = None
    best_score = float("inf")
    best_open_names = None
    elapsed = 0.0

    try:
        # Baseline
        baseline_piv, baseline_details = evaluate_baseline(app, couplers, buses)
        app.PrintPlain(f"Baseline PIv Total = {baseline_piv:.6f}")

        for row in baseline_details:
            detail_rows.append({
                "Scenario": "BASELINE",
                "N Open": 0,
                "Opened Couplers": "-",
                **row
            })

        t0 = time.time()
        scenario_no = 1

        # First 1 BC open, then 2 BC open
        for n_open in OPEN_COUNTS:
            if n_open > len(coupler_names):
                app.PrintPlain(f"Skip n_open={n_open} because available couplers only {len(coupler_names)}")
                continue

            app.PrintPlain(f"\n===== START EVALUATION: OPEN {n_open} BC =====")

            for combo in combinations(coupler_names, n_open):
                open_names = list(combo)
                scenario_label = ", ".join(open_names)

                app.PrintPlain(f"\n[{scenario_no}] OPEN ({n_open} BC): {scenario_label}")

                score, details, status = evaluate_open_set(app, couplers, buses, open_names)

                if score is None:
                    results.append({
                        "No": scenario_no,
                        "N Open": n_open,
                        "Opened Couplers": scenario_label,
                        "PIv Total": None,
                        "Delta vs Baseline": None,
                        "Status": status
                    })
                    scenario_no += 1
                    continue

                delta = score - baseline_piv

                results.append({
                    "No": scenario_no,
                    "N Open": n_open,
                    "Opened Couplers": scenario_label,
                    "PIv Total": round(score, 8),
                    "Delta vs Baseline": round(delta, 8),
                    "Status": status
                })

                for row in details:
                    detail_rows.append({
                        "Scenario": f"OPEN {scenario_label}",
                        "N Open": n_open,
                        "Opened Couplers": scenario_label,
                        **row
                    })

                if score < best_score:
                    best_score = score
                    best_open_names = open_names
                    app.PrintPlain(f"--> NEW BEST: {best_score:.6f}")

                scenario_no += 1

        elapsed = time.time() - t0

    finally:
        # Final guarantee: restore all couplers to original states
        restore_couplers(couplers)

    # ── SAVE RESULT ─────────────────────────────────────────────────
    df = pd.DataFrame(results)
    df_detail = pd.DataFrame(detail_rows)

    if not df.empty and "PIv Total" in df.columns:
        df_rank = (
            df[df["PIv Total"].notna()]
            .sort_values(by="PIv Total", ascending=True)
            .reset_index(drop=True)
        )
    else:
        df_rank = pd.DataFrame()

    df_best = pd.DataFrame([{
        "Best Opened Couplers": ", ".join(best_open_names) if best_open_names else None,
        "Best N Open": len(best_open_names) if best_open_names else None,
        "Best PIv Total": round(best_score, 8) if best_open_names else None,
        "Baseline PIv Total": round(baseline_piv, 8) if baseline_piv is not None else None,
        "Execution Time (s)": round(elapsed, 2)
    }])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df.to_excel(writer, sheet_name="All Scenario", index=False)
        df_rank.to_excel(writer, sheet_name="Ranking", index=False)
        df_best.to_excel(writer, sheet_name="Best", index=False)
        df_detail.to_excel(writer, sheet_name="Bus Detail", index=False)

    # ── FINAL LOG ───────────────────────────────────────────────────
    app.PrintPlain("\n===== FINAL RESULT =====")
    if baseline_piv is not None:
        app.PrintPlain(f"Baseline PIv Total : {baseline_piv:.6f}")
    else:
        app.PrintPlain("Baseline PIv Total : N/A")

    if best_open_names:
        app.PrintPlain(f"Best Opened Couplers : {', '.join(best_open_names)}")
        app.PrintPlain(f"Best N Open          : {len(best_open_names)}")
        app.PrintPlain(f"Best PIv Total       : {best_score:.6f}")
    else:
        app.PrintPlain("Best Opened Couplers : N/A")
        app.PrintPlain("Best N Open          : N/A")
        app.PrintPlain("Best PIv Total       : N/A")

    app.PrintPlain(f"Execution Time       : {elapsed:.2f} s")
    app.PrintPlain(f"Saved                : {OUTPUT_FILE}")


# ── RUN ─────────────────────────────────────────────────────────────
if __name__ == "__main__":
    run_bruteforce()
```

```python
# file: bruteforce_bc_pipf_core.py
import os
from itertools import combinations

import pandas as pd
import powerfactory

COUP_LIST = [
    "BC Waru",
    "BC Rungkut",
    "BC Sawahan",
    "BC Gunung Sari",
    "BC Tandes",
    "BC Altaprima",
]

SUBSTAT_LIST = [
    "GRESIK BARU5", "KRIAN5", "GRESIK LAMA55", "KARANGPILANG5", "SAWAHAN5",
    "DARMO GRANDE5", "ALTA PRIMA5", "SEGOROMADU5", "SAMBIKEREP5", "WILMAR5",
    "BAMBE5", "GUNUNG SARI5", "PETRO KIMIA5", "WARU5", "RUNGKUT5",
    "BUDURAN5", "SURABAYA SELATAN5", "SUKOLILO5", "HANIL JAYA STEEL5",
    "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5",
    "SIDOARJO5", "NEWBUDURAN5", "SEDATI5", "SIMPANG5", "GILITIMUR5",
    "NWARU5", "TANDES5", "PERAK5", "UNDAAN5", "KUPANG5",
    "KREMBANGAN5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "KEDINDING5", "BANGKALAN5",
    "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5"
]

OPEN_COUNTS = [1, 2]
W_I = 1.0
N_EXP = 1

OUTPUT_FILE = r"D:\Robbyo\S2 UGM\2026\Tesis\Hasil Uji\BruteForce_BC_PIpf_Core.xlsx"


def get_app():
    app = powerfactory.GetApplication()
    if not app:
        raise RuntimeError("Run inside DIgSILENT PowerFactory")
    return app


def run_loadflow(app):
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("ComLdf not found")
    lf.iopt_net = 0
    lf.iopt_o4 = 0
    return lf.Execute() == 0


def get_couplers(app):
    couplers = {}
    for coup in app.GetCalcRelevantObjects("*.ElmCoup"):
        name = coup.loc_name.strip()
        if name in COUP_LIST:
            couplers[name] = {
                "obj": coup,
                "orig_open": bool(coup.IsOpen()),
            }
    return couplers


def get_monitored_buses(app):
    subs = {
        s.loc_name.strip(): s
        for s in app.GetCalcRelevantObjects("*.ElmSubstat")
    }

    buses = []
    seen = set()

    for sub_name in SUBSTAT_LIST:
        sub = subs.get(sub_name)
        if not sub:
            continue

        for bus in sub.GetContents("*.ElmTerm", 1):
            key = id(bus)
            if key in seen:
                continue
            seen.add(key)
            buses.append({
                "substation": sub_name,
                "bus_name": bus.loc_name.strip(),
                "obj": bus,
            })

    return buses


def get_monitored_lines(app, buses):
    bus_objs = [b["obj"] for b in buses]
    lines = []
    seen = set()

    for line in app.GetCalcRelevantObjects("*.ElmLne"):
        try:
            if int(line.GetAttribute("outserv")) == 1:
                continue
        except Exception:
            pass

        try:
            bus1 = line.bus1.cterm
            bus2 = line.bus2.cterm
        except Exception:
            continue

        if bus1 in bus_objs:
            selected_side = "bus1"
            selected_bus = bus1
        elif bus2 in bus_objs:
            selected_side = "bus2"
            selected_bus = bus2
        else:
            continue

        line_key = id(line)
        if line_key in seen:
            continue
        seen.add(line_key)

        lines.append({
            "obj": line,
            "line_name": line.loc_name.strip(),
            "selected_side": selected_side,
            "selected_bus": selected_bus,
        })

    return lines


def restore_couplers(couplers):
    for info in couplers.values():
        obj = info["obj"]
        orig_open = info["orig_open"]
        try:
            current_open = bool(obj.IsOpen())
        except Exception:
            continue

        if current_open != orig_open:
            if orig_open:
                obj.Open()
            else:
                obj.Close()


def apply_open_scenario(couplers, open_names):
    open_set = set(open_names)
    for name, info in couplers.items():
        if name in open_set:
            info["obj"].Open()
        else:
            info["obj"].Close()


def calculate_bus_pipf(buses, lines):
    bus_info = {
        id(b["obj"]): {
            "Substation": b["substation"],
            "Bus": b["bus_name"],
            "PIpf Bus": 0.0,
        }
        for b in buses
    }

    line_rows = []

    for item in lines:
        line = item["obj"]
        selected_side = item["selected_side"]
        selected_bus = item["selected_bus"]
        bus_id = id(selected_bus)

        loading = line.GetAttribute("m:loading")

        if selected_side == "bus1":
            i_actual = line.GetAttribute("m:I1:bus1")
        else:
            i_actual = line.GetAttribute("m:I1:bus2")

        i_limit = line.GetAttribute("c:curnom")

        if loading is None:
            loading = 0.0

        if i_actual is None or i_limit is None or i_limit == 0:
            pipf_i = 0.0
            i_actual = 0.0 if i_actual is None else i_actual
            i_limit = 0.0 if i_limit is None else i_limit
        else:
            pipf_i = (W_I / (2 * N_EXP)) * ((i_actual / i_limit) ** (2 * N_EXP))

        if bus_id in bus_info:
            bus_info[bus_id]["PIpf Bus"] += pipf_i

        line_rows.append({
            "Substation": bus_info.get(bus_id, {}).get("Substation", ""),
            "Bus": bus_info.get(bus_id, {}).get("Bus", ""),
            "Line": item["line_name"],
            "Selected Side": selected_side,
            "Loading (%)": round(float(loading), 6),
            "I actual (kA)": round(float(i_actual), 6),
            "I limit (kA)": round(float(i_limit), 6),
            "PIpf_i": round(float(pipf_i), 8),
        })

    bus_rows = []
    pi_total = 0.0

    for row in bus_info.values():
        row["PIpf Bus"] = round(float(row["PIpf Bus"]), 8)
        pi_total += row["PIpf Bus"]
        bus_rows.append(row)

    return round(float(pi_total), 8), bus_rows, line_rows


def run_bruteforce():
    app = get_app()
    os.makedirs(os.path.dirname(OUTPUT_FILE), exist_ok=True)

    couplers = get_couplers(app)
    coupler_names = [name for name in COUP_LIST if name in couplers]
    buses = get_monitored_buses(app)
    lines = get_monitored_lines(app, buses)

    if not coupler_names:
        raise RuntimeError("No couplers found")
    if not buses:
        raise RuntimeError("No buses found")
    if not lines:
        raise RuntimeError("No lines found")

    summary_rows = []
    bus_detail_rows = []
    line_detail_rows = []

    try:
        scenario_no = 1

        for n_open in OPEN_COUNTS:
            for combo in combinations(coupler_names, n_open):
                open_names = list(combo)
                scenario_label = ", ".join(open_names)

                restore_couplers(couplers)
                apply_open_scenario(couplers, open_names)

                if not run_loadflow(app):
                    summary_rows.append({
                        "No": scenario_no,
                        "N Open": n_open,
                        "Opened BC": scenario_label,
                        "PI Total": None,
                        "Status": "LOADFLOW_FAILED",
                    })
                    scenario_no += 1
                    continue

                pi_total, bus_rows, line_rows = calculate_bus_pipf(buses, lines)

                summary_rows.append({
                    "No": scenario_no,
                    "N Open": n_open,
                    "Opened BC": scenario_label,
                    "PI Total": pi_total,
                    "Status": "OK",
                })

                for row in bus_rows:
                    bus_detail_rows.append({
                        "Scenario": scenario_label,
                        "N Open": n_open,
                        **row,
                    })

                for row in line_rows:
                    line_detail_rows.append({
                        "Scenario": scenario_label,
                        "N Open": n_open,
                        **row,
                    })

                scenario_no += 1

    finally:
        restore_couplers(couplers)

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        pd.DataFrame(summary_rows).to_excel(writer, sheet_name="Summary", index=False)
        pd.DataFrame(bus_detail_rows).to_excel(writer, sheet_name="PIpf per Bus", index=False)
        pd.DataFrame(line_detail_rows).to_excel(writer, sheet_name="Line Detail", index=False)

    app.PrintPlain(f"Saved: {OUTPUT_FILE}")


if __name__ == "__main__":
    run_bruteforce()
```

```python
# file: bruteforce_bc_pipf_fast.py
import os
import time
from itertools import combinations

import powerfactory
import pandas as pd

COUP_LIST = [
    "BC Waru",
    "BC Rungkut",
    "BC Sawahan",
    "BC Gunung Sari",
    "BC Tandes",
    "BC Altaprima",
]

SUBSTAT_LIST = [
    "GRESIK BARU5", "KRIAN5", "GRESIK LAMA55", "KARANGPILANG5", "SAWAHAN5",
    "DARMO GRANDE5", "ALTA PRIMA5", "SEGOROMADU5", "SAMBIKEREP5", "WILMAR5",
    "BAMBE5", "GUNUNG SARI5", "PETRO KIMIA5", "WARU5", "RUNGKUT5",
    "BUDURAN5", "SURABAYA SELATAN5", "SUKOLILO5", "HANIL JAYA STEEL5",
    "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5",
    "SIDOARJO5", "NEWBUDURAN5", "SEDATI5", "SIMPANG5", "GILITIMUR5",
    "NWARU5", "TANDES5", "PERAK5", "UNDAAN5", "KUPANG5",
    "KREMBANGAN5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "KEDINDING5", "BANGKALAN5",
    "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5"
]

OPEN_COUNTS = [1, 2]

W_I = 1.0
N_EXP = 1

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\Hasil Uji"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "BruteForce_BC_PIpf_1BC_2BC.xlsx")

LOG_EACH_SCENARIO = False
SAVE_ALL_LINE_DETAILS = False
SAVE_BEST_LINE_DETAILS = True


def get_app():
    app = powerfactory.GetApplication()
    if not app:
        raise RuntimeError("Run inside DIgSILENT PowerFactory")
    app.PrintPlain("=== BRUTE FORCE PIpf START (FAST MODE) ===")
    return app


def safe_get_attribute(obj, attr, default=None):
    try:
        value = obj.GetAttribute(attr)
        return default if value is None else value
    except Exception:
        return default


def safe_loc_name(obj, default=""):
    if obj is None:
        return default
    try:
        return obj.loc_name.strip()
    except Exception:
        return default


def get_obj_key(obj):
    if obj is None:
        return None

    try:
        return obj.GetFullName()
    except Exception:
        pass

    try:
        return f"{obj.GetClassName()}::{obj.loc_name.strip()}"
    except Exception:
        return str(obj)


def get_parent_substation_name(obj, max_depth=20):
    current = obj
    for _ in range(max_depth):
        if current is None:
            return None
        try:
            if current.GetClassName() == "ElmSubstat":
                return current.loc_name.strip()
            current = current.GetParent()
        except Exception:
            return None
    return None


def resolve_terminal(side_obj):
    if side_obj is None:
        return None

    try:
        if side_obj.GetClassName() == "ElmTerm":
            return side_obj
    except Exception:
        pass

    try:
        terminal = side_obj.cterm
        if terminal is not None:
            return terminal
    except Exception:
        pass

    return None


def run_loadflow(app):
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("ComLdf not found in active study case")

    lf.iopt_net = 0
    lf.iopt_o4 = 0
    return lf.Execute() == 0


def get_line_terminals(line_obj):
    cub1 = None
    cub2 = None

    try:
        cub1 = line_obj.bus1
    except Exception:
        pass

    try:
        cub2 = line_obj.bus2
    except Exception:
        pass

    if cub1 is None:
        cub1 = safe_get_attribute(line_obj, "bus1")
    if cub2 is None:
        cub2 = safe_get_attribute(line_obj, "bus2")

    term1 = resolve_terminal(cub1)
    term2 = resolve_terminal(cub2)
    return term1, term2


def cache_objects(app):
    couplers = {}
    for c in app.GetCalcRelevantObjects("*.ElmCoup"):
        name = safe_loc_name(c)
        if name in COUP_LIST:
            couplers[name] = {
                "obj": c,
                "orig_open": bool(c.IsOpen())
            }

    ordered_coupler_names = [name for name in COUP_LIST if name in couplers]
    app.PrintPlain(f"Found {len(ordered_coupler_names)} couplers from COUP_LIST")

    substations = {
        safe_loc_name(s): s
        for s in app.GetCalcRelevantObjects("*.ElmSubstat")
    }

    monitored_substation_names = set()
    monitored_bus_keys = set()
    buses = []

    for sub_name in SUBSTAT_LIST:
        sub = substations.get(sub_name)
        if sub is None:
            app.PrintPlain(f"Warning: Substation '{sub_name}' not found")
            continue

        monitored_substation_names.add(sub_name)

        all_terms = list(sub.GetContents("*.ElmTerm", 1) or [])
        preferred_terms = []

        for term in all_terms:
            usage = safe_get_attribute(term, "iUsage")
            if usage is None:
                preferred_terms.append(term)
                continue
            try:
                if int(usage) == 0:
                    preferred_terms.append(term)
            except Exception:
                preferred_terms.append(term)

        selected_terms = preferred_terms if preferred_terms else all_terms

        for term in selected_terms:
            term_key = get_obj_key(term)
            if not term_key or term_key in monitored_bus_keys:
                continue

            monitored_bus_keys.add(term_key)
            buses.append({
                "substation": sub_name,
                "bus_name": safe_loc_name(term),
                "obj": term,
                "key": term_key
            })

    app.PrintPlain(f"Total monitored buses from SUBSTAT_LIST = {len(buses)}")

    lines = []
    seen_line_keys = set()

    for line in app.GetCalcRelevantObjects("*.ElmLne"):
        outserv = safe_get_attribute(line, "outserv", 0)
        try:
            if int(outserv) == 1:
                continue
        except Exception:
            pass

        term1, term2 = get_line_terminals(line)

        term1_key = get_obj_key(term1)
        term2_key = get_obj_key(term2)
        term1_sub = get_parent_substation_name(term1)
        term2_sub = get_parent_substation_name(term2)

        term1_match = (
            term1_key in monitored_bus_keys
            or term1_sub in monitored_substation_names
        )
        term2_match = (
            term2_key in monitored_bus_keys
            or term2_sub in monitored_substation_names
        )

        if not (term1_match or term2_match):
            continue

        line_key = get_obj_key(line)
        if line_key in seen_line_keys:
            continue
        seen_line_keys.add(line_key)

        selected_side = "bus1" if term1_match else "bus2"

        lines.append({
            "line_name": safe_loc_name(line),
            "selected_side": selected_side,
            "obj": line,
            "i_limit": float(safe_get_attribute(line, "c:curnom", 0.0) or 0.0),
        })

    app.PrintPlain(f"Total monitored lines = {len(lines)}")
    return couplers, ordered_coupler_names, buses, lines


def restore_couplers(couplers):
    for info in couplers.values():
        obj = info["obj"]
        orig_open = info["orig_open"]

        try:
            current_open = bool(obj.IsOpen())
        except Exception:
            continue

        if current_open != orig_open:
            if orig_open:
                obj.Open()
            else:
                obj.Close()


def apply_open_scenario(couplers, open_names):
    open_set = set(open_names)

    for name, info in couplers.items():
        obj = info["obj"]
        if name in open_set:
            obj.Open()
        else:
            obj.Close()


def calculate_pipf_total(lines, save_details=False):
    total_pipf = 0.0
    details = [] if save_details else None

    for item in lines:
        line_obj = item["obj"]
        selected_side = item["selected_side"]
        i_limit = item["i_limit"]

        loading = safe_get_attribute(line_obj, "m:loading", 0.0)

        if selected_side == "bus1":
            i_actual = safe_get_attribute(line_obj, "m:I1:bus1", 0.0)
        else:
            i_actual = safe_get_attribute(line_obj, "m:I1:bus2", 0.0)

        try:
            loading = float(loading)
        except Exception:
            loading = 0.0

        try:
            i_actual = float(abs(i_actual))
        except Exception:
            i_actual = 0.0

        if i_limit <= 0.0:
            pipf_i = 0.0
        else:
            pipf_i = (W_I / (2 * N_EXP)) * ((i_actual / i_limit) ** (2 * N_EXP))

        total_pipf += pipf_i

        if save_details:
            details.append({
                "Line": item["line_name"],
                "Selected Side": selected_side,
                "Loading (%)": round(loading, 6),
                "I actual (kA)": round(i_actual, 6),
                "I limit (kA)": round(i_limit, 6),
                "PIpf_i": round(float(pipf_i), 8)
            })

    return total_pipf, (details or [])


def evaluate_baseline(app, couplers, lines):
    restore_couplers(couplers)

    t0 = time.perf_counter()
    if not run_loadflow(app):
        raise RuntimeError("Baseline load flow failed")
    lf_time = time.perf_counter() - t0

    t1 = time.perf_counter()
    total_pipf, details = calculate_pipf_total(lines, save_details=True)
    calc_time = time.perf_counter() - t1

    app.PrintPlain(f"Baseline timing | loadflow={lf_time:.2f}s | calc={calc_time:.2f}s")
    return total_pipf, details


def evaluate_open_set(app, couplers, lines, open_names, save_details=False):
    restore_couplers(couplers)

    try:
        apply_open_scenario(couplers, open_names)

        t0 = time.perf_counter()
        if not run_loadflow(app):
            return None, [], "LOADFLOW_FAILED", 0.0, 0.0
        lf_time = time.perf_counter() - t0

        t1 = time.perf_counter()
        total_pipf, details = calculate_pipf_total(lines, save_details=save_details)
        calc_time = time.perf_counter() - t1

        return total_pipf, details, "OK", lf_time, calc_time

    except Exception as exc:
        return None, [], f"ERROR: {exc}", 0.0, 0.0

    finally:
        restore_couplers(couplers)


def run_bruteforce():
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    app = get_app()

    t_cache0 = time.perf_counter()
    couplers, coupler_names, buses, lines = cache_objects(app)
    cache_time = time.perf_counter() - t_cache0
    app.PrintPlain(f"Cache objects time = {cache_time:.2f}s")

    if not couplers:
        raise RuntimeError("No couplers found from COUP_LIST")
    if not coupler_names:
        raise RuntimeError("No ordered coupler names available")
    if not buses:
        raise RuntimeError("No buses found from SUBSTAT_LIST")
    if not lines:
        raise RuntimeError("No lines found connected to SUBSTAT_LIST")

    results = []
    detail_rows = []

    baseline_pipf = None
    best_score = float("inf")
    best_open_names = None
    best_details = []

    total_lf_time = 0.0
    total_calc_time = 0.0
    elapsed = 0.0

    try:
        baseline_pipf, baseline_details = evaluate_baseline(app, couplers, lines)
        app.PrintPlain(f"Baseline PIpf Total = {baseline_pipf:.6f}")

        for row in baseline_details:
            detail_rows.append({
                "Scenario": "BASELINE",
                "N Open": 0,
                "Opened Couplers": "-",
                **row
            })

        t0 = time.perf_counter()
        scenario_no = 1

        for n_open in OPEN_COUNTS:
            if n_open > len(coupler_names):
                app.PrintPlain(f"Skip n_open={n_open}; available couplers only {len(coupler_names)}")
                continue

            app.PrintPlain(f"===== START EVALUATION: OPEN {n_open} BC =====")

            for combo in combinations(coupler_names, n_open):
                open_names = list(combo)
                scenario_label = ", ".join(open_names)

                if LOG_EACH_SCENARIO:
                    app.PrintPlain(f"[{scenario_no}] OPEN ({n_open} BC): {scenario_label}")

                save_details_now = SAVE_ALL_LINE_DETAILS

                score, details, status, lf_time, calc_time = evaluate_open_set(
                    app,
                    couplers,
                    lines,
                    open_names,
                    save_details=save_details_now
                )

                total_lf_time += lf_time
                total_calc_time += calc_time

                if score is None:
                    results.append({
                        "No": scenario_no,
                        "N Open": n_open,
                        "Opened Couplers": scenario_label,
                        "PIpf Total": None,
                        "Delta vs Baseline": None,
                        "Status": status,
                        "LF Time (s)": round(lf_time, 3),
                        "Calc Time (s)": round(calc_time, 3),
                    })
                    scenario_no += 1
                    continue

                delta = score - baseline_pipf

                results.append({
                    "No": scenario_no,
                    "N Open": n_open,
                    "Opened Couplers": scenario_label,
                    "PIpf Total": round(score, 8),
                    "Delta vs Baseline": round(delta, 8),
                    "Status": status,
                    "LF Time (s)": round(lf_time, 3),
                    "Calc Time (s)": round(calc_time, 3),
                })

                if SAVE_ALL_LINE_DETAILS:
                    for row in details:
                        detail_rows.append({
                            "Scenario": f"OPEN {scenario_label}",
                            "N Open": n_open,
                            "Opened Couplers": scenario_label,
                            **row
                        })

                if score < best_score:
                    best_score = score
                    best_open_names = open_names

                    if SAVE_BEST_LINE_DETAILS:
                        _, best_details = calculate_pipf_total(lines, save_details=True)

                    app.PrintPlain(
                        f"NEW BEST | scenario={scenario_no} | score={best_score:.6f} | "
                        f"lf={lf_time:.2f}s | calc={calc_time:.2f}s"
                    )

                scenario_no += 1

        elapsed = time.perf_counter() - t0

    finally:
        restore_couplers(couplers)

    if SAVE_BEST_LINE_DETAILS and best_open_names and not SAVE_ALL_LINE_DETAILS:
        best_label = ", ".join(best_open_names)
        for row in best_details:
            detail_rows.append({
                "Scenario": f"BEST OPEN {best_label}",
                "N Open": len(best_open_names),
                "Opened Couplers": best_label,
                **row
            })

    t_write0 = time.perf_counter()

    df = pd.DataFrame(results)
    df_detail = pd.DataFrame(detail_rows)

    if not df.empty and "PIpf Total" in df.columns:
        df_rank = (
            df[df["PIpf Total"].notna()]
            .sort_values(by="PIpf Total", ascending=True)
            .reset_index(drop=True)
        )
    else:
        df_rank = pd.DataFrame()

    df_best = pd.DataFrame([{
        "Best Opened Couplers": ", ".join(best_open_names) if best_open_names else None,
        "Best N Open": len(best_open_names) if best_open_names else None,
        "Best PIpf Total": round(best_score, 8) if best_open_names else None,
        "Baseline PIpf Total": round(baseline_pipf, 8) if baseline_pipf is not None else None,
        "Execution Time (s)": round(elapsed, 2),
        "Cache Time (s)": round(cache_time, 2),
        "Total Loadflow Time (s)": round(total_lf_time, 2),
        "Total PIpf Calc Time (s)": round(total_calc_time, 2),
    }])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df.to_excel(writer, sheet_name="All Scenario", index=False)
        df_rank.to_excel(writer, sheet_name="Ranking", index=False)
        df_best.to_excel(writer, sheet_name="Best", index=False)
        if not df_detail.empty:
            df_detail.to_excel(writer, sheet_name="Line Detail", index=False)

    write_time = time.perf_counter() - t_write0

    app.PrintPlain("===== FINAL RESULT =====")
    app.PrintPlain(f"Baseline PIpf Total   : {baseline_pipf:.6f}" if baseline_pipf is not None else "Baseline PIpf Total   : N/A")
    app.PrintPlain(f"Best Opened Couplers  : {', '.join(best_open_names)}" if best_open_names else "Best Opened Couplers  : N/A")
    app.PrintPlain(f"Best N Open           : {len(best_open_names)}" if best_open_names else "Best N Open           : N/A")
    app.PrintPlain(f"Best PIpf Total       : {best_score:.6f}" if best_open_names else "Best PIpf Total       : N/A")
    app.PrintPlain(f"Cache Time            : {cache_time:.2f} s")
    app.PrintPlain(f"Total Loadflow Time   : {total_lf_time:.2f} s")
    app.PrintPlain(f"Total PIpf Calc Time  : {total_calc_time:.2f} s")
    app.PrintPlain(f"Scenario Loop Time    : {elapsed:.2f} s")
    app.PrintPlain(f"Excel Write Time      : {write_time:.2f} s")
    app.PrintPlain(f"Saved                 : {OUTPUT_FILE}")


if __name__ == "__main__":
    run_bruteforce()
```

```python
# file: bruteforce_bc_pipf_fast.py
import os
import time
from itertools import combinations

import powerfactory
import pandas as pd

COUP_LIST = [
    "BC Waru",
    "BC Rungkut",
    "BC Sawahan",
    "BC Gunung Sari",
    "BC Tandes",
    "BC Altaprima",
]

SUBSTAT_LIST = [
    "GRESIK BARU5", "KRIAN5", "GRESIK LAMA55", "KARANGPILANG5", "SAWAHAN5",
    "DARMO GRANDE5", "ALTA PRIMA5", "SEGOROMADU5", "SAMBIKEREP5", "WILMAR5",
    "BAMBE5", "GUNUNG SARI5", "PETRO KIMIA5", "WARU5", "RUNGKUT5",
    "BUDURAN5", "SURABAYA SELATAN5", "SUKOLILO5", "HANIL JAYA STEEL5",
    "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5",
    "SIDOARJO5", "NEWBUDURAN5", "SEDATI5", "SIMPANG5", "GILITIMUR5",
    "NWARU5", "TANDES5", "PERAK5", "UNDAAN5", "KUPANG5",
    "KREMBANGAN5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "KEDINDING5", "BANGKALAN5",
    "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5"
]

OPEN_COUNTS = [1, 2]

W_I = 1.0
N_EXP = 1

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\Hasil Uji"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "BruteForce_BC_PIpf_1BC_2BC.xlsx")

LOG_EACH_SCENARIO = False
SAVE_ALL_LINE_DETAILS = False
SAVE_BEST_LINE_DETAILS = True


def get_app():
    app = powerfactory.GetApplication()
    if not app:
        raise RuntimeError("Run inside DIgSILENT PowerFactory")
    app.PrintPlain("=== BRUTE FORCE PIpf START (FAST MODE) ===")
    return app


def safe_get_attribute(obj, attr, default=None):
    try:
        value = obj.GetAttribute(attr)
        return default if value is None else value
    except Exception:
        return default


def safe_loc_name(obj, default=""):
    if obj is None:
        return default
    try:
        return obj.loc_name.strip()
    except Exception:
        return default


def get_obj_key(obj):
    if obj is None:
        return None

    try:
        return obj.GetFullName()
    except Exception:
        pass

    try:
        return f"{obj.GetClassName()}::{obj.loc_name.strip()}"
    except Exception:
        return str(obj)


def get_parent_substation_name(obj, max_depth=20):
    current = obj
    for _ in range(max_depth):
        if current is None:
            return None
        try:
            if current.GetClassName() == "ElmSubstat":
                return current.loc_name.strip()
            current = current.GetParent()
        except Exception:
            return None
    return None


def resolve_terminal(side_obj):
    if side_obj is None:
        return None

    try:
        if side_obj.GetClassName() == "ElmTerm":
            return side_obj
    except Exception:
        pass

    try:
        terminal = side_obj.cterm
        if terminal is not None:
            return terminal
    except Exception:
        pass

    return None


def run_loadflow(app):
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("ComLdf not found in active study case")

    lf.iopt_net = 0
    lf.iopt_o4 = 0
    return lf.Execute() == 0


def get_line_terminals(line_obj):
    cub1 = None
    cub2 = None

    try:
        cub1 = line_obj.bus1
    except Exception:
        pass

    try:
        cub2 = line_obj.bus2
    except Exception:
        pass

    if cub1 is None:
        cub1 = safe_get_attribute(line_obj, "bus1")
    if cub2 is None:
        cub2 = safe_get_attribute(line_obj, "bus2")

    term1 = resolve_terminal(cub1)
    term2 = resolve_terminal(cub2)
    return term1, term2


def cache_objects(app):
    couplers = {}
    for c in app.GetCalcRelevantObjects("*.ElmCoup"):
        name = safe_loc_name(c)
        if name in COUP_LIST:
            couplers[name] = {
                "obj": c,
                "orig_open": bool(c.IsOpen())
            }

    ordered_coupler_names = [name for name in COUP_LIST if name in couplers]
    app.PrintPlain(f"Found {len(ordered_coupler_names)} couplers from COUP_LIST")

    substations = {
        safe_loc_name(s): s
        for s in app.GetCalcRelevantObjects("*.ElmSubstat")
    }

    monitored_substation_names = set()
    monitored_bus_keys = set()
    buses = []

    for sub_name in SUBSTAT_LIST:
        sub = substations.get(sub_name)
        if sub is None:
            app.PrintPlain(f"Warning: Substation '{sub_name}' not found")
            continue

        monitored_substation_names.add(sub_name)

        all_terms = list(sub.GetContents("*.ElmTerm", 1) or [])
        preferred_terms = []

        for term in all_terms:
            usage = safe_get_attribute(term, "iUsage")
            if usage is None:
                preferred_terms.append(term)
                continue
            try:
                if int(usage) == 0:
                    preferred_terms.append(term)
            except Exception:
                preferred_terms.append(term)

        selected_terms = preferred_terms if preferred_terms else all_terms

        for term in selected_terms:
            term_key = get_obj_key(term)
            if not term_key or term_key in monitored_bus_keys:
                continue

            monitored_bus_keys.add(term_key)
            buses.append({
                "substation": sub_name,
                "bus_name": safe_loc_name(term),
                "obj": term,
                "key": term_key
            })

    app.PrintPlain(f"Total monitored buses from SUBSTAT_LIST = {len(buses)}")

    lines = []
    seen_line_keys = set()

    for line in app.GetCalcRelevantObjects("*.ElmLne"):
        outserv = safe_get_attribute(line, "outserv", 0)
        try:
            if int(outserv) == 1:
                continue
        except Exception:
            pass

        term1, term2 = get_line_terminals(line)

        term1_key = get_obj_key(term1)
        term2_key = get_obj_key(term2)
        term1_sub = get_parent_substation_name(term1)
        term2_sub = get_parent_substation_name(term2)

        term1_match = (
            term1_key in monitored_bus_keys
            or term1_sub in monitored_substation_names
        )
        term2_match = (
            term2_key in monitored_bus_keys
            or term2_sub in monitored_substation_names
        )

        if not (term1_match or term2_match):
            continue

        line_key = get_obj_key(line)
        if line_key in seen_line_keys:
            continue
        seen_line_keys.add(line_key)

        selected_side = "bus1" if term1_match else "bus2"

        lines.append({
            "line_name": safe_loc_name(line),
            "selected_side": selected_side,
            "obj": line,
            "i_limit": float(safe_get_attribute(line, "c:curnom", 0.0) or 0.0),
        })

    app.PrintPlain(f"Total monitored lines = {len(lines)}")
    return couplers, ordered_coupler_names, buses, lines


def restore_couplers(couplers):
    for info in couplers.values():
        obj = info["obj"]
        orig_open = info["orig_open"]

        try:
            current_open = bool(obj.IsOpen())
        except Exception:
            continue

        if current_open != orig_open:
            if orig_open:
                obj.Open()
            else:
                obj.Close()


def apply_open_scenario(couplers, open_names):
    open_set = set(open_names)

    for name, info in couplers.items():
        obj = info["obj"]
        if name in open_set:
            obj.Open()
        else:
            obj.Close()


def calculate_pipf_total(lines, save_details=False):
    total_pipf = 0.0
    details = [] if save_details else None

    for item in lines:
        line_obj = item["obj"]
        selected_side = item["selected_side"]
        i_limit = item["i_limit"]

        loading = safe_get_attribute(line_obj, "m:loading", 0.0)

        if selected_side == "bus1":
            i_actual = safe_get_attribute(line_obj, "m:I1:bus1", 0.0)
        else:
            i_actual = safe_get_attribute(line_obj, "m:I1:bus2", 0.0)

        try:
            loading = float(loading)
        except Exception:
            loading = 0.0

        try:
            i_actual = float(abs(i_actual))
        except Exception:
            i_actual = 0.0

        if i_limit <= 0.0:
            pipf_i = 0.0
        else:
            pipf_i = (W_I / (2 * N_EXP)) * ((i_actual / i_limit) ** (2 * N_EXP))

        total_pipf += pipf_i

        if save_details:
            details.append({
                "Line": item["line_name"],
                "Selected Side": selected_side,
                "Loading (%)": round(loading, 6),
                "I actual (kA)": round(i_actual, 6),
                "I limit (kA)": round(i_limit, 6),
                "PIpf_i": round(float(pipf_i), 8)
            })

    return total_pipf, (details or [])


def evaluate_baseline(app, couplers, lines):
    restore_couplers(couplers)

    t0 = time.perf_counter()
    if not run_loadflow(app):
        raise RuntimeError("Baseline load flow failed")
    lf_time = time.perf_counter() - t0

    t1 = time.perf_counter()
    total_pipf, details = calculate_pipf_total(lines, save_details=True)
    calc_time = time.perf_counter() - t1

    app.PrintPlain(f"Baseline timing | loadflow={lf_time:.2f}s | calc={calc_time:.2f}s")
    return total_pipf, details


def evaluate_open_set(app, couplers, lines, open_names, save_details=False):
    restore_couplers(couplers)

    try:
        apply_open_scenario(couplers, open_names)

        t0 = time.perf_counter()
        if not run_loadflow(app):
            return None, [], "LOADFLOW_FAILED", 0.0, 0.0
        lf_time = time.perf_counter() - t0

        t1 = time.perf_counter()
        total_pipf, details = calculate_pipf_total(lines, save_details=save_details)
        calc_time = time.perf_counter() - t1

        return total_pipf, details, "OK", lf_time, calc_time

    except Exception as exc:
        return None, [], f"ERROR: {exc}", 0.0, 0.0

    finally:
        restore_couplers(couplers)


def run_bruteforce():
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    app = get_app()

    t_cache0 = time.perf_counter()
    couplers, coupler_names, buses, lines = cache_objects(app)
    cache_time = time.perf_counter() - t_cache0
    app.PrintPlain(f"Cache objects time = {cache_time:.2f}s")

    if not couplers:
        raise RuntimeError("No couplers found from COUP_LIST")
    if not coupler_names:
        raise RuntimeError("No ordered coupler names available")
    if not buses:
        raise RuntimeError("No buses found from SUBSTAT_LIST")
    if not lines:
        raise RuntimeError("No lines found connected to SUBSTAT_LIST")

    results = []
    detail_rows = []

    baseline_pipf = None
    best_score = float("inf")
    best_open_names = None
    best_details = []

    total_lf_time = 0.0
    total_calc_time = 0.0
    elapsed = 0.0

    try:
        baseline_pipf, baseline_details = evaluate_baseline(app, couplers, lines)
        app.PrintPlain(f"Baseline PIpf Total = {baseline_pipf:.6f}")

        for row in baseline_details:
            detail_rows.append({
                "Scenario": "BASELINE",
                "N Open": 0,
                "Opened Couplers": "-",
                **row
            })

        t0 = time.perf_counter()
        scenario_no = 1

        for n_open in OPEN_COUNTS:
            if n_open > len(coupler_names):
                app.PrintPlain(f"Skip n_open={n_open}; available couplers only {len(coupler_names)}")
                continue

            app.PrintPlain(f"===== START EVALUATION: OPEN {n_open} BC =====")

            for combo in combinations(coupler_names, n_open):
                open_names = list(combo)
                scenario_label = ", ".join(open_names)

                if LOG_EACH_SCENARIO:
                    app.PrintPlain(f"[{scenario_no}] OPEN ({n_open} BC): {scenario_label}")

                save_details_now = SAVE_ALL_LINE_DETAILS

                score, details, status, lf_time, calc_time = evaluate_open_set(
                    app,
                    couplers,
                    lines,
                    open_names,
                    save_details=save_details_now
                )

                total_lf_time += lf_time
                total_calc_time += calc_time

                if score is None:
                    results.append({
                        "No": scenario_no,
                        "N Open": n_open,
                        "Opened Couplers": scenario_label,
                        "PIpf Total": None,
                        "Delta vs Baseline": None,
                        "Status": status,
                        "LF Time (s)": round(lf_time, 3),
                        "Calc Time (s)": round(calc_time, 3),
                    })
                    scenario_no += 1
                    continue

                delta = score - baseline_pipf

                results.append({
                    "No": scenario_no,
                    "N Open": n_open,
                    "Opened Couplers": scenario_label,
                    "PIpf Total": round(score, 8),
                    "Delta vs Baseline": round(delta, 8),
                    "Status": status,
                    "LF Time (s)": round(lf_time, 3),
                    "Calc Time (s)": round(calc_time, 3),
                })

                if SAVE_ALL_LINE_DETAILS:
                    for row in details:
                        detail_rows.append({
                            "Scenario": f"OPEN {scenario_label}",
                            "N Open": n_open,
                            "Opened Couplers": scenario_label,
                            **row
                        })

                if score < best_score:
                    best_score = score
                    best_open_names = open_names

                    if SAVE_BEST_LINE_DETAILS:
                        _, best_details = calculate_pipf_total(lines, save_details=True)

                    app.PrintPlain(
                        f"NEW BEST | scenario={scenario_no} | score={best_score:.6f} | "
                        f"lf={lf_time:.2f}s | calc={calc_time:.2f}s"
                    )

                scenario_no += 1

        elapsed = time.perf_counter() - t0

    finally:
        restore_couplers(couplers)

    if SAVE_BEST_LINE_DETAILS and best_open_names and not SAVE_ALL_LINE_DETAILS:
        best_label = ", ".join(best_open_names)
        for row in best_details:
            detail_rows.append({
                "Scenario": f"BEST OPEN {best_label}",
                "N Open": len(best_open_names),
                "Opened Couplers": best_label,
                **row
            })

    t_write0 = time.perf_counter()

    df = pd.DataFrame(results)
    df_detail = pd.DataFrame(detail_rows)

    if not df.empty and "PIpf Total" in df.columns:
        df_rank = (
            df[df["PIpf Total"].notna()]
            .sort_values(by="PIpf Total", ascending=True)
            .reset_index(drop=True)
        )
    else:
        df_rank = pd.DataFrame()

    df_best = pd.DataFrame([{
        "Best Opened Couplers": ", ".join(best_open_names) if best_open_names else None,
        "Best N Open": len(best_open_names) if best_open_names else None,
        "Best PIpf Total": round(best_score, 8) if best_open_names else None,
        "Baseline PIpf Total": round(baseline_pipf, 8) if baseline_pipf is not None else None,
        "Execution Time (s)": round(elapsed, 2),
        "Cache Time (s)": round(cache_time, 2),
        "Total Loadflow Time (s)": round(total_lf_time, 2),
        "Total PIpf Calc Time (s)": round(total_calc_time, 2),
    }])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df.to_excel(writer, sheet_name="All Scenario", index=False)
        df_rank.to_excel(writer, sheet_name="Ranking", index=False)
        df_best.to_excel(writer, sheet_name="Best", index=False)
        if not df_detail.empty:
            df_detail.to_excel(writer, sheet_name="Line Detail", index=False)

    write_time = time.perf_counter() - t_write0

    app.PrintPlain("===== FINAL RESULT =====")
    app.PrintPlain(f"Baseline PIpf Total   : {baseline_pipf:.6f}" if baseline_pipf is not None else "Baseline PIpf Total   : N/A")
    app.PrintPlain(f"Best Opened Couplers  : {', '.join(best_open_names)}" if best_open_names else "Best Opened Couplers  : N/A")
    app.PrintPlain(f"Best N Open           : {len(best_open_names)}" if best_open_names else "Best N Open           : N/A")
    app.PrintPlain(f"Best PIpf Total       : {best_score:.6f}" if best_open_names else "Best PIpf Total       : N/A")
    app.PrintPlain(f"Cache Time            : {cache_time:.2f} s")
    app.PrintPlain(f"Total Loadflow Time   : {total_lf_time:.2f} s")
    app.PrintPlain(f"Total PIpf Calc Time  : {total_calc_time:.2f} s")
    app.PrintPlain(f"Scenario Loop Time    : {elapsed:.2f} s")
    app.PrintPlain(f"Excel Write Time      : {write_time:.2f} s")
    app.PrintPlain(f"Saved                 : {OUTPUT_FILE}")


if __name__ == "__main__":
    run_bruteforce()
```

```python
# file: bruteforce_bc_pipf_fast_fixed.py
import os
import time
from itertools import combinations

import powerfactory
import pandas as pd

COUP_LIST = [
    "BC Waru",
    "BC Rungkut",
    "BC Sawahan",
    "BC Gunung Sari",
    "BC Tandes",
    "BC Altaprima",
]

SUBSTAT_LIST = [
    "GRESIK BARU5", "KRIAN5", "GRESIK LAMA55", "KARANGPILANG5", "SAWAHAN5",
    "DARMO GRANDE5", "ALTA PRIMA5", "SEGOROMADU5", "SAMBIKEREP5", "WILMAR5",
    "BAMBE5", "GUNUNG SARI5", "PETRO KIMIA5", "WARU5", "RUNGKUT5",
    "BUDURAN5", "SURABAYA SELATAN5", "SUKOLILO5", "HANIL JAYA STEEL5",
    "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5",
    "SIDOARJO5", "NEWBUDURAN5", "SEDATI5", "SIMPANG5", "GILITIMUR5",
    "NWARU5", "TANDES5", "PERAK5", "UNDAAN5", "KUPANG5",
    "KREMBANGAN5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "KEDINDING5", "BANGKALAN5",
    "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5"
]

OPEN_COUNTS = [1, 2]

W_I = 1.0
N_EXP = 1

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\Hasil Uji"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "BruteForce_BC_PIpf_1BC_2BC.xlsx")

LOG_EACH_SCENARIO = False
SAVE_ALL_LINE_DETAILS = False
SAVE_BEST_LINE_DETAILS = True


def get_app():
    app = powerfactory.GetApplication()
    if not app:
        raise RuntimeError("Run inside DIgSILENT PowerFactory")
    app.PrintPlain("=== BRUTE FORCE PIpf START (FAST MODE) ===")
    return app


def safe_get_attribute(obj, attr, default=None):
    try:
        value = obj.GetAttribute(attr)
        return default if value is None else value
    except Exception:
        return default


def safe_loc_name(obj, default=""):
    if obj is None:
        return default
    try:
        return obj.loc_name.strip()
    except Exception:
        return default


def get_obj_key(obj):
    if obj is None:
        return None

    try:
        return obj.GetFullName()
    except Exception:
        pass

    try:
        return f"{obj.GetClassName()}::{obj.loc_name.strip()}"
    except Exception:
        return str(obj)


def get_parent_substation_name(obj, max_depth=20):
    current = obj
    for _ in range(max_depth):
        if current is None:
            return None
        try:
            if current.GetClassName() == "ElmSubstat":
                return current.loc_name.strip()
            current = current.GetParent()
        except Exception:
            return None
    return None


def resolve_terminal(side_obj):
    if side_obj is None:
        return None

    try:
        if side_obj.GetClassName() == "ElmTerm":
            return side_obj
    except Exception:
        pass

    try:
        terminal = side_obj.cterm
        if terminal is not None:
            return terminal
    except Exception:
        pass

    return None


def run_loadflow(app):
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("ComLdf not found in active study case")

    lf.iopt_net = 0
    lf.iopt_o4 = 0
    return lf.Execute() == 0


def get_line_terminals(line_obj):
    cub1 = None
    cub2 = None

    try:
        cub1 = line_obj.bus1
    except Exception:
        pass

    try:
        cub2 = line_obj.bus2
    except Exception:
        pass

    if cub1 is None:
        cub1 = safe_get_attribute(line_obj, "bus1")
    if cub2 is None:
        cub2 = safe_get_attribute(line_obj, "bus2")

    term1 = resolve_terminal(cub1)
    term2 = resolve_terminal(cub2)
    return term1, term2


def cache_objects(app):
    couplers = {}
    for c in app.GetCalcRelevantObjects("*.ElmCoup"):
        name = safe_loc_name(c)
        if name in COUP_LIST:
            couplers[name] = {
                "obj": c,
                "orig_open": bool(c.IsOpen())
            }

    ordered_coupler_names = [name for name in COUP_LIST if name in couplers]
    app.PrintPlain(f"Found {len(ordered_coupler_names)} couplers from COUP_LIST")

    substations = {
        safe_loc_name(s): s
        for s in app.GetCalcRelevantObjects("*.ElmSubstat")
    }

    monitored_substation_names = set()
    monitored_bus_keys = set()
    buses = []

    for sub_name in SUBSTAT_LIST:
        sub = substations.get(sub_name)
        if sub is None:
            app.PrintPlain(f"Warning: Substation '{sub_name}' not found")
            continue

        monitored_substation_names.add(sub_name)

        all_terms = list(sub.GetContents("*.ElmTerm", 1) or [])
        preferred_terms = []

        for term in all_terms:
            usage = safe_get_attribute(term, "iUsage")
            if usage is None:
                preferred_terms.append(term)
                continue
            try:
                if int(usage) == 0:
                    preferred_terms.append(term)
            except Exception:
                preferred_terms.append(term)

        selected_terms = preferred_terms if preferred_terms else all_terms

        for term in selected_terms:
            term_key = get_obj_key(term)
            if not term_key or term_key in monitored_bus_keys:
                continue

            monitored_bus_keys.add(term_key)
            buses.append({
                "substation": sub_name,
                "bus_name": safe_loc_name(term),
                "obj": term,
                "key": term_key
            })

    app.PrintPlain(f"Total monitored buses from SUBSTAT_LIST = {len(buses)}")

    lines = []
    seen_line_keys = set()

    for line in app.GetCalcRelevantObjects("*.ElmLne"):
        outserv = safe_get_attribute(line, "outserv", 0)
        try:
            if int(outserv) == 1:
                continue
        except Exception:
            pass

        term1, term2 = get_line_terminals(line)

        term1_key = get_obj_key(term1)
        term2_key = get_obj_key(term2)
        term1_sub = get_parent_substation_name(term1)
        term2_sub = get_parent_substation_name(term2)

        term1_match = (
            term1_key in monitored_bus_keys
            or term1_sub in monitored_substation_names
        )
        term2_match = (
            term2_key in monitored_bus_keys
            or term2_sub in monitored_substation_names
        )

        if not (term1_match or term2_match):
            continue

        line_key = get_obj_key(line)
        if line_key in seen_line_keys:
            continue
        seen_line_keys.add(line_key)

        selected_side = "bus1" if term1_match else "bus2"

        lines.append({
            "line_name": safe_loc_name(line),
            "selected_side": selected_side,
            "obj": line,
        })

    app.PrintPlain(f"Total monitored lines = {len(lines)}")
    return couplers, ordered_coupler_names, buses, lines


def restore_couplers(couplers):
    for info in couplers.values():
        obj = info["obj"]
        orig_open = info["orig_open"]

        try:
            current_open = bool(obj.IsOpen())
        except Exception:
            continue

        if current_open != orig_open:
            if orig_open:
                obj.Open()
            else:
                obj.Close()


def apply_open_scenario(couplers, open_names):
    open_set = set(open_names)

    for name, info in couplers.items():
        obj = info["obj"]
        if name in open_set:
            obj.Open()
        else:
            obj.Close()


def calculate_pipf_total(lines, save_details=False):
    total_pipf = 0.0
    details = [] if save_details else []

    for item in lines:
        line_obj = item["obj"]
        selected_side = item["selected_side"]

        loading = safe_get_attribute(line_obj, "m:loading", 0.0)

        if selected_side == "bus1":
            i_actual = safe_get_attribute(line_obj, "m:I1:bus1", 0.0)
        else:
            i_actual = safe_get_attribute(line_obj, "m:I1:bus2", 0.0)

        i_limit = safe_get_attribute(line_obj, "c:curnom", 0.0)

        try:
            loading = float(loading)
        except Exception:
            loading = 0.0

        try:
            i_actual = float(abs(i_actual))
        except Exception:
            i_actual = 0.0

        try:
            i_limit = float(i_limit)
        except Exception:
            i_limit = 0.0

        if i_limit <= 0.0:
            pipf_i = 0.0
        else:
            pipf_i = (W_I / (2 * N_EXP)) * ((i_actual / i_limit) ** (2 * N_EXP))

        total_pipf += pipf_i

        if save_details:
            details.append({
                "Line": item["line_name"],
                "Selected Side": selected_side,
                "Loading (%)": round(loading, 6),
                "I actual (kA)": round(i_actual, 6),
                "I limit (kA)": round(i_limit, 6),
                "PIpf_i": round(float(pipf_i), 8)
            })

    return total_pipf, details


def evaluate_baseline(app, couplers, lines):
    restore_couplers(couplers)

    t0 = time.perf_counter()
    if not run_loadflow(app):
        raise RuntimeError("Baseline load flow failed")
    lf_time = time.perf_counter() - t0

    t1 = time.perf_counter()
    total_pipf, details = calculate_pipf_total(lines, save_details=True)
    calc_time = time.perf_counter() - t1

    app.PrintPlain(f"Baseline timing | loadflow={lf_time:.2f}s | calc={calc_time:.2f}s")
    return total_pipf, details


def evaluate_open_set(app, couplers, lines, open_names, save_details=False):
    restore_couplers(couplers)

    try:
        apply_open_scenario(couplers, open_names)

        t0 = time.perf_counter()
        if not run_loadflow(app):
            return None, [], "LOADFLOW_FAILED", 0.0, 0.0
        lf_time = time.perf_counter() - t0

        t1 = time.perf_counter()
        total_pipf, details = calculate_pipf_total(lines, save_details=save_details)
        calc_time = time.perf_counter() - t1

        return total_pipf, details, "OK", lf_time, calc_time

    except Exception as exc:
        return None, [], f"ERROR: {exc}", 0.0, 0.0

    finally:
        restore_couplers(couplers)


def run_bruteforce():
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    app = get_app()

    t_cache0 = time.perf_counter()
    couplers, coupler_names, buses, lines = cache_objects(app)
    cache_time = time.perf_counter() - t_cache0
    app.PrintPlain(f"Cache objects time = {cache_time:.2f}s")

    if not couplers:
        raise RuntimeError("No couplers found from COUP_LIST")
    if not coupler_names:
        raise RuntimeError("No ordered coupler names available")
    if not buses:
        raise RuntimeError("No buses found from SUBSTAT_LIST")
    if not lines:
        raise RuntimeError("No lines found connected to SUBSTAT_LIST")

    results = []
    detail_rows = []

    baseline_pipf = None
    best_score = float("inf")
    best_open_names = None
    best_details = []

    total_lf_time = 0.0
    total_calc_time = 0.0
    elapsed = 0.0

    try:
        baseline_pipf, baseline_details = evaluate_baseline(app, couplers, lines)
        app.PrintPlain(f"Baseline PIpf Total = {baseline_pipf:.6f}")

        for row in baseline_details:
            detail_rows.append({
                "Scenario": "BASELINE",
                "N Open": 0,
                "Opened Couplers": "-",
                **row
            })

        t0 = time.perf_counter()
        scenario_no = 1

        for n_open in OPEN_COUNTS:
            if n_open > len(coupler_names):
                app.PrintPlain(f"Skip n_open={n_open}; available couplers only {len(coupler_names)}")
                continue

            app.PrintPlain(f"===== START EVALUATION: OPEN {n_open} BC =====")

            for combo in combinations(coupler_names, n_open):
                open_names = list(combo)
                scenario_label = ", ".join(open_names)

                if LOG_EACH_SCENARIO:
                    app.PrintPlain(f"[{scenario_no}] OPEN ({n_open} BC): {scenario_label}")

                save_details_now = SAVE_ALL_LINE_DETAILS or SAVE_BEST_LINE_DETAILS

                score, details, status, lf_time, calc_time = evaluate_open_set(
                    app,
                    couplers,
                    lines,
                    open_names,
                    save_details=save_details_now
                )

                total_lf_time += lf_time
                total_calc_time += calc_time

                if score is None:
                    results.append({
                        "No": scenario_no,
                        "N Open": n_open,
                        "Opened Couplers": scenario_label,
                        "PIpf Total": None,
                        "Delta vs Baseline": None,
                        "Status": status,
                        "LF Time (s)": round(lf_time, 3),
                        "Calc Time (s)": round(calc_time, 3),
                    })
                    scenario_no += 1
                    continue

                delta = score - baseline_pipf

                results.append({
                    "No": scenario_no,
                    "N Open": n_open,
                    "Opened Couplers": scenario_label,
                    "PIpf Total": round(score, 8),
                    "Delta vs Baseline": round(delta, 8),
                    "Status": status,
                    "LF Time (s)": round(lf_time, 3),
                    "Calc Time (s)": round(calc_time, 3),
                })

                if SAVE_ALL_LINE_DETAILS:
                    for row in details:
                        detail_rows.append({
                            "Scenario": f"OPEN {scenario_label}",
                            "N Open": n_open,
                            "Opened Couplers": scenario_label,
                            **row
                        })

                if score < best_score:
                    best_score = score
                    best_open_names = open_names
                    best_details = list(details)

                    app.PrintPlain(
                        f"NEW BEST | scenario={scenario_no} | score={best_score:.6f} | "
                        f"lf={lf_time:.2f}s | calc={calc_time:.2f}s"
                    )

                scenario_no += 1

        elapsed = time.perf_counter() - t0

    finally:
        restore_couplers(couplers)

    if SAVE_BEST_LINE_DETAILS and best_open_names and not SAVE_ALL_LINE_DETAILS:
        best_label = ", ".join(best_open_names)
        for row in best_details:
            detail_rows.append({
                "Scenario": f"BEST OPEN {best_label}",
                "N Open": len(best_open_names),
                "Opened Couplers": best_label,
                **row
            })

    t_write0 = time.perf_counter()

    df = pd.DataFrame(results)
    df_detail = pd.DataFrame(detail_rows)

    if not df.empty and "PIpf Total" in df.columns:
        df_rank = (
            df[df["PIpf Total"].notna()]
            .sort_values(by="PIpf Total", ascending=True)
            .reset_index(drop=True)
        )
    else:
        df_rank = pd.DataFrame()

    df_best = pd.DataFrame([{
        "Best Opened Couplers": ", ".join(best_open_names) if best_open_names else None,
        "Best N Open": len(best_open_names) if best_open_names else None,
        "Best PIpf Total": round(best_score, 8) if best_open_names else None,
        "Baseline PIpf Total": round(baseline_pipf, 8) if baseline_pipf is not None else None,
        "Execution Time (s)": round(elapsed, 2),
        "Cache Time (s)": round(cache_time, 2),
        "Total Loadflow Time (s)": round(total_lf_time, 2),
        "Total PIpf Calc Time (s)": round(total_calc_time, 2),
    }])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df.to_excel(writer, sheet_name="All Scenario", index=False)
        df_rank.to_excel(writer, sheet_name="Ranking", index=False)
        df_best.to_excel(writer, sheet_name="Best", index=False)
        if not df_detail.empty:
            df_detail.to_excel(writer, sheet_name="Line Detail", index=False)

    write_time = time.perf_counter() - t_write0

    app.PrintPlain("===== FINAL RESULT =====")
    app.PrintPlain(f"Baseline PIpf Total   : {baseline_pipf:.6f}" if baseline_pipf is not None else "Baseline PIpf Total   : N/A")
    app.PrintPlain(f"Best Opened Couplers  : {', '.join(best_open_names)}" if best_open_names else "Best Opened Couplers  : N/A")
    app.PrintPlain(f"Best N Open           : {len(best_open_names)}" if best_open_names else "Best N Open           : N/A")
    app.PrintPlain(f"Best PIpf Total       : {best_score:.6f}" if best_open_names else "Best PIpf Total       : N/A")
    app.PrintPlain(f"Cache Time            : {cache_time:.2f} s")
    app.PrintPlain(f"Total Loadflow Time   : {total_lf_time:.2f} s")
    app.PrintPlain(f"Total PIpf Calc Time  : {total_calc_time:.2f} s")
    app.PrintPlain(f"Scenario Loop Time    : {elapsed:.2f} s")
    app.PrintPlain(f"Excel Write Time      : {write_time:.2f} s")
    app.PrintPlain(f"Saved                 : {OUTPUT_FILE}")


if __name__ == "__main__":
    run_bruteforce()
```

```python
import math
import os
import time
from itertools import combinations
from typing import Any, Dict, List, Optional, Sequence, Tuple

import pandas as pd

try:
    import powerfactory
except ImportError:
    powerfactory = None


COUP_LIST = [
    "BC Waru",
    "BC Rungkut",
    "BC Sawahan",
    "BC Gunung Sari",
    "BC Tandes",
    "BC Altaprima","BC Undaan","BC Kupang", "BC Sambikerep"
]

SUBSTAT_LIST = [
    "GRESIK BARU5", "KRIAN5", "GRESIK LAMA55", "KARANGPILANG5", "SAWAHAN5",
    "DARMO GRANDE5", "ALTA PRIMA5", "SEGOROMADU5", "SAMBIKEREP5", "WILMAR5",
    "BAMBE5", "GUNUNG SARI5", "PETRO KIMIA5", "WARU5", "RUNGKUT5",
     "BUDURAN5", "SURABAYA SELATAN5", "SUKOLILO5", "HANIL JAYA STEEL5",
    "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5",
    "SIDOARJO5", "NEWBUDURAN5", "SEDATI5", "SIMPANG5", "GILITIMUR5",
    "NWARU5", "TANDES5", "PERAK5", "UNDAAN5", "KUPANG5",
    "KREMBANGAN5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "KEDINDING5", "BANGKALAN5",
    "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5",
]

OPEN_COUNTS = [1, 2]

# "exact"   -> selected couplers open, all other monitored couplers closed
# "relative" -> selected couplers open, all other monitored couplers keep original state
SCENARIO_MODE = "exact"

# ===============================
# PARAMETER PIpf
# PIpf_i = Wi/(2n) * (Iactual / Ilimit)^(2n)
# ===============================
W_I_PF = 1.0
N_EXP_PF = 1

# ===============================
# PARAMETER PIsc
# PIsc_i = Wi/(2n) * (Ikss / SCC_limit)^(2n), jika Ikss > limit
# ===============================
SCC_LIMIT_KA = 40.0
W_I_SC = 1.0
N_EXP_SC = 1

# ===============================
# PARAMETER PIv Mode A
# PIv_i = Wvi/(2n) * ( ||Vi| - |Vi_rated|| / ΔVlim )^(2n)
# ===============================
W_V = 1.0
N_EXP_V = 1
V_RATED_PU_DEFAULT = 1.0
DV_LIMIT_PU_DEFAULT = 0.05

# ===============================
# PARAMETER OBJECTIVE GABUNGAN
# PI_total = a*PIpf_norm + b*PIsc_norm + c*PIv_norm
# dengan normalisasi berbasis baseline
# ===============================
WEIGHT_PF = 0.33
WEIGHT_SC = 0.33
WEIGHT_V = 0.34

# proteksi jika baseline = 0
EPSILON = 1e-12
ZERO_BASELINE_PENALTY = 1_000_000.0

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\Hasil Uji"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "BruteForce_BC_PIpf_PIsc_PIv_PI_Total_Final.xlsx")

LOG_EACH_SCENARIO = False
SAVE_ALL_LINE_DETAILS = False
SAVE_ALL_BUS_SC_DETAILS = False
SAVE_ALL_BUS_V_DETAILS = False
SAVE_BEST_DETAILS = True


def log(app: Any, message: str) -> None:
    try:
        app.PrintPlain(message)
    except Exception:
        pass


def get_app() -> Any:
    if powerfactory is None:
        raise RuntimeError("powerfactory module not available. Run this inside DIgSILENT PowerFactory.")
    app = powerfactory.GetApplication()
    if not app:
        raise RuntimeError("Run this script inside DIgSILENT PowerFactory.")
    log(app, "=== BRUTE FORCE PIpf + PIsc + PIv + PI TOTAL START ===")
    return app


def safe_get_attribute(obj: Any, attr: str, default: Any = None) -> Any:
    try:
        value = obj.GetAttribute(attr)
        return default if value is None else value
    except Exception:
        return default


def safe_loc_name(obj: Any, default: str = "") -> str:
    if obj is None:
        return default
    try:
        return obj.loc_name.strip()
    except Exception:
        return default


def get_obj_key(obj: Any) -> Optional[str]:
    if obj is None:
        return None
    try:
        return obj.GetFullName()
    except Exception:
        pass
    try:
        return f"{obj.GetClassName()}::{obj.loc_name.strip()}"
    except Exception:
        return str(obj)


def as_float(value: Any, default: Optional[float] = None) -> Optional[float]:
    try:
        number = float(value)
    except Exception:
        return default
    if not math.isfinite(number):
        return default
    return number


def get_parent_substation_name(obj: Any, max_depth: int = 20) -> Optional[str]:
    current = obj
    for _ in range(max_depth):
        if current is None:
            return None
        try:
            if current.GetClassName() == "ElmSubstat":
                return current.loc_name.strip()
            current = current.GetParent()
        except Exception:
            return None
    return None


def resolve_terminal(side_obj: Any) -> Any:
    if side_obj is None:
        return None

    try:
        if side_obj.GetClassName() == "ElmTerm":
            return side_obj
    except Exception:
        pass

    try:
        terminal = side_obj.cterm
        if terminal is not None:
            return terminal
    except Exception:
        pass

    return None


def run_loadflow(app: Any) -> bool:
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("ComLdf not found in active study case.")
    lf.iopt_net = 0
    lf.iopt_o4 = 0
    return lf.Execute() == 0


def prepare_short_circuit(app: Any) -> Any:
    sc = app.GetFromStudyCase("ComShc")
    if sc is None:
        raise RuntimeError("ComShc not found in active study case.")
    sc.iopt_mde = 0
    sc.Rf = 0
    sc.Xf = 0
    sc.iopt_allbus = 0
    sc.iopt_shc = "3psc"
    return sc


def get_line_terminals(line_obj: Any) -> Tuple[Any, Any]:
    cub1 = None
    cub2 = None

    try:
        cub1 = line_obj.bus1
    except Exception:
        pass

    try:
        cub2 = line_obj.bus2
    except Exception:
        pass

    if cub1 is None:
        cub1 = safe_get_attribute(line_obj, "bus1")
    if cub2 is None:
        cub2 = safe_get_attribute(line_obj, "bus2")

    term1 = resolve_terminal(cub1)
    term2 = resolve_terminal(cub2)
    return term1, term2


def cache_couplers(app: Any) -> Tuple[Dict[str, Dict[str, Any]], List[str]]:
    couplers: Dict[str, Dict[str, Any]] = {}
    for coupler in app.GetCalcRelevantObjects("*.ElmCoup"):
        name = safe_loc_name(coupler)
        if name in COUP_LIST:
            couplers[name] = {
                "obj": coupler,
                "orig_open": bool(coupler.IsOpen()),
            }

    ordered_names = [name for name in COUP_LIST if name in couplers]
    log(app, f"Found {len(ordered_names)} couplers from COUP_LIST")
    return couplers, ordered_names


def cache_buses(app: Any) -> Tuple[List[Dict[str, Any]], set, set]:
    substations = {
        safe_loc_name(substation): substation
        for substation in app.GetCalcRelevantObjects("*.ElmSubstat")
    }

    buses: List[Dict[str, Any]] = []
    monitored_substation_names = set()
    monitored_bus_keys = set()

    for sub_name in SUBSTAT_LIST:
        substation = substations.get(sub_name)
        if substation is None:
            log(app, f"Warning: Substation '{sub_name}' not found")
            continue

        monitored_substation_names.add(sub_name)
        all_terms = list(substation.GetContents("*.ElmTerm", 1) or [])

        for bus in all_terms:
            usage = safe_get_attribute(bus, "iUsage")
            try:
                if int(usage) != 0:
                    continue
            except Exception:
                continue

            key = get_obj_key(bus)
            if not key or key in monitored_bus_keys:
                continue

            monitored_bus_keys.add(key)
            buses.append({
                "substation": sub_name,
                "bus_name": safe_loc_name(bus),
                "obj": bus,
                "key": key,
            })

    log(app, f"Total monitored buses from SUBSTAT_LIST = {len(buses)}")
    return buses, monitored_bus_keys, monitored_substation_names


def cache_lines(app: Any, monitored_bus_keys: set, monitored_substation_names: set) -> List[Dict[str, Any]]:
    lines: List[Dict[str, Any]] = []
    seen_line_keys = set()

    for line in app.GetCalcRelevantObjects("*.ElmLne"):
        outserv = safe_get_attribute(line, "outserv", 0)
        try:
            if int(outserv) == 1:
                continue
        except Exception:
            pass

        term1, term2 = get_line_terminals(line)

        term1_key = get_obj_key(term1)
        term2_key = get_obj_key(term2)
        term1_sub = get_parent_substation_name(term1)
        term2_sub = get_parent_substation_name(term2)

        term1_match = term1_key in monitored_bus_keys or term1_sub in monitored_substation_names
        term2_match = term2_key in monitored_bus_keys or term2_sub in monitored_substation_names

        if not (term1_match or term2_match):
            continue

        line_key = get_obj_key(line)
        if not line_key or line_key in seen_line_keys:
            continue
        seen_line_keys.add(line_key)

        if term1_match:
            selected_side = "bus1"
            selected_term = term1
            selected_substation = term1_sub or ""
        else:
            selected_side = "bus2"
            selected_term = term2
            selected_substation = term2_sub or ""

        lines.append({
            "line_name": safe_loc_name(line),
            "selected_side": selected_side,
            "obj": line,
            "bus_name": safe_loc_name(selected_term),
            "substation": selected_substation,
        })

    log(app, f"Total monitored lines = {len(lines)}")
    return lines


def restore_couplers(couplers: Dict[str, Dict[str, Any]]) -> None:
    for info in couplers.values():
        obj = info["obj"]
        orig_open = info["orig_open"]

        try:
            current_open = bool(obj.IsOpen())
        except Exception:
            continue

        if current_open != orig_open:
            try:
                if orig_open:
                    obj.Open()
                else:
                    obj.Close()
            except Exception:
                continue


def apply_open_scenario(
    couplers: Dict[str, Dict[str, Any]],
    open_names: Sequence[str],
    scenario_mode: str = SCENARIO_MODE,
) -> None:
    open_set = set(open_names)

    if scenario_mode not in {"exact", "relative"}:
        raise ValueError(f"Unsupported SCENARIO_MODE: {scenario_mode}")

    for name, info in couplers.items():
        obj = info["obj"]
        if name in open_set:
            obj.Open()
        elif scenario_mode == "exact":
            obj.Close()


def calculate_pipf_total(lines: Sequence[Dict[str, Any]], save_details: bool = False) -> Tuple[float, List[Dict[str, Any]]]:
    total_pipf = 0.0
    details: List[Dict[str, Any]] = []

    for item in lines:
        line_obj = item["obj"]
        selected_side = item["selected_side"]

        loading = as_float(safe_get_attribute(line_obj, "m:loading", 0.0), 0.0)

        if selected_side == "bus1":
            i_actual = as_float(safe_get_attribute(line_obj, "m:I1:bus1", 0.0), 0.0)
        else:
            i_actual = as_float(safe_get_attribute(line_obj, "m:I1:bus2", 0.0), 0.0)

        i_limit = as_float(safe_get_attribute(line_obj, "c:curnom", 0.0), 0.0)

        i_actual = abs(i_actual or 0.0)
        i_limit = i_limit or 0.0

        if i_limit <= 0.0:
            pipf_i = 0.0
        else:
            pipf_i = (W_I_PF / (2 * N_EXP_PF)) * ((i_actual / i_limit) ** (2 * N_EXP_PF))

        total_pipf += pipf_i

        if save_details:
            details.append({
                "Substation": item["substation"],
                "Bus": item["bus_name"],
                "Line": item["line_name"],
                "Selected Side": selected_side,
                "Loading (%)": round(loading, 6),
                "I actual (kA)": round(i_actual, 6),
                "I limit (kA)": round(i_limit, 6),
                "PIpf_i": round(float(pipf_i), 8),
            })

    return total_pipf, details


def calculate_pisc_total(
    sc: Any,
    buses: Sequence[Dict[str, Any]],
    save_details: bool = False,
) -> Tuple[Optional[float], List[Dict[str, Any]], str]:
    total_pisc = 0.0
    details: List[Dict[str, Any]] = []

    for item in buses:
        bus = item["obj"]
        sc.shcobj = bus

        if sc.Execute() != 0:
            return None, details, f"SC_EXECUTE_FAILED: {item['substation']} / {item['bus_name']}"

        scc_i = as_float(safe_get_attribute(bus, "m:Ikss", None), None)
        if scc_i is None:
            return None, details, f"SC_RESULT_MISSING: {item['substation']} / {item['bus_name']}"

        if scc_i > SCC_LIMIT_KA:
            pisc_i = (W_I_SC / (2 * N_EXP_SC)) * ((scc_i / SCC_LIMIT_KA) ** (2 * N_EXP_SC))
        else:
            pisc_i = 0.0

        total_pisc += pisc_i

        if save_details:
            details.append({
                "Substation": item["substation"],
                "Bus": item["bus_name"],
                "SCC 3PH (kA)": round(scc_i, 6),
                "SCC Limit (kA)": round(SCC_LIMIT_KA, 6),
                "PIsc_i": round(float(pisc_i), 8),
            })

    return total_pisc, details, "OK"


def calculate_piv_total(
    buses: Sequence[Dict[str, Any]],
    save_details: bool = False,
) -> Tuple[Optional[float], List[Dict[str, Any]], str]:
    total_piv = 0.0
    details: List[Dict[str, Any]] = []

    for item in buses:
        sub_name = item["substation"]
        bus_name = item["bus_name"]
        bus_obj = item["obj"]

        vpu = as_float(safe_get_attribute(bus_obj, "m:u", None), None)
        vkv = as_float(safe_get_attribute(bus_obj, "m:Ul", None), None)

        if vpu is None:
            return None, details, f"VOLTAGE_RESULT_MISSING: {sub_name} / {bus_name}"

        deviation = abs(abs(vpu) - abs(V_RATED_PU_DEFAULT))

        if DV_LIMIT_PU_DEFAULT <= 0.0:
            piv_i = 0.0
        else:
            piv_i = (W_V / (2 * N_EXP_V)) * ((deviation / DV_LIMIT_PU_DEFAULT) ** (2 * N_EXP_V))

        total_piv += piv_i

        if vpu > (V_RATED_PU_DEFAULT + DV_LIMIT_PU_DEFAULT):
            status = "HIGH"
        elif vpu < (V_RATED_PU_DEFAULT - DV_LIMIT_PU_DEFAULT):
            status = "LOW"
        else:
            status = "NORMAL"

        if save_details:
            details.append({
                "Substation": sub_name,
                "Bus": bus_name,
                "Vpu": round(vpu, 6),
                "VkV": round(vkv, 3) if vkv is not None else None,
                "Weight": W_V,
                "Vrated_pu": V_RATED_PU_DEFAULT,
                "DeltaVlim_pu": DV_LIMIT_PU_DEFAULT,
                "Deviation": round(deviation, 8),
                "PIv_i": round(float(piv_i), 8),
                "Status": status,
            })

    return total_piv, details, "OK"


def normalize_by_baseline(value: Optional[float], baseline: Optional[float]) -> Optional[float]:
    if value is None or baseline is None:
        return None

    if abs(baseline) <= EPSILON:
        if abs(value) <= EPSILON:
            return 1.0
        return ZERO_BASELINE_PENALTY

    return value / baseline


def calculate_pi_total(
    pipf_total: float,
    pisc_total: float,
    piv_total: float,
    baseline_pipf: float,
    baseline_pisc: float,
    baseline_piv: float,
) -> Tuple[Optional[float], Optional[float], Optional[float], Optional[float]]:
    pipf_norm = normalize_by_baseline(pipf_total, baseline_pipf)
    pisc_norm = normalize_by_baseline(pisc_total, baseline_pisc)
    piv_norm = normalize_by_baseline(piv_total, baseline_piv)

    if pipf_norm is None or pisc_norm is None or piv_norm is None:
        return None, pipf_norm, pisc_norm, piv_norm

    pi_total = (
        WEIGHT_PF * pipf_norm +
        WEIGHT_SC * pisc_norm +
        WEIGHT_V * piv_norm
    )

    return pi_total, pipf_norm, pisc_norm, piv_norm


def evaluate_baseline(
    app: Any,
    couplers: Dict[str, Dict[str, Any]],
    lines: Sequence[Dict[str, Any]],
    buses: Sequence[Dict[str, Any]],
) -> Tuple[float, List[Dict[str, Any]], float, List[Dict[str, Any]], float, List[Dict[str, Any]]]:
    restore_couplers(couplers)

    if not run_loadflow(app):
        raise RuntimeError("Baseline load flow failed.")

    pipf_total, line_details = calculate_pipf_total(lines, save_details=True)

    piv_total, v_details, piv_status = calculate_piv_total(buses, save_details=True)
    if piv_total is None:
        raise RuntimeError(f"Baseline invalid: {piv_status}")

    sc = prepare_short_circuit(app)
    pisc_total, sc_details, pisc_status = calculate_pisc_total(sc, buses, save_details=True)
    if pisc_total is None:
        raise RuntimeError(f"Baseline invalid: {pisc_status}")

    return pipf_total, line_details, pisc_total, sc_details, piv_total, v_details


def evaluate_open_set(
    app: Any,
    couplers: Dict[str, Dict[str, Any]],
    lines: Sequence[Dict[str, Any]],
    buses: Sequence[Dict[str, Any]],
    open_names: Sequence[str],
    save_line_details: bool = False,
    save_bus_sc_details: bool = False,
    save_bus_v_details: bool = False,
) -> Tuple[Optional[float], List[Dict[str, Any]], Optional[float], List[Dict[str, Any]], Optional[float], List[Dict[str, Any]], str]:
    restore_couplers(couplers)

    try:
        apply_open_scenario(couplers, open_names, SCENARIO_MODE)

        if not run_loadflow(app):
            return None, [], None, [], None, [], "LOADFLOW_FAILED"

        pipf_total, line_details = calculate_pipf_total(lines, save_details=save_line_details)

        piv_total, v_details, piv_status = calculate_piv_total(buses, save_details=save_bus_v_details)
        if piv_total is None:
            return None, line_details, None, [], None, v_details, piv_status

        sc = prepare_short_circuit(app)
        pisc_total, sc_details, pisc_status = calculate_pisc_total(sc, buses, save_details=save_bus_sc_details)
        if pisc_total is None:
            return pipf_total, line_details, None, sc_details, piv_total, v_details, pisc_status

        return pipf_total, line_details, pisc_total, sc_details, piv_total, v_details, "OK"

    except Exception as exc:
        return None, [], None, [], None, [], f"ERROR: {exc}"

    finally:
        restore_couplers(couplers)


def build_ranking(df_all: pd.DataFrame, score_column: str) -> pd.DataFrame:
    if df_all.empty or score_column not in df_all.columns:
        return pd.DataFrame()

    df_valid = df_all[(df_all["Status"] == "OK") & (df_all[score_column].notna())]
    if df_valid.empty:
        return pd.DataFrame()

    return df_valid.sort_values(by=score_column, ascending=True).reset_index(drop=True)


def append_detail_rows(
    target_rows: List[Dict[str, Any]],
    scenario_name: str,
    n_open: int,
    opened_couplers: str,
    rows: Sequence[Dict[str, Any]],
) -> None:
    for row in rows:
        target_rows.append({
            "Scenario": scenario_name,
            "N Open": n_open,
            "Opened Couplers": opened_couplers,
            **row,
        })


def run_bruteforce() -> None:
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    app = get_app()

    cache_time = 0.0
    elapsed = 0.0

    t_cache0 = time.perf_counter()
    couplers, coupler_names = cache_couplers(app)
    buses, monitored_bus_keys, monitored_substation_names = cache_buses(app)
    lines = cache_lines(app, monitored_bus_keys, monitored_substation_names)
    cache_time = time.perf_counter() - t_cache0
    log(app, f"Cache objects time = {cache_time:.2f}s")

    if not couplers:
        raise RuntimeError("No couplers found from COUP_LIST.")
    if not coupler_names:
        raise RuntimeError("No ordered coupler names available.")
    if not buses:
        raise RuntimeError("No buses found from SUBSTAT_LIST.")
    if not lines:
        raise RuntimeError("No lines found connected to SUBSTAT_LIST.")

    scenario_rows: List[Dict[str, Any]] = []
    line_detail_rows: List[Dict[str, Any]] = []
    bus_sc_detail_rows: List[Dict[str, Any]] = []
    bus_v_detail_rows: List[Dict[str, Any]] = []

    baseline_pipf: Optional[float] = None
    baseline_pisc: Optional[float] = None
    baseline_piv: Optional[float] = None

    best_pipf_score = float("inf")
    best_pipf_open_names: Optional[List[str]] = None
    best_pipf_line_details: List[Dict[str, Any]] = []

    best_pisc_score = float("inf")
    best_pisc_open_names: Optional[List[str]] = None
    best_pisc_bus_details: List[Dict[str, Any]] = []

    best_piv_score = float("inf")
    best_piv_open_names: Optional[List[str]] = None
    best_piv_bus_details: List[Dict[str, Any]] = []

    best_total_score = float("inf")
    best_total_open_names: Optional[List[str]] = None
    best_total_line_details: List[Dict[str, Any]] = []
    best_total_sc_details: List[Dict[str, Any]] = []
    best_total_v_details: List[Dict[str, Any]] = []

    (
        baseline_pipf,
        baseline_line_details,
        baseline_pisc,
        baseline_sc_details,
        baseline_piv,
        baseline_v_details,
    ) = evaluate_baseline(app, couplers, lines, buses)

    baseline_pi_total = WEIGHT_PF + WEIGHT_SC + WEIGHT_V

    log(app, f"Baseline PIpf Total = {baseline_pipf:.6f}")
    log(app, f"Baseline PIsc Total = {baseline_pisc:.6f}")
    log(app, f"Baseline PIv  Total = {baseline_piv:.6f}")
    log(app, f"Baseline PI Total   = {baseline_pi_total:.6f}")

    scenario_rows.append({
        "No": 0,
        "N Open": 0,
        "Opened Couplers": "-",

        "PIpf Total": round(baseline_pipf, 8),
        "Delta PIpf vs Baseline": 0.0,
        "PIpf Norm": 1.0,

        "PIsc Total": round(baseline_pisc, 8),
        "Delta PIsc vs Baseline": 0.0,
        "PIsc Norm": 1.0,

        "PIv Total": round(baseline_piv, 8),
        "Delta PIv vs Baseline": 0.0,
        "PIv Norm": 1.0,

        "Weight PF": WEIGHT_PF,
        "Weight SC": WEIGHT_SC,
        "Weight V": WEIGHT_V,
        "PI Total": round(baseline_pi_total, 8),

        "Status": "OK",
        "Scenario Type": "BASELINE",
    })

    append_detail_rows(line_detail_rows, "BASELINE", 0, "-", baseline_line_details)
    append_detail_rows(bus_sc_detail_rows, "BASELINE", 0, "-", baseline_sc_details)
    append_detail_rows(bus_v_detail_rows, "BASELINE", 0, "-", baseline_v_details)

    total_scenarios = sum(math.comb(len(coupler_names), n) for n in OPEN_COUNTS if n <= len(coupler_names))
    log(app, f"Total evaluated combinations = {total_scenarios}")

    t0 = time.perf_counter()
    scenario_no = 1

    try:
        for n_open in OPEN_COUNTS:
            if n_open > len(coupler_names):
                log(app, f"Skip n_open={n_open}; available couplers only {len(coupler_names)}")
                continue

            log(app, f"===== START EVALUATION: OPEN {n_open} BC =====")

            for combo in combinations(coupler_names, n_open):
                open_names = list(combo)
                scenario_label = ", ".join(open_names)

                if LOG_EACH_SCENARIO:
                    log(app, f"[{scenario_no}] OPEN ({n_open} BC): {scenario_label}")

                save_line_details_now = SAVE_ALL_LINE_DETAILS or SAVE_BEST_DETAILS
                save_bus_sc_details_now = SAVE_ALL_BUS_SC_DETAILS or SAVE_BEST_DETAILS
                save_bus_v_details_now = SAVE_ALL_BUS_V_DETAILS or SAVE_BEST_DETAILS

                (
                    pipf_total,
                    line_details,
                    pisc_total,
                    sc_details,
                    piv_total,
                    v_details,
                    status,
                ) = evaluate_open_set(
                    app=app,
                    couplers=couplers,
                    lines=lines,
                    buses=buses,
                    open_names=open_names,
                    save_line_details=save_line_details_now,
                    save_bus_sc_details=save_bus_sc_details_now,
                    save_bus_v_details=save_bus_v_details_now,
                )

                if pipf_total is None or pisc_total is None or piv_total is None:
                    scenario_rows.append({
                        "No": scenario_no,
                        "N Open": n_open,
                        "Opened Couplers": scenario_label,

                        "PIpf Total": round(pipf_total, 8) if pipf_total is not None else None,
                        "Delta PIpf vs Baseline": round(pipf_total - baseline_pipf, 8) if pipf_total is not None else None,
                        "PIpf Norm": None,

                        "PIsc Total": round(pisc_total, 8) if pisc_total is not None else None,
                        "Delta PIsc vs Baseline": round(pisc_total - baseline_pisc, 8) if pisc_total is not None else None,
                        "PIsc Norm": None,

                        "PIv Total": round(piv_total, 8) if piv_total is not None else None,
                        "Delta PIv vs Baseline": round(piv_total - baseline_piv, 8) if piv_total is not None else None,
                        "PIv Norm": None,

                        "Weight PF": WEIGHT_PF,
                        "Weight SC": WEIGHT_SC,
                        "Weight V": WEIGHT_V,
                        "PI Total": None,

                        "Status": status,
                        "Scenario Type": "OPEN_SET",
                    })
                    scenario_no += 1
                    continue

                delta_pipf = pipf_total - baseline_pipf
                delta_pisc = pisc_total - baseline_pisc
                delta_piv = piv_total - baseline_piv

                pi_total, pipf_norm, pisc_norm, piv_norm = calculate_pi_total(
                    pipf_total=pipf_total,
                    pisc_total=pisc_total,
                    piv_total=piv_total,
                    baseline_pipf=baseline_pipf,
                    baseline_pisc=baseline_pisc,
                    baseline_piv=baseline_piv,
                )

                scenario_rows.append({
                    "No": scenario_no,
                    "N Open": n_open,
                    "Opened Couplers": scenario_label,

                    "PIpf Total": round(pipf_total, 8),
                    "Delta PIpf vs Baseline": round(delta_pipf, 8),
                    "PIpf Norm": round(pipf_norm, 8) if pipf_norm is not None else None,

                    "PIsc Total": round(pisc_total, 8),
                    "Delta PIsc vs Baseline": round(delta_pisc, 8),
                    "PIsc Norm": round(pisc_norm, 8) if pisc_norm is not None else None,

                    "PIv Total": round(piv_total, 8),
                    "Delta PIv vs Baseline": round(delta_piv, 8),
                    "PIv Norm": round(piv_norm, 8) if piv_norm is not None else None,

                    "Weight PF": WEIGHT_PF,
                    "Weight SC": WEIGHT_SC,
                    "Weight V": WEIGHT_V,
                    "PI Total": round(pi_total, 8) if pi_total is not None else None,

                    "Status": status,
                    "Scenario Type": "OPEN_SET",
                })

                if SAVE_ALL_LINE_DETAILS:
                    append_detail_rows(
                        line_detail_rows,
                        f"OPEN {scenario_label}",
                        n_open,
                        scenario_label,
                        line_details,
                    )

                if SAVE_ALL_BUS_SC_DETAILS:
                    append_detail_rows(
                        bus_sc_detail_rows,
                        f"OPEN {scenario_label}",
                        n_open,
                        scenario_label,
                        sc_details,
                    )

                if SAVE_ALL_BUS_V_DETAILS:
                    append_detail_rows(
                        bus_v_detail_rows,
                        f"OPEN {scenario_label}",
                        n_open,
                        scenario_label,
                        v_details,
                    )

                if pipf_total < best_pipf_score:
                    best_pipf_score = pipf_total
                    best_pipf_open_names = open_names
                    best_pipf_line_details = list(line_details)
                    log(app, f"NEW BEST PIpf     | scenario={scenario_no} | PIpf={best_pipf_score:.6f}")

                if pisc_total < best_pisc_score:
                    best_pisc_score = pisc_total
                    best_pisc_open_names = open_names
                    best_pisc_bus_details = list(sc_details)
                    log(app, f"NEW BEST PIsc     | scenario={scenario_no} | PIsc={best_pisc_score:.6f}")

                if piv_total < best_piv_score:
                    best_piv_score = piv_total
                    best_piv_open_names = open_names
                    best_piv_bus_details = list(v_details)
                    log(app, f"NEW BEST PIv      | scenario={scenario_no} | PIv={best_piv_score:.6f}")

                if pi_total is not None and pi_total < best_total_score:
                    best_total_score = pi_total
                    best_total_open_names = open_names
                    best_total_line_details = list(line_details)
                    best_total_sc_details = list(sc_details)
                    best_total_v_details = list(v_details)
                    log(app, f"NEW BEST PI TOTAL | scenario={scenario_no} | PI Total={best_total_score:.6f}")

                scenario_no += 1

        elapsed = time.perf_counter() - t0

    finally:
        restore_couplers(couplers)

    if SAVE_BEST_DETAILS and best_pipf_open_names and not SAVE_ALL_LINE_DETAILS:
        best_pipf_label = ", ".join(best_pipf_open_names)
        append_detail_rows(
            line_detail_rows,
            f"BEST PIpf OPEN {best_pipf_label}",
            len(best_pipf_open_names),
            best_pipf_label,
            best_pipf_line_details,
        )

    if SAVE_BEST_DETAILS and best_pisc_open_names and not SAVE_ALL_BUS_SC_DETAILS:
        best_pisc_label = ", ".join(best_pisc_open_names)
        append_detail_rows(
            bus_sc_detail_rows,
            f"BEST PIsc OPEN {best_pisc_label}",
            len(best_pisc_open_names),
            best_pisc_label,
            best_pisc_bus_details,
        )

    if SAVE_BEST_DETAILS and best_piv_open_names and not SAVE_ALL_BUS_V_DETAILS:
        best_piv_label = ", ".join(best_piv_open_names)
        append_detail_rows(
            bus_v_detail_rows,
            f"BEST PIv OPEN {best_piv_label}",
            len(best_piv_open_names),
            best_piv_label,
            best_piv_bus_details,
        )

    if SAVE_BEST_DETAILS and best_total_open_names:
        best_total_label = ", ".join(best_total_open_names)

        if not SAVE_ALL_LINE_DETAILS:
            append_detail_rows(
                line_detail_rows,
                f"BEST PI TOTAL OPEN {best_total_label}",
                len(best_total_open_names),
                best_total_label,
                best_total_line_details,
            )

        if not SAVE_ALL_BUS_SC_DETAILS:
            append_detail_rows(
                bus_sc_detail_rows,
                f"BEST PI TOTAL OPEN {best_total_label}",
                len(best_total_open_names),
                best_total_label,
                best_total_sc_details,
            )

        if not SAVE_ALL_BUS_V_DETAILS:
            append_detail_rows(
                bus_v_detail_rows,
                f"BEST PI TOTAL OPEN {best_total_label}",
                len(best_total_open_names),
                best_total_label,
                best_total_v_details,
            )

    df_all = pd.DataFrame(scenario_rows)
    df_line = pd.DataFrame(line_detail_rows)
    df_bus_sc = pd.DataFrame(bus_sc_detail_rows)
    df_bus_v = pd.DataFrame(bus_v_detail_rows)

    df_open_only = df_all[df_all["Scenario Type"] == "OPEN_SET"]

    df_rank_pipf = build_ranking(df_open_only, "PIpf Total")
    df_rank_pisc = build_ranking(df_open_only, "PIsc Total")
    df_rank_piv = build_ranking(df_open_only, "PIv Total")
    df_rank_total = build_ranking(df_open_only, "PI Total")

    df_best = pd.DataFrame([{
        "Scenario Mode": SCENARIO_MODE,
        "Weight PF": WEIGHT_PF,
        "Weight SC": WEIGHT_SC,
        "Weight V": WEIGHT_V,
        "Baseline PI Total": round(baseline_pi_total, 8),

        "Best PIpf Opened Couplers": ", ".join(best_pipf_open_names) if best_pipf_open_names else None,
        "Best PIpf N Open": len(best_pipf_open_names) if best_pipf_open_names else None,
        "Best PIpf Total": round(best_pipf_score, 8) if best_pipf_open_names else None,
        "Baseline PIpf Total": round(baseline_pipf, 8) if baseline_pipf is not None else None,

        "Best PIsc Opened Couplers": ", ".join(best_pisc_open_names) if best_pisc_open_names else None,
        "Best PIsc N Open": len(best_pisc_open_names) if best_pisc_open_names else None,
        "Best PIsc Total": round(best_pisc_score, 8) if best_pisc_open_names else None,
        "Baseline PIsc Total": round(baseline_pisc, 8) if baseline_pisc is not None else None,

        "Best PIv Opened Couplers": ", ".join(best_piv_open_names) if best_piv_open_names else None,
        "Best PIv N Open": len(best_piv_open_names) if best_piv_open_names else None,
        "Best PIv Total": round(best_piv_score, 8) if best_piv_open_names else None,
        "Baseline PIv Total": round(baseline_piv, 8) if baseline_piv is not None else None,

        "Best PI Total Opened Couplers": ", ".join(best_total_open_names) if best_total_open_names else None,
        "Best PI Total N Open": len(best_total_open_names) if best_total_open_names else None,
        "Best PI Total": round(best_total_score, 8) if best_total_open_names else None,

        "Execution Time (s)": round(elapsed, 2),
        "Cache Time (s)": round(cache_time, 2),
    }])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df_all.to_excel(writer, sheet_name="All Scenario", index=False)
        df_rank_pipf.to_excel(writer, sheet_name="Ranking PIpf", index=False)
        df_rank_pisc.to_excel(writer, sheet_name="Ranking PIsc", index=False)
        df_rank_piv.to_excel(writer, sheet_name="Ranking PIv", index=False)
        df_rank_total.to_excel(writer, sheet_name="Ranking PI Total", index=False)
        df_best.to_excel(writer, sheet_name="Best", index=False)

        if not df_line.empty:
            df_line.to_excel(writer, sheet_name="Line Detail PIpf", index=False)

        if not df_bus_sc.empty:
            df_bus_sc.to_excel(writer, sheet_name="Bus Detail PIsc", index=False)

        if not df_bus_v.empty:
            df_bus_v.to_excel(writer, sheet_name="Bus Detail PIv", index=False)

    log(app, "===== FINAL RESULT =====")
    log(app, f"Scenario Mode          : {SCENARIO_MODE}")

    log(app, f"Baseline PIpf Total    : {baseline_pipf:.6f}" if baseline_pipf is not None else "Baseline PIpf Total    : N/A")
    log(app, f"Best PIpf Couplers     : {', '.join(best_pipf_open_names)}" if best_pipf_open_names else "Best PIpf Couplers     : N/A")
    log(app, f"Best PIpf Total        : {best_pipf_score:.6f}" if best_pipf_open_names else "Best PIpf Total        : N/A")

    log(app, f"Baseline PIsc Total    : {baseline_pisc:.6f}" if baseline_pisc is not None else "Baseline PIsc Total    : N/A")
    log(app, f"Best PIsc Couplers     : {', '.join(best_pisc_open_names)}" if best_pisc_open_names else "Best PIsc Couplers     : N/A")
    log(app, f"Best PIsc Total        : {best_pisc_score:.6f}" if best_pisc_open_names else "Best PIsc Total        : N/A")

    log(app, f"Baseline PIv Total     : {baseline_piv:.6f}" if baseline_piv is not None else "Baseline PIv Total     : N/A")
    log(app, f"Best PIv Couplers      : {', '.join(best_piv_open_names)}" if best_piv_open_names else "Best PIv Couplers      : N/A")
    log(app, f"Best PIv Total         : {best_piv_score:.6f}" if best_piv_open_names else "Best PIv Total         : N/A")

    log(app, f"Baseline PI Total      : {baseline_pi_total:.6f}")
    log(app, f"Best PI Total Couplers : {', '.join(best_total_open_names)}" if best_total_open_names else "Best PI Total Couplers : N/A")
    log(app, f"Best PI Total          : {best_total_score:.6f}" if best_total_open_names else "Best PI Total          : N/A")

    log(app, f"Execution Time         : {elapsed:.2f} s")
    log(app, f"Saved                  : {OUTPUT_FILE}")


if __name__ == "__main__":
    run_bruteforce()
```

```python
import math
import os
import time
from itertools import combinations
from typing import Any, Dict, List, Optional, Sequence, Tuple

import pandas as pd

try:
    import powerfactory
except ImportError:
    powerfactory = None


COUP_LIST = [
    "BC Waru",
    "BC Rungkut",
    "BC Sawahan",
    "BC Gunung Sari",
    "BC Tandes",
    "BC Altaprima",
]

SUBSTAT_LIST = [
    "GRESIK BARU5", "KRIAN5", "GRESIK LAMA55", "KARANGPILANG5", "SAWAHAN5",
    "DARMO GRANDE5", "ALTA PRIMA5", "SEGOROMADU5", "SAMBIKEREP5", "WILMAR5",
    "BAMBE5", "GUNUNG SARI5", "PETRO KIMIA5", "WARU5", "RUNGKUT5",
    "BUDURAN5", "SURABAYA SELATAN5", "SUKOLILO5", "HANIL JAYA STEEL5",
    "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5",
    "SIDOARJO5", "NEWBUDURAN5", "SEDATI5", "SIMPANG5", "GILITIMUR5",
    "NWARU5", "TANDES5", "PERAK5", "UNDAAN5", "KUPANG5",
    "KREMBANGAN5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "KEDINDING5", "BANGKALAN5",
    "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5",
]

OPEN_COUNTS = [1, 2]

# "exact"   -> selected couplers open, all other monitored couplers closed
# "relative" -> selected couplers open, all other monitored couplers keep original state
SCENARIO_MODE = "exact"

# ===============================
# PARAMETER PIpf
# PIpf_i = Wi/(2n) * (Iactual / Ilimit)^(2n)
# ===============================
W_I_PF = 1.0
N_EXP_PF = 1

# ===============================
# PARAMETER PIsc
# PIsc_i = Wi/(2n) * (Ikss / SCC_limit)^(2n), jika Ikss > limit
# ===============================
SCC_LIMIT_KA = 40.0
W_I_SC = 1.0
N_EXP_SC = 1

# ===============================
# PARAMETER PIv Mode A
# PIv_i = Wvi/(2n) * ( ||Vi| - |Vi_rated|| / ΔVlim )^(2n)
# ===============================
W_V = 1.0
N_EXP_V = 1
V_RATED_PU_DEFAULT = 1.0
DV_LIMIT_PU_DEFAULT = 0.05

# ===============================
# AHP SENSITIVITY STUDY
# PI_total = a*PIpf_norm + b*PIsc_norm + c*PIv_norm
# ===============================
AHP_CASES: Dict[str, List[List[float]]] = {
    "PF_DOMINANT": [
        [1,   3,   5],
        [1/3, 1,   3],
        [1/5, 1/3, 1],
    ],
    "BALANCED": [
        [1,   1,   3],
        [1,   1,   3],
        [1/3, 1/3, 1],
    ],
    "SC_DOMINANT": [
        [1,   1/3, 3],
        [3,   1,   5],
        [1/3, 1/5, 1],
    ],
}

RI_TABLE = {
    1: 0.00,
    2: 0.00,
    3: 0.58,
    4: 0.90,
    5: 1.12,
    6: 1.24,
    7: 1.32,
    8: 1.41,
    9: 1.45,
    10: 1.49,
}

# proteksi jika baseline = 0
EPSILON = 1e-12
ZERO_BASELINE_PENALTY = 1_000_000.0

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\Hasil Uji"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "BruteForce_BC_PIpf_PIsc_PIv_PI_Total_AHP_Sensitivity.xlsx")

LOG_EACH_SCENARIO = False
SAVE_ALL_LINE_DETAILS = False
SAVE_ALL_BUS_SC_DETAILS = False
SAVE_ALL_BUS_V_DETAILS = False
SAVE_BEST_DETAILS = True


def log(app: Any, message: str) -> None:
    try:
        app.PrintPlain(message)
    except Exception:
        pass


def get_app() -> Any:
    if powerfactory is None:
        raise RuntimeError("powerfactory module not available. Run this inside DIgSILENT PowerFactory.")
    app = powerfactory.GetApplication()
    if not app:
        raise RuntimeError("Run this script inside DIgSILENT PowerFactory.")
    log(app, "=== BRUTE FORCE PIpf + PIsc + PIv + PI TOTAL + AHP SENSITIVITY START ===")
    return app


def ahp_weights(matrix: List[List[float]]) -> Tuple[List[float], float, float]:
    """
    Return:
        weights   : normalized AHP weights
        lambda_max: max eigen approximation
        cr        : consistency ratio
    Method: geometric mean
    """
    n = len(matrix)
    if n == 0:
        raise ValueError("AHP matrix is empty.")
    if any(len(row) != n for row in matrix):
        raise ValueError("AHP matrix must be square.")

    geo_means: List[float] = []
    for row in matrix:
        prod = 1.0
        for value in row:
            if value <= 0:
                raise ValueError("AHP matrix values must be > 0.")
            prod *= float(value)
        geo_means.append(prod ** (1.0 / n))

    total_geo = sum(geo_means)
    if total_geo <= 0:
        raise ValueError("Invalid AHP matrix: zero geometric mean sum.")

    weights = [g / total_geo for g in geo_means]

    aw: List[float] = []
    for i in range(n):
        aw_i = 0.0
        for j in range(n):
            aw_i += matrix[i][j] * weights[j]
        aw.append(aw_i)

    lambda_max = sum(aw[i] / weights[i] for i in range(n)) / n

    ci = (lambda_max - n) / (n - 1) if n > 1 else 0.0
    ri = RI_TABLE.get(n, 1.49)
    cr = (ci / ri) if ri > 0 else 0.0

    return weights, lambda_max, cr


def safe_get_attribute(obj: Any, attr: str, default: Any = None) -> Any:
    try:
        value = obj.GetAttribute(attr)
        return default if value is None else value
    except Exception:
        return default


def safe_loc_name(obj: Any, default: str = "") -> str:
    if obj is None:
        return default
    try:
        return obj.loc_name.strip()
    except Exception:
        return default


def get_obj_key(obj: Any) -> Optional[str]:
    if obj is None:
        return None
    try:
        return obj.GetFullName()
    except Exception:
        pass
    try:
        return f"{obj.GetClassName()}::{obj.loc_name.strip()}"
    except Exception:
        return str(obj)


def as_float(value: Any, default: Optional[float] = None) -> Optional[float]:
    try:
        number = float(value)
    except Exception:
        return default
    if not math.isfinite(number):
        return default
    return number


def get_parent_substation_name(obj: Any, max_depth: int = 20) -> Optional[str]:
    current = obj
    for _ in range(max_depth):
        if current is None:
            return None
        try:
            if current.GetClassName() == "ElmSubstat":
                return current.loc_name.strip()
            current = current.GetParent()
        except Exception:
            return None
    return None


def resolve_terminal(side_obj: Any) -> Any:
    if side_obj is None:
        return None

    try:
        if side_obj.GetClassName() == "ElmTerm":
            return side_obj
    except Exception:
        pass

    try:
        terminal = side_obj.cterm
        if terminal is not None:
            return terminal
    except Exception:
        pass

    return None


def run_loadflow(app: Any) -> bool:
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("ComLdf not found in active study case.")
    lf.iopt_net = 0
    lf.iopt_o4 = 0
    return lf.Execute() == 0


def prepare_short_circuit(app: Any) -> Any:
    sc = app.GetFromStudyCase("ComShc")
    if sc is None:
        raise RuntimeError("ComShc not found in active study case.")
    sc.iopt_mde = 0
    sc.Rf = 0
    sc.Xf = 0
    sc.iopt_allbus = 0
    sc.iopt_shc = "3psc"
    return sc


def get_line_terminals(line_obj: Any) -> Tuple[Any, Any]:
    cub1 = None
    cub2 = None

    try:
        cub1 = line_obj.bus1
    except Exception:
        pass

    try:
        cub2 = line_obj.bus2
    except Exception:
        pass

    if cub1 is None:
        cub1 = safe_get_attribute(line_obj, "bus1")
    if cub2 is None:
        cub2 = safe_get_attribute(line_obj, "bus2")

    term1 = resolve_terminal(cub1)
    term2 = resolve_terminal(cub2)
    return term1, term2


def cache_couplers(app: Any) -> Tuple[Dict[str, Dict[str, Any]], List[str]]:
    couplers: Dict[str, Dict[str, Any]] = {}
    for coupler in app.GetCalcRelevantObjects("*.ElmCoup"):
        name = safe_loc_name(coupler)
        if name in COUP_LIST:
            couplers[name] = {
                "obj": coupler,
                "orig_open": bool(coupler.IsOpen()),
            }

    ordered_names = [name for name in COUP_LIST if name in couplers]
    log(app, f"Found {len(ordered_names)} couplers from COUP_LIST")
    return couplers, ordered_names


def cache_buses(app: Any) -> Tuple[List[Dict[str, Any]], set, set]:
    substations = {
        safe_loc_name(substation): substation
        for substation in app.GetCalcRelevantObjects("*.ElmSubstat")
    }

    buses: List[Dict[str, Any]] = []
    monitored_substation_names = set()
    monitored_bus_keys = set()

    for sub_name in SUBSTAT_LIST:
        substation = substations.get(sub_name)
        if substation is None:
            log(app, f"Warning: Substation '{sub_name}' not found")
            continue

        monitored_substation_names.add(sub_name)
        all_terms = list(substation.GetContents("*.ElmTerm", 1) or [])

        for bus in all_terms:
            usage = safe_get_attribute(bus, "iUsage")
            try:
                if int(usage) != 0:
                    continue
            except Exception:
                continue

            key = get_obj_key(bus)
            if not key or key in monitored_bus_keys:
                continue

            monitored_bus_keys.add(key)
            buses.append({
                "substation": sub_name,
                "bus_name": safe_loc_name(bus),
                "obj": bus,
                "key": key,
            })

    log(app, f"Total monitored buses from SUBSTAT_LIST = {len(buses)}")
    return buses, monitored_bus_keys, monitored_substation_names


def cache_lines(app: Any, monitored_bus_keys: set, monitored_substation_names: set) -> List[Dict[str, Any]]:
    lines: List[Dict[str, Any]] = []
    seen_line_keys = set()

    for line in app.GetCalcRelevantObjects("*.ElmLne"):
        outserv = safe_get_attribute(line, "outserv", 0)
        try:
            if int(outserv) == 1:
                continue
        except Exception:
            pass

        term1, term2 = get_line_terminals(line)

        term1_key = get_obj_key(term1)
        term2_key = get_obj_key(term2)
        term1_sub = get_parent_substation_name(term1)
        term2_sub = get_parent_substation_name(term2)

        term1_match = term1_key in monitored_bus_keys or term1_sub in monitored_substation_names
        term2_match = term2_key in monitored_bus_keys or term2_sub in monitored_substation_names

        if not (term1_match or term2_match):
            continue

        line_key = get_obj_key(line)
        if not line_key or line_key in seen_line_keys:
            continue
        seen_line_keys.add(line_key)

        if term1_match:
            selected_side = "bus1"
            selected_term = term1
            selected_substation = term1_sub or ""
        else:
            selected_side = "bus2"
            selected_term = term2
            selected_substation = term2_sub or ""

        lines.append({
            "line_name": safe_loc_name(line),
            "selected_side": selected_side,
            "obj": line,
            "bus_name": safe_loc_name(selected_term),
            "substation": selected_substation,
        })

    log(app, f"Total monitored lines = {len(lines)}")
    return lines


def restore_couplers(couplers: Dict[str, Dict[str, Any]]) -> None:
    for info in couplers.values():
        obj = info["obj"]
        orig_open = info["orig_open"]

        try:
            current_open = bool(obj.IsOpen())
        except Exception:
            continue

        if current_open != orig_open:
            try:
                if orig_open:
                    obj.Open()
                else:
                    obj.Close()
            except Exception:
                continue


def apply_open_scenario(
    couplers: Dict[str, Dict[str, Any]],
    open_names: Sequence[str],
    scenario_mode: str = SCENARIO_MODE,
) -> None:
    open_set = set(open_names)

    if scenario_mode not in {"exact", "relative"}:
        raise ValueError(f"Unsupported SCENARIO_MODE: {scenario_mode}")

    for name, info in couplers.items():
        obj = info["obj"]
        if name in open_set:
            obj.Open()
        elif scenario_mode == "exact":
            obj.Close()


def calculate_pipf_total(lines: Sequence[Dict[str, Any]], save_details: bool = False) -> Tuple[float, List[Dict[str, Any]]]:
    total_pipf = 0.0
    details: List[Dict[str, Any]] = []

    for item in lines:
        line_obj = item["obj"]
        selected_side = item["selected_side"]

        loading = as_float(safe_get_attribute(line_obj, "m:loading", 0.0), 0.0)

        if selected_side == "bus1":
            i_actual = as_float(safe_get_attribute(line_obj, "m:I1:bus1", 0.0), 0.0)
        else:
            i_actual = as_float(safe_get_attribute(line_obj, "m:I1:bus2", 0.0), 0.0)

        i_limit = as_float(safe_get_attribute(line_obj, "c:curnom", 0.0), 0.0)

        i_actual = abs(i_actual or 0.0)
        i_limit = i_limit or 0.0

        if i_limit <= 0.0:
            pipf_i = 0.0
        else:
            pipf_i = (W_I_PF / (2 * N_EXP_PF)) * ((i_actual / i_limit) ** (2 * N_EXP_PF))

        total_pipf += pipf_i

        if save_details:
            details.append({
                "Substation": item["substation"],
                "Bus": item["bus_name"],
                "Line": item["line_name"],
                "Selected Side": selected_side,
                "Loading (%)": round(loading, 6),
                "I actual (kA)": round(i_actual, 6),
                "I limit (kA)": round(i_limit, 6),
                "PIpf_i": round(float(pipf_i), 8),
            })

    return total_pipf, details


def calculate_pisc_total(
    sc: Any,
    buses: Sequence[Dict[str, Any]],
    save_details: bool = False,
) -> Tuple[Optional[float], List[Dict[str, Any]], str]:
    total_pisc = 0.0
    details: List[Dict[str, Any]] = []

    for item in buses:
        bus = item["obj"]
        sc.shcobj = bus

        if sc.Execute() != 0:
            return None, details, f"SC_EXECUTE_FAILED: {item['substation']} / {item['bus_name']}"

        scc_i = as_float(safe_get_attribute(bus, "m:Ikss", None), None)
        if scc_i is None:
            return None, details, f"SC_RESULT_MISSING: {item['substation']} / {item['bus_name']}"

        if scc_i > SCC_LIMIT_KA:
            pisc_i = (W_I_SC / (2 * N_EXP_SC)) * ((scc_i / SCC_LIMIT_KA) ** (2 * N_EXP_SC))
        else:
            pisc_i = 0.0

        total_pisc += pisc_i

        if save_details:
            details.append({
                "Substation": item["substation"],
                "Bus": item["bus_name"],
                "SCC 3PH (kA)": round(scc_i, 6),
                "SCC Limit (kA)": round(SCC_LIMIT_KA, 6),
                "PIsc_i": round(float(pisc_i), 8),
            })

    return total_pisc, details, "OK"


def calculate_piv_total(
    buses: Sequence[Dict[str, Any]],
    save_details: bool = False,
) -> Tuple[Optional[float], List[Dict[str, Any]], str]:
    total_piv = 0.0
    details: List[Dict[str, Any]] = []

    for item in buses:
        sub_name = item["substation"]
        bus_name = item["bus_name"]
        bus_obj = item["obj"]

        vpu = as_float(safe_get_attribute(bus_obj, "m:u", None), None)
        vkv = as_float(safe_get_attribute(bus_obj, "m:Ul", None), None)

        if vpu is None:
            return None, details, f"VOLTAGE_RESULT_MISSING: {sub_name} / {bus_name}"

        deviation = abs(abs(vpu) - abs(V_RATED_PU_DEFAULT))

        if DV_LIMIT_PU_DEFAULT <= 0.0:
            piv_i = 0.0
        else:
            piv_i = (W_V / (2 * N_EXP_V)) * ((deviation / DV_LIMIT_PU_DEFAULT) ** (2 * N_EXP_V))

        total_piv += piv_i

        if vpu > (V_RATED_PU_DEFAULT + DV_LIMIT_PU_DEFAULT):
            status = "HIGH"
        elif vpu < (V_RATED_PU_DEFAULT - DV_LIMIT_PU_DEFAULT):
            status = "LOW"
        else:
            status = "NORMAL"

        if save_details:
            details.append({
                "Substation": sub_name,
                "Bus": bus_name,
                "Vpu": round(vpu, 6),
                "VkV": round(vkv, 3) if vkv is not None else None,
                "Weight": W_V,
                "Vrated_pu": V_RATED_PU_DEFAULT,
                "DeltaVlim_pu": DV_LIMIT_PU_DEFAULT,
                "Deviation": round(deviation, 8),
                "PIv_i": round(float(piv_i), 8),
                "Status": status,
            })

    return total_piv, details, "OK"


def normalize_by_baseline(value: Optional[float], baseline: Optional[float]) -> Optional[float]:
    if value is None or baseline is None:
        return None

    if abs(baseline) <= EPSILON:
        if abs(value) <= EPSILON:
            return 1.0
        return ZERO_BASELINE_PENALTY

    return value / baseline


def calculate_pi_total(
    pipf_total: float,
    pisc_total: float,
    piv_total: float,
    baseline_pipf: float,
    baseline_pisc: float,
    baseline_piv: float,
    weight_pf: float,
    weight_sc: float,
    weight_v: float,
) -> Tuple[Optional[float], Optional[float], Optional[float], Optional[float]]:
    pipf_norm = normalize_by_baseline(pipf_total, baseline_pipf)
    pisc_norm = normalize_by_baseline(pisc_total, baseline_pisc)
    piv_norm = normalize_by_baseline(piv_total, baseline_piv)

    if pipf_norm is None or pisc_norm is None or piv_norm is None:
        return None, pipf_norm, pisc_norm, piv_norm

    pi_total = (
        weight_pf * pipf_norm +
        weight_sc * pisc_norm +
        weight_v * piv_norm
    )

    return pi_total, pipf_norm, pisc_norm, piv_norm


def evaluate_baseline(
    app: Any,
    couplers: Dict[str, Dict[str, Any]],
    lines: Sequence[Dict[str, Any]],
    buses: Sequence[Dict[str, Any]],
) -> Tuple[float, List[Dict[str, Any]], float, List[Dict[str, Any]], float, List[Dict[str, Any]]]:
    restore_couplers(couplers)

    if not run_loadflow(app):
        raise RuntimeError("Baseline load flow failed.")

    pipf_total, line_details = calculate_pipf_total(lines, save_details=True)

    piv_total, v_details, piv_status = calculate_piv_total(buses, save_details=True)
    if piv_total is None:
        raise RuntimeError(f"Baseline invalid: {piv_status}")

    sc = prepare_short_circuit(app)
    pisc_total, sc_details, pisc_status = calculate_pisc_total(sc, buses, save_details=True)
    if pisc_total is None:
        raise RuntimeError(f"Baseline invalid: {pisc_status}")

    return pipf_total, line_details, pisc_total, sc_details, piv_total, v_details


def evaluate_open_set(
    app: Any,
    couplers: Dict[str, Dict[str, Any]],
    lines: Sequence[Dict[str, Any]],
    buses: Sequence[Dict[str, Any]],
    open_names: Sequence[str],
    save_line_details: bool = False,
    save_bus_sc_details: bool = False,
    save_bus_v_details: bool = False,
) -> Tuple[Optional[float], List[Dict[str, Any]], Optional[float], List[Dict[str, Any]], Optional[float], List[Dict[str, Any]], str]:
    restore_couplers(couplers)

    try:
        apply_open_scenario(couplers, open_names, SCENARIO_MODE)

        if not run_loadflow(app):
            return None, [], None, [], None, [], "LOADFLOW_FAILED"

        pipf_total, line_details = calculate_pipf_total(lines, save_details=save_line_details)

        piv_total, v_details, piv_status = calculate_piv_total(buses, save_details=save_bus_v_details)
        if piv_total is None:
            return None, line_details, None, [], None, v_details, piv_status

        sc = prepare_short_circuit(app)
        pisc_total, sc_details, pisc_status = calculate_pisc_total(sc, buses, save_details=save_bus_sc_details)
        if pisc_total is None:
            return pipf_total, line_details, None, sc_details, piv_total, v_details, pisc_status

        return pipf_total, line_details, pisc_total, sc_details, piv_total, v_details, "OK"

    except Exception as exc:
        return None, [], None, [], None, [], f"ERROR: {exc}"

    finally:
        restore_couplers(couplers)


def build_ranking(df_all: pd.DataFrame, score_column: str) -> pd.DataFrame:
    if df_all.empty or score_column not in df_all.columns:
        return pd.DataFrame()

    df_valid = df_all[(df_all["Status"] == "OK") & (df_all[score_column].notna())]
    if df_valid.empty:
        return pd.DataFrame()

    return df_valid.sort_values(by=score_column, ascending=True).reset_index(drop=True)


def append_detail_rows(
    target_rows: List[Dict[str, Any]],
    case_name: str,
    scenario_name: str,
    n_open: int,
    opened_couplers: str,
    rows: Sequence[Dict[str, Any]],
) -> None:
    for row in rows:
        target_rows.append({
            "Case": case_name,
            "Scenario": scenario_name,
            "N Open": n_open,
            "Opened Couplers": opened_couplers,
            **row,
        })


def run_bruteforce(
    case_name: str,
    weight_pf: float,
    weight_sc: float,
    weight_v: float,
) -> Dict[str, pd.DataFrame]:
    app = get_app()

    cache_time = 0.0
    elapsed = 0.0

    t_cache0 = time.perf_counter()
    couplers, coupler_names = cache_couplers(app)
    buses, monitored_bus_keys, monitored_substation_names = cache_buses(app)
    lines = cache_lines(app, monitored_bus_keys, monitored_substation_names)
    cache_time = time.perf_counter() - t_cache0
    log(app, f"[{case_name}] Cache objects time = {cache_time:.2f}s")

    if not couplers:
        raise RuntimeError("No couplers found from COUP_LIST.")
    if not coupler_names:
        raise RuntimeError("No ordered coupler names available.")
    if not buses:
        raise RuntimeError("No buses found from SUBSTAT_LIST.")
    if not lines:
        raise RuntimeError("No lines found connected to SUBSTAT_LIST.")

    scenario_rows: List[Dict[str, Any]] = []
    line_detail_rows: List[Dict[str, Any]] = []
    bus_sc_detail_rows: List[Dict[str, Any]] = []
    bus_v_detail_rows: List[Dict[str, Any]] = []

    baseline_pipf: Optional[float] = None
    baseline_pisc: Optional[float] = None
    baseline_piv: Optional[float] = None

    best_pipf_score = float("inf")
    best_pipf_open_names: Optional[List[str]] = None
    best_pipf_line_details: List[Dict[str, Any]] = []

    best_pisc_score = float("inf")
    best_pisc_open_names: Optional[List[str]] = None
    best_pisc_bus_details: List[Dict[str, Any]] = []

    best_piv_score = float("inf")
    best_piv_open_names: Optional[List[str]] = None
    best_piv_bus_details: List[Dict[str, Any]] = []

    best_total_score = float("inf")
    best_total_open_names: Optional[List[str]] = None
    best_total_line_details: List[Dict[str, Any]] = []
    best_total_sc_details: List[Dict[str, Any]] = []
    best_total_v_details: List[Dict[str, Any]] = []

    (
        baseline_pipf,
        baseline_line_details,
        baseline_pisc,
        baseline_sc_details,
        baseline_piv,
        baseline_v_details,
    ) = evaluate_baseline(app, couplers, lines, buses)

    baseline_pi_total = weight_pf + weight_sc + weight_v

    log(app, f"[{case_name}] Weight PF = {weight_pf:.6f}")
    log(app, f"[{case_name}] Weight SC = {weight_sc:.6f}")
    log(app, f"[{case_name}] Weight V  = {weight_v:.6f}")
    log(app, f"[{case_name}] Baseline PIpf Total = {baseline_pipf:.6f}")
    log(app, f"[{case_name}] Baseline PIsc Total = {baseline_pisc:.6f}")
    log(app, f"[{case_name}] Baseline PIv  Total = {baseline_piv:.6f}")
    log(app, f"[{case_name}] Baseline PI Total   = {baseline_pi_total:.6f}")

    scenario_rows.append({
        "Case": case_name,
        "No": 0,
        "N Open": 0,
        "Opened Couplers": "-",

        "PIpf Total": round(baseline_pipf, 8),
        "Delta PIpf vs Baseline": 0.0,
        "PIpf Norm": 1.0,

        "PIsc Total": round(baseline_pisc, 8),
        "Delta PIsc vs Baseline": 0.0,
        "PIsc Norm": 1.0,

        "PIv Total": round(baseline_piv, 8),
        "Delta PIv vs Baseline": 0.0,
        "PIv Norm": 1.0,

        "Weight PF": weight_pf,
        "Weight SC": weight_sc,
        "Weight V": weight_v,
        "PI Total": round(baseline_pi_total, 8),

        "Status": "OK",
        "Scenario Type": "BASELINE",
    })

    append_detail_rows(line_detail_rows, case_name, "BASELINE", 0, "-", baseline_line_details)
    append_detail_rows(bus_sc_detail_rows, case_name, "BASELINE", 0, "-", baseline_sc_details)
    append_detail_rows(bus_v_detail_rows, case_name, "BASELINE", 0, "-", baseline_v_details)

    total_scenarios = sum(math.comb(len(coupler_names), n) for n in OPEN_COUNTS if n <= len(coupler_names))
    log(app, f"[{case_name}] Total evaluated combinations = {total_scenarios}")

    t0 = time.perf_counter()
    scenario_no = 1

    try:
        for n_open in OPEN_COUNTS:
            if n_open > len(coupler_names):
                log(app, f"[{case_name}] Skip n_open={n_open}; available couplers only {len(coupler_names)}")
                continue

            log(app, f"[{case_name}] ===== START EVALUATION: OPEN {n_open} BC =====")

            for combo in combinations(coupler_names, n_open):
                open_names = list(combo)
                scenario_label = ", ".join(open_names)

                if LOG_EACH_SCENARIO:
                    log(app, f"[{case_name}] [{scenario_no}] OPEN ({n_open} BC): {scenario_label}")

                save_line_details_now = SAVE_ALL_LINE_DETAILS or SAVE_BEST_DETAILS
                save_bus_sc_details_now = SAVE_ALL_BUS_SC_DETAILS or SAVE_BEST_DETAILS
                save_bus_v_details_now = SAVE_ALL_BUS_V_DETAILS or SAVE_BEST_DETAILS

                (
                    pipf_total,
                    line_details,
                    pisc_total,
                    sc_details,
                    piv_total,
                    v_details,
                    status,
                ) = evaluate_open_set(
                    app=app,
                    couplers=couplers,
                    lines=lines,
                    buses=buses,
                    open_names=open_names,
                    save_line_details=save_line_details_now,
                    save_bus_sc_details=save_bus_sc_details_now,
                    save_bus_v_details=save_bus_v_details_now,
                )

                if pipf_total is None or pisc_total is None or piv_total is None:
                    scenario_rows.append({
                        "Case": case_name,
                        "No": scenario_no,
                        "N Open": n_open,
                        "Opened Couplers": scenario_label,

                        "PIpf Total": round(pipf_total, 8) if pipf_total is not None else None,
                        "Delta PIpf vs Baseline": round(pipf_total - baseline_pipf, 8) if pipf_total is not None else None,
                        "PIpf Norm": None,

                        "PIsc Total": round(pisc_total, 8) if pisc_total is not None else None,
                        "Delta PIsc vs Baseline": round(pisc_total - baseline_pisc, 8) if pisc_total is not None else None,
                        "PIsc Norm": None,

                        "PIv Total": round(piv_total, 8) if piv_total is not None else None,
                        "Delta PIv vs Baseline": round(piv_total - baseline_piv, 8) if piv_total is not None else None,
                        "PIv Norm": None,

                        "Weight PF": weight_pf,
                        "Weight SC": weight_sc,
                        "Weight V": weight_v,
                        "PI Total": None,

                        "Status": status,
                        "Scenario Type": "OPEN_SET",
                    })
                    scenario_no += 1
                    continue

                delta_pipf = pipf_total - baseline_pipf
                delta_pisc = pisc_total - baseline_pisc
                delta_piv = piv_total - baseline_piv

                pi_total, pipf_norm, pisc_norm, piv_norm = calculate_pi_total(
                    pipf_total=pipf_total,
                    pisc_total=pisc_total,
                    piv_total=piv_total,
                    baseline_pipf=baseline_pipf,
                    baseline_pisc=baseline_pisc,
                    baseline_piv=baseline_piv,
                    weight_pf=weight_pf,
                    weight_sc=weight_sc,
                    weight_v=weight_v,
                )

                scenario_rows.append({
                    "Case": case_name,
                    "No": scenario_no,
                    "N Open": n_open,
                    "Opened Couplers": scenario_label,

                    "PIpf Total": round(pipf_total, 8),
                    "Delta PIpf vs Baseline": round(delta_pipf, 8),
                    "PIpf Norm": round(pipf_norm, 8) if pipf_norm is not None else None,

                    "PIsc Total": round(pisc_total, 8),
                    "Delta PIsc vs Baseline": round(delta_pisc, 8),
                    "PIsc Norm": round(pisc_norm, 8) if pisc_norm is not None else None,

                    "PIv Total": round(piv_total, 8),
                    "Delta PIv vs Baseline": round(delta_piv, 8),
                    "PIv Norm": round(piv_norm, 8) if piv_norm is not None else None,

                    "Weight PF": weight_pf,
                    "Weight SC": weight_sc,
                    "Weight V": weight_v,
                    "PI Total": round(pi_total, 8) if pi_total is not None else None,

                    "Status": status,
                    "Scenario Type": "OPEN_SET",
                })

                if SAVE_ALL_LINE_DETAILS:
                    append_detail_rows(
                        line_detail_rows,
                        case_name,
                        f"OPEN {scenario_label}",
                        n_open,
                        scenario_label,
                        line_details,
                    )

                if SAVE_ALL_BUS_SC_DETAILS:
                    append_detail_rows(
                        bus_sc_detail_rows,
                        case_name,
                        f"OPEN {scenario_label}",
                        n_open,
                        scenario_label,
                        sc_details,
                    )

                if SAVE_ALL_BUS_V_DETAILS:
                    append_detail_rows(
                        bus_v_detail_rows,
                        case_name,
                        f"OPEN {scenario_label}",
                        n_open,
                        scenario_label,
                        v_details,
                    )

                if pipf_total < best_pipf_score:
                    best_pipf_score = pipf_total
                    best_pipf_open_names = open_names
                    best_pipf_line_details = list(line_details)
                    log(app, f"[{case_name}] NEW BEST PIpf     | scenario={scenario_no} | PIpf={best_pipf_score:.6f}")

                if pisc_total < best_pisc_score:
                    best_pisc_score = pisc_total
                    best_pisc_open_names = open_names
                    best_pisc_bus_details = list(sc_details)
                    log(app, f"[{case_name}] NEW BEST PIsc     | scenario={scenario_no} | PIsc={best_pisc_score:.6f}")

                if piv_total < best_piv_score:
                    best_piv_score = piv_total
                    best_piv_open_names = open_names
                    best_piv_bus_details = list(v_details)
                    log(app, f"[{case_name}] NEW BEST PIv      | scenario={scenario_no} | PIv={best_piv_score:.6f}")

                if pi_total is not None and pi_total < best_total_score:
                    best_total_score = pi_total
                    best_total_open_names = open_names
                    best_total_line_details = list(line_details)
                    best_total_sc_details = list(sc_details)
                    best_total_v_details = list(v_details)
                    log(app, f"[{case_name}] NEW BEST PI TOTAL | scenario={scenario_no} | PI Total={best_total_score:.6f}")

                scenario_no += 1

        elapsed = time.perf_counter() - t0

    finally:
        restore_couplers(couplers)

    if SAVE_BEST_DETAILS and best_pipf_open_names and not SAVE_ALL_LINE_DETAILS:
        best_pipf_label = ", ".join(best_pipf_open_names)
        append_detail_rows(
            line_detail_rows,
            case_name,
            f"BEST PIpf OPEN {best_pipf_label}",
            len(best_pipf_open_names),
            best_pipf_label,
            best_pipf_line_details,
        )

    if SAVE_BEST_DETAILS and best_pisc_open_names and not SAVE_ALL_BUS_SC_DETAILS:
        best_pisc_label = ", ".join(best_pisc_open_names)
        append_detail_rows(
            bus_sc_detail_rows,
            case_name,
            f"BEST PIsc OPEN {best_pisc_label}",
            len(best_pisc_open_names),
            best_pisc_label,
            best_pisc_bus_details,
        )

    if SAVE_BEST_DETAILS and best_piv_open_names and not SAVE_ALL_BUS_V_DETAILS:
        best_piv_label = ", ".join(best_piv_open_names)
        append_detail_rows(
            bus_v_detail_rows,
            case_name,
            f"BEST PIv OPEN {best_piv_label}",
            len(best_piv_open_names),
            best_piv_label,
            best_piv_bus_details,
        )

    if SAVE_BEST_DETAILS and best_total_open_names:
        best_total_label = ", ".join(best_total_open_names)

        if not SAVE_ALL_LINE_DETAILS:
            append_detail_rows(
                line_detail_rows,
                case_name,
                f"BEST PI TOTAL OPEN {best_total_label}",
                len(best_total_open_names),
                best_total_label,
                best_total_line_details,
            )

        if not SAVE_ALL_BUS_SC_DETAILS:
            append_detail_rows(
                bus_sc_detail_rows,
                case_name,
                f"BEST PI TOTAL OPEN {best_total_label}",
                len(best_total_open_names),
                best_total_label,
                best_total_sc_details,
            )

        if not SAVE_ALL_BUS_V_DETAILS:
            append_detail_rows(
                bus_v_detail_rows,
                case_name,
                f"BEST PI TOTAL OPEN {best_total_label}",
                len(best_total_open_names),
                best_total_label,
                best_total_v_details,
            )

    df_all = pd.DataFrame(scenario_rows)
    df_line = pd.DataFrame(line_detail_rows)
    df_bus_sc = pd.DataFrame(bus_sc_detail_rows)
    df_bus_v = pd.DataFrame(bus_v_detail_rows)

    df_open_only = df_all[df_all["Scenario Type"] == "OPEN_SET"]

    df_rank_pipf = build_ranking(df_open_only, "PIpf Total")
    df_rank_pisc = build_ranking(df_open_only, "PIsc Total")
    df_rank_piv = build_ranking(df_open_only, "PIv Total")
    df_rank_total = build_ranking(df_open_only, "PI Total")

    df_best = pd.DataFrame([{
        "Case": case_name,
        "Scenario Mode": SCENARIO_MODE,
        "Weight PF": weight_pf,
        "Weight SC": weight_sc,
        "Weight V": weight_v,
        "Baseline PI Total": round(baseline_pi_total, 8),

        "Best PIpf Opened Couplers": ", ".join(best_pipf_open_names) if best_pipf_open_names else None,
        "Best PIpf N Open": len(best_pipf_open_names) if best_pipf_open_names else None,
        "Best PIpf Total": round(best_pipf_score, 8) if best_pipf_open_names else None,
        "Baseline PIpf Total": round(baseline_pipf, 8) if baseline_pipf is not None else None,

        "Best PIsc Opened Couplers": ", ".join(best_pisc_open_names) if best_pisc_open_names else None,
        "Best PIsc N Open": len(best_pisc_open_names) if best_pisc_open_names else None,
        "Best PIsc Total": round(best_pisc_score, 8) if best_pisc_open_names else None,
        "Baseline PIsc Total": round(baseline_pisc, 8) if baseline_pisc is not None else None,

        "Best PIv Opened Couplers": ", ".join(best_piv_open_names) if best_piv_open_names else None,
        "Best PIv N Open": len(best_piv_open_names) if best_piv_open_names else None,
        "Best PIv Total": round(best_piv_score, 8) if best_piv_open_names else None,
        "Baseline PIv Total": round(baseline_piv, 8) if baseline_piv is not None else None,

        "Best PI Total Opened Couplers": ", ".join(best_total_open_names) if best_total_open_names else None,
        "Best PI Total N Open": len(best_total_open_names) if best_total_open_names else None,
        "Best PI Total": round(best_total_score, 8) if best_total_open_names else None,

        "Execution Time (s)": round(elapsed, 2),
        "Cache Time (s)": round(cache_time, 2),
    }])

    log(app, "===== FINAL RESULT =====")
    log(app, f"Case                   : {case_name}")
    log(app, f"Weights                : PF={weight_pf:.6f}, SC={weight_sc:.6f}, V={weight_v:.6f}")
    log(app, f"Baseline PIpf Total    : {baseline_pipf:.6f}" if baseline_pipf is not None else "Baseline PIpf Total    : N/A")
    log(app, f"Baseline PIsc Total    : {baseline_pisc:.6f}" if baseline_pisc is not None else "Baseline PIsc Total    : N/A")
    log(app, f"Baseline PIv Total     : {baseline_piv:.6f}" if baseline_piv is not None else "Baseline PIv Total     : N/A")
    log(app, f"Best PIpf Couplers     : {', '.join(best_pipf_open_names)}" if best_pipf_open_names else "Best PIpf Couplers     : N/A")
    log(app, f"Best PIsc Couplers     : {', '.join(best_pisc_open_names)}" if best_pisc_open_names else "Best PIsc Couplers     : N/A")
    log(app, f"Best PIv Couplers      : {', '.join(best_piv_open_names)}" if best_piv_open_names else "Best PIv Couplers      : N/A")
    log(app, f"Best PI Total Couplers : {', '.join(best_total_open_names)}" if best_total_open_names else "Best PI Total Couplers : N/A")
    log(app, f"Best PI Total          : {best_total_score:.6f}" if best_total_open_names else "Best PI Total          : N/A")
    log(app, f"Execution Time         : {elapsed:.2f} s")

    return {
        "all": df_all,
        "rank_pipf": df_rank_pipf,
        "rank_pisc": df_rank_pisc,
        "rank_piv": df_rank_piv,
        "rank_total": df_rank_total,
        "best": df_best,
        "line": df_line,
        "bus_sc": df_bus_sc,
        "bus_v": df_bus_v,
    }


def run_sensitivity_study() -> None:
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    summary_rows: List[Dict[str, Any]] = []
    app = get_app()

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        for case_name, matrix in AHP_CASES.items():
            weights, lambda_max, cr = ahp_weights(matrix)
            weight_pf, weight_sc, weight_v = weights

            log(app, f"===== START AHP CASE: {case_name} =====")
            log(app, f"{case_name} lambda_max = {lambda_max:.6f}")
            log(app, f"{case_name} CR         = {cr:.6f}")
            log(app, f"{case_name} weights    = PF={weight_pf:.6f}, SC={weight_sc:.6f}, V={weight_v:.6f}")

            results = run_bruteforce(
                case_name=case_name,
                weight_pf=weight_pf,
                weight_sc=weight_sc,
                weight_v=weight_v,
            )

            results["all"].to_excel(writer, sheet_name=f"{case_name}_All"[:31], index=False)
            results["rank_total"].to_excel(writer, sheet_name=f"{case_name}_RankTot"[:31], index=False)
            results["rank_pipf"].to_excel(writer, sheet_name=f"{case_name}_RankPF"[:31], index=False)
            results["rank_pisc"].to_excel(writer, sheet_name=f"{case_name}_RankSC"[:31], index=False)
            results["rank_piv"].to_excel(writer, sheet_name=f"{case_name}_RankV"[:31], index=False)
            results["best"].to_excel(writer, sheet_name=f"{case_name}_Best"[:31], index=False)

            if not results["line"].empty:
                results["line"].to_excel(writer, sheet_name=f"{case_name}_Line"[:31], index=False)

            if not results["bus_sc"].empty:
                results["bus_sc"].to_excel(writer, sheet_name=f"{case_name}_BusSC"[:31], index=False)

            if not results["bus_v"].empty:
                results["bus_v"].to_excel(writer, sheet_name=f"{case_name}_BusV"[:31], index=False)

            best_row = results["best"].iloc[0].to_dict()
            summary_rows.append({
                "Case": case_name,
                "Weight PF": round(weight_pf, 8),
                "Weight SC": round(weight_sc, 8),
                "Weight V": round(weight_v, 8),
                "Lambda Max": round(lambda_max, 8),
                "CR": round(cr, 8),
                "Best PI Total Opened Couplers": best_row.get("Best PI Total Opened Couplers"),
                "Best PI Total N Open": best_row.get("Best PI Total N Open"),
                "Best PI Total": best_row.get("Best PI Total"),
                "Best PIpf Opened Couplers": best_row.get("Best PIpf Opened Couplers"),
                "Best PIsc Opened Couplers": best_row.get("Best PIsc Opened Couplers"),
                "Best PIv Opened Couplers": best_row.get("Best PIv Opened Couplers"),
                "Execution Time (s)": best_row.get("Execution Time (s)"),
            })

        df_summary = pd.DataFrame(summary_rows)
        df_summary.to_excel(writer, sheet_name="AHP_Summary", index=False)

    log(app, "===== SENSITIVITY STUDY FINISHED =====")
    log(app, f"Saved: {OUTPUT_FILE}")


if __name__ == "__main__":
    run_sensitivity_study()
```

```python
import os
import powerfactory
import pandas as pd

# ==========================
# INPUT
# ==========================
substat_list = [
"KRIAN5", "GRESIK BARU5", "GRESIK LAMA55", "KARANGPILANG5", "DARMO GRANDE5", "TANDES5", "ALTA PRIMA5", "SEGOROMADU5", 
"SAMBIKEREP5", "WILMAR5", "BAMBE5", "SAWAHAN5", "PETRO KIMIA5", "PERAK5", "KUPANG5", "UNDAAN5", "KREMBANGAN5",
"GUNUNG SARI5", "UJUNG5", "INDOFOOD SUKSES MAKMUR5", "RUNGKUT5", "GEMBONG5", "WARU5", "SURABAYA SELATAN5", "SUKOLILO5",
"HANIL JAYA STEEL5", "BUDURAN5", "JATIM TAMAN STEEL5", "KALISARI5", "WONOKROMO5", "NGAGEL5", "KENJERAN5", "NEWBUDURAN5", 
"SIDOARJO5", "SIMPANG5", "SEDATI5", "KEDINDING5", "GILITIMUR5", "BANGKALAN5", "SAMPANG5", "PAMEKASAN5", "GULUKGULUK5", "SUMENEP5"
]

VMIN = 0.95
VMAX = 1.10

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\DIGSILENT DATA\Beban Maksmimal\Skenario 1"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "Bus_Voltage_All_Substations_Skenario_1_Average.xlsx")


# ==========================
# POWERFACTORY INIT
# ==========================
def get_app():
    app = powerfactory.GetApplication()
    if app is None:
        raise RuntimeError("Cannot connect to PowerFactory")
    return app


# ==========================
# RUN LOAD FLOW
# ==========================
def run_loadflow(app):
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        raise RuntimeError("Load flow object 'ComLdf' not found")

    lf.iopt_net = 0
    lf.iopt_o4 = 0

    rc = lf.Execute()
    if rc != 0:
        raise RuntimeError("Load flow failed")

    print("Load Flow Completed\n")


# ==========================
# READ BUS VOLTAGE
# ==========================
def read_bus_voltage(app, substation_names):
    all_subs = app.GetCalcRelevantObjects("*.ElmSubstat")
    sub_map = {s.loc_name.strip(): s for s in all_subs}

    rows = []
    not_found = []

    for sub_name in substation_names:
        sub = sub_map.get(sub_name.strip())

        if sub is None:
            print(f"Substation not found: {sub_name}")
            not_found.append(sub_name)
            continue

        print(f"Reading buses in substation: {sub_name}")

        terms = sub.GetContents("*.ElmTerm", 1)

        for bus in terms:
            try:
                if bus.GetAttribute("iUsage") != 0:
                    continue
            except Exception:
                continue

            bus_name = bus.loc_name.strip()
            vpu = bus.GetAttribute("m:u")
            vkv = bus.GetAttribute("m:Ul")

            if vpu is None or vkv is None:
                status = "NO_RESULT"
            elif vpu < VMIN:
                status = "LOW"
            elif vpu > VMAX:
                status = "HIGH"
            else:
                status = "NORMAL"

            rows.append({
                "Substation": sub_name,
                "Bus Name": bus_name,
                "Voltage (pu)": round(vpu, 6) if vpu is not None else None,
                "Voltage (kV)": round(vkv, 3) if vkv is not None else None,
                "Status": status
            })

    return rows, not_found


# ==========================
# SAVE TO EXCEL
# ==========================
def save_to_excel(rows, not_found, output_file):
    os.makedirs(os.path.dirname(output_file), exist_ok=True)

    df = pd.DataFrame(rows)

    if not df.empty:
        summary = (
            df.groupby(["Substation", "Status"])
            .size()
            .reset_index(name="Count")
            .sort_values(["Substation", "Status"])
        )
    else:
        summary = pd.DataFrame(columns=["Substation", "Status", "Count"])

    df_not_found = pd.DataFrame({"Substation Not Found": not_found})

    with pd.ExcelWriter(output_file) as writer:
        df.to_excel(writer, sheet_name="Bus Voltage", index=False)
        summary.to_excel(writer, sheet_name="Summary", index=False)
        df_not_found.to_excel(writer, sheet_name="Not Found", index=False)

    print(f"\nSaved to Excel: {output_file}")


# ==========================
# MAIN
# ==========================
def main():
    app = get_app()
    run_loadflow(app)

    rows, not_found = read_bus_voltage(app, substat_list)
    save_to_excel(rows, not_found, OUTPUT_FILE)

    print(f"Total buses recorded: {len(rows)}")
    print(f"Substations not found: {len(not_found)}")
    print(f"Voltage limits used: VMIN = {VMIN:.2f}, VMAX = {VMAX:.2f}")


if __name__ == "__main__":
    main()
```

```python
import os
import math
import random
import powerfactory
import pandas as pd

app = powerfactory.GetApplication()

# ── CONFIG ─────────────────────────────
BAY_LIST = [
    "sawahan-krian1", "sawahan-krian2", "sawahan_td1", "sawahan_td2", "sawahan_td3",
    "swhan-gnsri1", "swhan-gnsri2", "swhan-kpang1", "swhan-kpang2",
    "swhan-krban1", "swhan-krban2", "swhan-tandes1", "swhan-tandes2",
    "swhan-udaan1", "swhan-udaan2"
]

POP_SIZE = 40
N_GEN = 1
MUT_PROB = 0.02
CROSSOVER_PROB = 0.80
ELITE_SIZE = 1

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\DIGSILENT DATA"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "GA_DS_Sawahan_Result.xlsx")

# 1 = Bus A  -> DSA close, DSB open
# 0 = Bus B  -> DSA open,  DSB close


# ── BUILD DS MAP ───────────────────────
def build_ds_map():
    ds_map = {}
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        ds_map[obj.loc_name.lower()] = obj
    return ds_map


# ── SAVE ORIGINAL STATE ────────────────
def save_original_state(ds_map):
    original = {}
    for bay in BAY_LIST:
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa in ds_map:
            original[dsa] = ds_map[dsa].IsOpen()
        if dsb in ds_map:
            original[dsb] = ds_map[dsb].IsOpen()

    return original


# ── RESTORE ORIGINAL STATE ─────────────
def restore_original_state(ds_map, original):
    for name, is_open in original.items():
        if name not in ds_map:
            continue

        obj = ds_map[name]
        if is_open:
            obj.Open()
        else:
            obj.Close()


# ── APPLY CONFIG ───────────────────────
def apply_config(ds_map, ind):
    for i, bay in enumerate(BAY_LIST):
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa not in ds_map or dsb not in ds_map:
            continue

        if ind[i] == 1:
            ds_map[dsa].Close()
            ds_map[dsb].Open()
        else:
            ds_map[dsa].Open()
            ds_map[dsb].Close()


# ── LOADFLOW ───────────────────────────
def run_lf():
    app.Rebuild()
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        return False
    return lf.Execute() == 0


# ── GET BC ─────────────────────────────
def get_bc():
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        if obj.loc_name.lower() == "bc sawahan":
            return obj
    return None


# ── FITNESS ────────────────────────────
def fitness(ds_map, bc, ind):
    apply_config(ds_map, ind)

    if not run_lf():
        return 9999.0, None, None

    p = bc.GetAttribute("m:P:bus1")
    q = bc.GetAttribute("m:Q:bus1")

    if p is None or q is None:
        return 9999.0, None, None

    s = math.sqrt(p * p + q * q)
    return s, p, q


# ── INIT POP ───────────────────────────
def init_pop(n):
    return [[random.randint(0, 1) for _ in range(n)] for _ in range(POP_SIZE)]


# ── MUTATION ───────────────────────────
def mutate(ind):
    return [bit ^ 1 if random.random() < MUT_PROB else bit for bit in ind]


# ── CROSSOVER ──────────────────────────
def crossover(a, b):
    if len(a) <= 1:
        return a[:], b[:]

    if random.random() > CROSSOVER_PROB:
        return a[:], b[:]

    pt = random.randint(1, len(a) - 1)
    return a[:pt] + b[pt:], b[:pt] + a[pt:]


# ── FORMAT BINARY ──────────────────────
def format_binary(ind):
    return "".join(str(bit) for bit in ind)


# ── BUILD CONFIG DETAIL ────────────────
def build_config_detail(ind):
    rows = []
    for i, bay in enumerate(BAY_LIST):
        bit = ind[i]

        if bit == 1:
            bus = "Bus A"
            dsa_state = "CLOSE"
            dsb_state = "OPEN"
        else:
            bus = "Bus B"
            dsa_state = "OPEN"
            dsb_state = "CLOSE"

        rows.append({
            "No": i + 1,
            "Bay": bay,
            "Bit": bit,
            "Bus": bus,
            "DSA": dsa_state,
            "DSB": dsb_state
        })
    return rows


# ── SAVE TO EXCEL ──────────────────────
def save_to_excel(gen_rows, all_individual_rows, elite_rows, best_summary, best_config_rows):
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    df_gen = pd.DataFrame(gen_rows)
    df_all = pd.DataFrame(all_individual_rows)
    df_elite = pd.DataFrame(elite_rows)
    df_best = pd.DataFrame([best_summary])
    df_cfg = pd.DataFrame(best_config_rows)
    df_param = pd.DataFrame([
        {"Parameter": "POP_SIZE", "Value": POP_SIZE},
        {"Parameter": "N_GEN", "Value": N_GEN},
        {"Parameter": "MUT_PROB", "Value": MUT_PROB},
        {"Parameter": "CROSSOVER_PROB", "Value": CROSSOVER_PROB},
        {"Parameter": "ELITE_SIZE", "Value": ELITE_SIZE},
        {"Parameter": "N_BAY", "Value": len(BAY_LIST)},
        {"Parameter": "BC_TARGET", "Value": "BC Sawahan"},
        {"Parameter": "Binary Meaning", "Value": "1=Bus A, 0=Bus B"},
        {"Parameter": "Output File", "Value": OUTPUT_FILE},
    ])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df_gen.to_excel(writer, sheet_name="Generation History", index=False)
        df_all.to_excel(writer, sheet_name="All Individuals", index=False)
        df_elite.to_excel(writer, sheet_name="Elite History", index=False)
        df_best.to_excel(writer, sheet_name="Best Summary", index=False)
        df_cfg.to_excel(writer, sheet_name="Best Config Detail", index=False)
        df_param.to_excel(writer, sheet_name="GA Parameters", index=False)


# ── TOURNAMENT SELECTION ───────────────
def tournament_select(pop, scores):
    i, j = random.sample(range(len(pop)), 2)
    return pop[i][:] if scores[i] < scores[j] else pop[j][:]


# ── MAIN GA ────────────────────────────
def run():
    if app is None:
        raise RuntimeError("Cannot connect to PowerFactory")

    if ELITE_SIZE < 0 or ELITE_SIZE >= POP_SIZE:
        raise ValueError("ELITE_SIZE must be >= 0 and < POP_SIZE")

    ds_map = build_ds_map()
    bc = get_bc()

    if bc is None:
        raise RuntimeError("BC Sawahan tidak ditemukan")

    original_state = save_original_state(ds_map)
    pop = init_pop(len(BAY_LIST))

    best = None
    best_score = 1e9
    best_p = None
    best_q = None

    gen_rows = []
    all_individual_rows = []
    elite_rows = []

    try:
        for gen in range(N_GEN):
            evaluated = []

            for idx, ind in enumerate(pop, start=1):
                s, p, q = fitness(ds_map, bc, ind)
                evaluated.append((ind[:], s, p, q))

                all_individual_rows.append({
                    "Generation": gen + 1,
                    "Individual No": idx,
                    "Config List": str(ind),
                    "Config Binary": format_binary(ind),
                    "S": round(s, 6),
                    "P BC (MW)": round(p, 6) if p is not None else None,
                    "Q BC (Mvar)": round(q, 6) if q is not None else None
                })

            ranked = sorted(evaluated, key=lambda x: x[1])

            gen_best_ind = ranked[0][0][:]
            gen_best_score = ranked[0][1]
            gen_best_p = ranked[0][2]
            gen_best_q = ranked[0][3]

            if gen_best_score < best_score:
                best_score = gen_best_score
                best = gen_best_ind[:]
                best_p = gen_best_p
                best_q = gen_best_q

            gen_rows.append({
                "Generation": gen + 1,
                "Generation Best S": round(gen_best_score, 6),
                "Generation Best Binary": format_binary(gen_best_ind),
                "Generation Best P BC (MW)": round(gen_best_p, 6) if gen_best_p is not None else None,
                "Generation Best Q BC (Mvar)": round(gen_best_q, 6) if gen_best_q is not None else None,
                "Global Best S": round(best_score, 6),
                "Global Best Binary": format_binary(best),
                "Global Best P BC (MW)": round(best_p, 6) if best_p is not None else None,
                "Global Best Q BC (Mvar)": round(best_q, 6) if best_q is not None else None
            })

            # Elite History
            for elite_rank, (elite_ind, elite_s, elite_p, elite_q) in enumerate(ranked[:ELITE_SIZE], start=1):
                elite_rows.append({
                    "Generation": gen + 1,
                    "Elite Rank": elite_rank,
                    "Config List": str(elite_ind),
                    "Config Binary": format_binary(elite_ind),
                    "S": round(elite_s, 6),
                    "P BC (MW)": round(elite_p, 6) if elite_p is not None else None,
                    "Q BC (Mvar)": round(elite_q, 6) if elite_q is not None else None
                })

            print(
                f"Gen {gen+1:>2} | "
                f"Gen Best = {gen_best_score:.4f} | "
                f"Global Best = {best_score:.4f} | "
                f"Binary = {format_binary(best)}"
            )

            scores = [item[1] for item in evaluated]
            pop_only = [item[0][:] for item in evaluated]

            # ── ELITISM ──
            new_pop = [ind[:] for ind, _, _, _ in ranked[:ELITE_SIZE]]

            # ── Fill rest with selection + crossover + mutation ──
            while len(new_pop) < POP_SIZE:
                parent1 = tournament_select(pop_only, scores)
                parent2 = tournament_select(pop_only, scores)

                child1, child2 = crossover(parent1, parent2)

                child1 = mutate(child1)
                new_pop.append(child1)

                if len(new_pop) < POP_SIZE:
                    child2 = mutate(child2)
                    new_pop.append(child2)

            pop = new_pop[:POP_SIZE]

        print("\n=== HASIL TERBAIK ===")
        print("Config List :", best)
        print("Config Bin  :", format_binary(best))
        print("S           :", round(best_score, 4))
        print("P BC        :", round(best_p, 4) if best_p is not None else None)
        print("Q BC        :", round(best_q, 4) if best_q is not None else None)
        print("\nKeterangan  : 1 = Bus A, 0 = Bus B")

        best_summary = {
            "Best Config List": str(best),
            "Best Config Binary": format_binary(best),
            "Best S": round(best_score, 6),
            "Best P BC (MW)": round(best_p, 6) if best_p is not None else None,
            "Best Q BC (Mvar)": round(best_q, 6) if best_q is not None else None,
            "BC Target": "BC Sawahan",
            "Meaning": "1=Bus A, 0=Bus B"
        }

        best_config_rows = build_config_detail(best)

        save_to_excel(gen_rows, all_individual_rows, elite_rows, best_summary, best_config_rows)
        print("\nHasil berhasil disimpan ke Excel:")
        print(OUTPUT_FILE)

    finally:
        restore_original_state(ds_map, original_state)
        print("\nKondisi DS sudah dikembalikan ke kondisi awal.")


if __name__ == "__main__":
    run()
```

```python
import os
import math
from itertools import product

import powerfactory
import pandas as pd

app = powerfactory.GetApplication()

# ── CONFIG ─────────────────────────────
BAY_LIST = [
    "sawahan-krian1", "sawahan-krian2", "sawahan_td1", "sawahan_td2", "sawahan_td3",
    "swhan-gnsri1", "swhan-gnsri2", "swhan-kpang1", "swhan-kpang2",
    "swhan-krban1", "swhan-krban2", "swhan-tandes1", "swhan-tandes2",
    "swhan-udaan1", "swhan-udaan2"
]

# Constraint target GI
TARGET_SUB_NAME = "SAWAHAN5"
VMIN = 0.95
VMAX = 1.10
MAX_LINE_LOADING = 80.0   # %

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\DIGSILENT DATA"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "BF_DS_Sawahan_Result.xlsx")

# 1 = Bus A  -> DSA close, DSB open
# 0 = Bus B  -> DSA open,  DSB close


# ── BUILD DS MAP ───────────────────────
def build_ds_map():
    ds_map = {}
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        ds_map[obj.loc_name.lower()] = obj
    return ds_map


# ── SAVE ORIGINAL STATE ────────────────
def save_original_state(ds_map):
    original = {}
    for bay in BAY_LIST:
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa in ds_map:
            original[dsa] = ds_map[dsa].IsOpen()
        if dsb in ds_map:
            original[dsb] = ds_map[dsb].IsOpen()

    return original


# ── RESTORE ORIGINAL STATE ─────────────
def restore_original_state(ds_map, original):
    for name, is_open in original.items():
        if name not in ds_map:
            continue

        obj = ds_map[name]
        if is_open:
            obj.Open()
        else:
            obj.Close()


# ── APPLY CONFIG ───────────────────────
def apply_config(ds_map, ind):
    for i, bay in enumerate(BAY_LIST):
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa not in ds_map or dsb not in ds_map:
            continue

        if ind[i] == 1:
            ds_map[dsa].Close()
            ds_map[dsb].Open()
        else:
            ds_map[dsa].Open()
            ds_map[dsb].Close()


# ── LOADFLOW ───────────────────────────
def run_lf():
    app.Rebuild()
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        return False
    return lf.Execute() == 0


# ── GET BC ─────────────────────────────
def get_bc():
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        if obj.loc_name.lower() == "bc sawahan":
            return obj
    return None


# ── GET TARGET SUBSTATION ──────────────
def get_target_substation():
    for sub in app.GetCalcRelevantObjects("*.ElmSubstat"):
        if sub.loc_name.strip().lower() == TARGET_SUB_NAME.lower():
            return sub
    return None


# ── GET TARGET BUSES ───────────────────
def get_target_buses(target_sub):
    buses = []
    terms = target_sub.GetContents("*.ElmTerm", 1)

    for bus in terms:
        try:
            if bus.GetAttribute("iUsage") != 0:
                continue
        except Exception:
            continue
        buses.append(bus)

    return buses


# ── CHECK CONSTRAINTS ──────────────────
def check_constraints(target_buses, all_lines):
    voltage_violation = False
    loading_violation = False

    min_v = None
    max_v = None
    max_loading = 0.0

    # Voltage constraint on buses in target GI
    for bus in target_buses:
        try:
            vpu = bus.GetAttribute("m:u")
        except Exception:
            continue

        if vpu is None:
            continue

        if min_v is None or vpu < min_v:
            min_v = vpu
        if max_v is None or vpu > max_v:
            max_v = vpu

        if vpu < VMIN or vpu > VMAX:
            voltage_violation = True

    # Line loading constraint on lines connected to target GI
    for line in all_lines:
        try:
            bus1 = line.bus1.cterm
            bus2 = line.bus2.cterm
        except Exception:
            continue

        if bus1 in target_buses or bus2 in target_buses:
            try:
                loading = line.GetAttribute("m:loading")
            except Exception:
                continue

            if loading is None:
                continue

            if loading > max_loading:
                max_loading = loading

            if loading >= MAX_LINE_LOADING:
                loading_violation = True

    reject_reasons = []
    if voltage_violation:
        reject_reasons.append("VOLTAGE_LIMIT")
    if loading_violation:
        reject_reasons.append("LINE_LOADING_LIMIT")

    feasible = not voltage_violation and not loading_violation
    reject_reason = "OK" if feasible else ";".join(reject_reasons)

    return feasible, reject_reason, max_loading, min_v, max_v


# ── EVALUATE CONFIG ────────────────────
def evaluate_config(ds_map, bc, target_buses, all_lines, ind):
    apply_config(ds_map, ind)

    if not run_lf():
        return 9999.0, None, None, False, "LOADFLOW_FAILED", None, None, None

    p = bc.GetAttribute("m:P:bus1")
    q = bc.GetAttribute("m:Q:bus1")

    if p is None or q is None:
        return 9999.0, None, None, False, "BC_RESULT_MISSING", None, None, None

    feasible, reject_reason, max_loading, min_v, max_v = check_constraints(target_buses, all_lines)

    if not feasible:
        return 9999.0, p, q, False, reject_reason, max_loading, min_v, max_v

    s = math.sqrt(p * p + q * q)
    return s, p, q, True, reject_reason, max_loading, min_v, max_v


# ── FORMAT BINARY ──────────────────────
def format_binary(ind):
    return "".join(str(bit) for bit in ind)


# ── BUILD CONFIG DETAIL ────────────────
def build_config_detail(ind):
    rows = []
    for i, bay in enumerate(BAY_LIST):
        bit = ind[i]

        if bit == 1:
            bus = "Bus A"
            dsa_state = "CLOSE"
            dsb_state = "OPEN"
        else:
            bus = "Bus B"
            dsa_state = "OPEN"
            dsb_state = "CLOSE"

        rows.append({
            "No": i + 1,
            "Bay": bay,
            "Bit": bit,
            "Bus": bus,
            "DSA": dsa_state,
            "DSB": dsb_state
        })
    return rows


# ── SAVE TO EXCEL ──────────────────────
def save_to_excel(all_rows, best_summary, best_config_rows):
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    df_all = pd.DataFrame(all_rows)
    df_best = pd.DataFrame([best_summary])
    df_cfg = pd.DataFrame(best_config_rows)

    if not df_all.empty:
        df_rank = df_all.sort_values(
            by=["S", "Feasible", "Max Loading (%)"],
            ascending=[True, False, True]
        ).reset_index(drop=True)
        df_rank.insert(0, "Rank", range(1, len(df_rank) + 1))

        reject_summary = (
            df_all.groupby(["Feasible", "Reject Reason"])
            .size()
            .reset_index(name="Count")
            .sort_values(["Feasible", "Reject Reason"])
        )
    else:
        df_rank = pd.DataFrame()
        reject_summary = pd.DataFrame()

    df_param = pd.DataFrame([
        {"Parameter": "METHOD", "Value": "BRUTE FORCE"},
        {"Parameter": "TOTAL_CONFIG", "Value": 2 ** len(BAY_LIST)},
        {"Parameter": "TARGET_SUB_NAME", "Value": TARGET_SUB_NAME},
        {"Parameter": "VMIN", "Value": VMIN},
        {"Parameter": "VMAX", "Value": VMAX},
        {"Parameter": "MAX_LINE_LOADING", "Value": MAX_LINE_LOADING},
        {"Parameter": "N_BAY", "Value": len(BAY_LIST)},
        {"Parameter": "BC_TARGET", "Value": "BC Sawahan"},
        {"Parameter": "Binary Meaning", "Value": "1=Bus A, 0=Bus B"},
        {"Parameter": "Reject Rule", "Value": "Loading>=80% or V outside 0.95-1.10 pu"},
        {"Parameter": "Output File", "Value": OUTPUT_FILE},
    ])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df_all.to_excel(writer, sheet_name="All Configurations", index=False)
        df_rank.to_excel(writer, sheet_name="Ranking", index=False)
        df_best.to_excel(writer, sheet_name="Best Summary", index=False)
        df_cfg.to_excel(writer, sheet_name="Best Config Detail", index=False)
        reject_summary.to_excel(writer, sheet_name="Reject Summary", index=False)
        df_param.to_excel(writer, sheet_name="Parameters", index=False)


# ── MAIN BRUTE FORCE ───────────────────
def run():
    if app is None:
        raise RuntimeError("Cannot connect to PowerFactory")

    ds_map = build_ds_map()
    bc = get_bc()
    target_sub = get_target_substation()

    if bc is None:
        raise RuntimeError("BC Sawahan tidak ditemukan")

    if target_sub is None:
        raise RuntimeError(f"Target substation '{TARGET_SUB_NAME}' tidak ditemukan")

    target_buses = get_target_buses(target_sub)
    if not target_buses:
        raise RuntimeError(f"Tidak ada bus yang valid di GI '{TARGET_SUB_NAME}'")

    all_lines = app.GetCalcRelevantObjects("*.ElmLne")
    original_state = save_original_state(ds_map)

    best = None
    best_score = 1e9
    best_p = None
    best_q = None
    best_feasible = False
    best_reject_reason = None
    best_max_loading = None
    best_min_v = None
    best_max_v = None

    all_rows = []
    total_config = 2 ** len(BAY_LIST)

    try:
        for idx, ind_tuple in enumerate(product([0, 1], repeat=len(BAY_LIST)), start=1):
            ind = list(ind_tuple)

            s, p, q, feasible, reject_reason, max_loading, min_v, max_v = evaluate_config(
                ds_map, bc, target_buses, all_lines, ind
            )

            bin_str = format_binary(ind)

            all_rows.append({
                "Config No": idx,
                "Config List": str(ind),
                "Config Binary": bin_str,
                "S": round(s, 6),
                "P BC (MW)": round(p, 6) if p is not None else None,
                "Q BC (Mvar)": round(q, 6) if q is not None else None,
                "Feasible": feasible,
                "Reject Reason": reject_reason,
                "Max Loading (%)": round(max_loading, 6) if max_loading is not None else None,
                "Min Vpu": round(min_v, 6) if min_v is not None else None,
                "Max Vpu": round(max_v, 6) if max_v is not None else None
            })

            if s < best_score:
                best_score = s
                best = ind[:]
                best_p = p
                best_q = q
                best_feasible = feasible
                best_reject_reason = reject_reason
                best_max_loading = max_loading
                best_min_v = min_v
                best_max_v = max_v

            if idx % 500 == 0 or idx == total_config:
                print(
                    f"[{idx}/{total_config}] "
                    f"Best S = {best_score:.4f} | "
                    f"Best Binary = {format_binary(best)}"
                )

        print("\n=== HASIL TERBAIK ===")
        print("Config List :", best)
        print("Config Bin  :", format_binary(best))
        print("S           :", round(best_score, 4))
        print("P BC        :", round(best_p, 4) if best_p is not None else None)
        print("Q BC        :", round(best_q, 4) if best_q is not None else None)
        print("Feasible    :", best_feasible)
        print("Reject      :", best_reject_reason)
        print("Max Loading :", round(best_max_loading, 4) if best_max_loading is not None else None)
        print("Min Vpu     :", round(best_min_v, 4) if best_min_v is not None else None)
        print("Max Vpu     :", round(best_max_v, 4) if best_max_v is not None else None)
        print("\nKeterangan  : 1 = Bus A, 0 = Bus B")

        best_summary = {
            "Best Config List": str(best),
            "Best Config Binary": format_binary(best),
            "Best S": round(best_score, 6),
            "Best P BC (MW)": round(best_p, 6) if best_p is not None else None,
            "Best Q BC (Mvar)": round(best_q, 6) if best_q is not None else None,
            "Feasible": best_feasible,
            "Reject Reason": best_reject_reason,
            "Max Loading (%)": round(best_max_loading, 6) if best_max_loading is not None else None,
            "Min Vpu": round(best_min_v, 6) if best_min_v is not None else None,
            "Max Vpu": round(best_max_v, 6) if best_max_v is not None else None,
            "BC Target": "BC Sawahan",
            "Target GI": TARGET_SUB_NAME,
            "Meaning": "1=Bus A, 0=Bus B"
        }

        best_config_rows = build_config_detail(best)

        save_to_excel(all_rows, best_summary, best_config_rows)
        print("\nHasil berhasil disimpan ke Excel:")
        print(OUTPUT_FILE)

    finally:
        restore_original_state(ds_map, original_state)
        print("\nKondisi DS sudah dikembalikan ke kondisi awal.")


if __name__ == "__main__":
    run()
```

```python
import os
import math
import random
import powerfactory
import pandas as pd

app = powerfactory.GetApplication()

# ── CONFIG ─────────────────────────────
BAY_LIST = [
    "waru-ptjts",
    "waru-nbdrn",
    "waru-bdran 2",
    "4waru5_td3",
    "4waru5_td4",
    "4waru5_td5",
    "4waru5_td6",
    "darmo-waru 2",
    "darmo-waru 1",
    "waru-gnsri 1",
    "waru-gnsri 2",
    "smkrp-waru#1",
    "smkrp-waru#2",
    "4waru5_td7",
    "waru-rnkut 1",
    "waru-rnkut 2",
    "nwaru-waru1",
    "nwaru-waru2"
]

POP_SIZE = 40
N_GEN = 5
MUT_PROB = 0.02
CROSSOVER_PROB = 0.80
ELITE_SIZE = 1

# Constraint target GI
TARGET_SUB_NAME = "WARU5"
VMIN = 0.95
VMAX = 1.10
MAX_LINE_LOADING = 80.0   # %

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\DIGSILENT DATA"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "GA_DS_Waru_Result.xlsx")

# 1 = Bus A  -> DSA close, DSB open
# 0 = Bus B  -> DSA open,  DSB close

# Kondisi awal dari tabel (opsional untuk verifikasi)
INITIAL_BINARY_FROM_TABLE = "101000000011101000"


# ── BUILD DS MAP ───────────────────────
def build_ds_map():
    ds_map = {}
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        ds_map[obj.loc_name.lower().strip()] = obj
    return ds_map


# ── SAVE ORIGINAL STATE ────────────────
def save_original_state(ds_map):
    original = {}
    for bay in BAY_LIST:
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa in ds_map:
            original[dsa] = ds_map[dsa].IsOpen()
        if dsb in ds_map:
            original[dsb] = ds_map[dsb].IsOpen()

    return original


# ── RESTORE ORIGINAL STATE ─────────────
def restore_original_state(ds_map, original):
    for name, is_open in original.items():
        if name not in ds_map:
            continue

        obj = ds_map[name]
        if is_open:
            obj.Open()
        else:
            obj.Close()


# ── APPLY CONFIG ───────────────────────
def apply_config(ds_map, ind):
    for i, bay in enumerate(BAY_LIST):
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa not in ds_map or dsb not in ds_map:
            continue

        if ind[i] == 1:
            ds_map[dsa].Close()
            ds_map[dsb].Open()
        else:
            ds_map[dsa].Open()
            ds_map[dsb].Close()


# ── LOADFLOW ───────────────────────────
def run_lf():
    app.Rebuild()
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        return False
    return lf.Execute() == 0


# ── GET BC ─────────────────────────────
def get_bc():
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        if obj.loc_name.lower().strip() == "bc waru":
            return obj
    return None


# ── GET TARGET SUBSTATION ──────────────
def get_target_substation():
    for sub in app.GetCalcRelevantObjects("*.ElmSubstat"):
        if sub.loc_name.strip().lower() == TARGET_SUB_NAME.lower():
            return sub
    return None


# ── GET TARGET BUSES ───────────────────
def get_target_buses(target_sub):
    buses = []
    terms = target_sub.GetContents("*.ElmTerm", 1)

    for bus in terms:
        try:
            if bus.GetAttribute("iUsage") != 0:
                continue
        except Exception:
            continue
        buses.append(bus)

    return buses


# ── CHECK CONSTRAINTS ──────────────────
def check_constraints(target_buses, all_lines):
    voltage_violation = False
    loading_violation = False

    min_v = None
    max_v = None
    max_loading = 0.0

    # 1) Voltage constraint on buses in target GI
    for bus in target_buses:
        try:
            vpu = bus.GetAttribute("m:u")
        except Exception:
            continue

        if vpu is None:
            continue

        if min_v is None or vpu < min_v:
            min_v = vpu
        if max_v is None or vpu > max_v:
            max_v = vpu

        if vpu < VMIN or vpu > VMAX:
            voltage_violation = True

    # 2) Line loading constraint on lines connected to target GI
    for line in all_lines:
        try:
            bus1 = line.bus1.cterm
            bus2 = line.bus2.cterm
        except Exception:
            continue

        if bus1 in target_buses or bus2 in target_buses:
            try:
                loading = line.GetAttribute("m:loading")
            except Exception:
                continue

            if loading is None:
                continue

            if loading > max_loading:
                max_loading = loading

            if loading >= MAX_LINE_LOADING:
                loading_violation = True

    reject_reasons = []
    if voltage_violation:
        reject_reasons.append("VOLTAGE_LIMIT")
    if loading_violation:
        reject_reasons.append("LINE_LOADING_LIMIT")

    feasible = not voltage_violation and not loading_violation
    reject_reason = "OK" if feasible else ";".join(reject_reasons)

    return feasible, reject_reason, max_loading, min_v, max_v


# ── FITNESS ────────────────────────────
def fitness(ds_map, bc, target_buses, all_lines, ind):
    apply_config(ds_map, ind)

    if not run_lf():
        return 9999.0, None, None, False, "LOADFLOW_FAILED", None, None, None

    p = bc.GetAttribute("m:P:bus1")
    q = bc.GetAttribute("m:Q:bus1")

    if p is None or q is None:
        return 9999.0, None, None, False, "BC_RESULT_MISSING", None, None, None

    feasible, reject_reason, max_loading, min_v, max_v = check_constraints(target_buses, all_lines)

    if not feasible:
        return 9999.0, p, q, False, reject_reason, max_loading, min_v, max_v

    s = math.sqrt(p * p + q * q)
    return s, p, q, True, reject_reason, max_loading, min_v, max_v


# ── INIT POP ───────────────────────────
def init_pop(n):
    return [[random.randint(0, 1) for _ in range(n)] for _ in range(POP_SIZE)]


# ── MUTATION ───────────────────────────
def mutate(ind):
    return [bit ^ 1 if random.random() < MUT_PROB else bit for bit in ind]


# ── CROSSOVER ──────────────────────────
def crossover(a, b):
    if len(a) <= 1:
        return a[:], b[:]

    if random.random() > CROSSOVER_PROB:
        return a[:], b[:]

    pt = random.randint(1, len(a) - 1)
    return a[:pt] + b[pt:], b[:pt] + a[pt:]


# ── FORMAT BINARY ──────────────────────
def format_binary(ind):
    return "".join(str(bit) for bit in ind)


# ── BUILD CONFIG DETAIL ────────────────
def build_config_detail(ind):
    rows = []
    for i, bay in enumerate(BAY_LIST):
        bit = ind[i]

        if bit == 1:
            bus = "Bus A"
            dsa_state = "CLOSE"
            dsb_state = "OPEN"
        else:
            bus = "Bus B"
            dsa_state = "OPEN"
            dsb_state = "CLOSE"

        rows.append({
            "No": i + 1,
            "Bay": bay,
            "Bit": bit,
            "Bus": bus,
            "DSA": dsa_state,
            "DSB": dsb_state
        })
    return rows


# ── SAVE TO EXCEL ──────────────────────
def save_to_excel(gen_rows, all_individual_rows, elite_rows, best_summary, best_config_rows):
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    df_gen = pd.DataFrame(gen_rows)
    df_all = pd.DataFrame(all_individual_rows)
    df_elite = pd.DataFrame(elite_rows)
    df_best = pd.DataFrame([best_summary])
    df_cfg = pd.DataFrame(best_config_rows)
    df_param = pd.DataFrame([
        {"Parameter": "POP_SIZE", "Value": POP_SIZE},
        {"Parameter": "N_GEN", "Value": N_GEN},
        {"Parameter": "MUT_PROB", "Value": MUT_PROB},
        {"Parameter": "CROSSOVER_PROB", "Value": CROSSOVER_PROB},
        {"Parameter": "ELITE_SIZE", "Value": ELITE_SIZE},
        {"Parameter": "TARGET_SUB_NAME", "Value": TARGET_SUB_NAME},
        {"Parameter": "VMIN", "Value": VMIN},
        {"Parameter": "VMAX", "Value": VMAX},
        {"Parameter": "MAX_LINE_LOADING", "Value": MAX_LINE_LOADING},
        {"Parameter": "N_BAY", "Value": len(BAY_LIST)},
        {"Parameter": "BC_TARGET", "Value": "BC Waru"},
        {"Parameter": "Binary Meaning", "Value": "1=Bus A, 0=Bus B"},
        {"Parameter": "Initial Binary From Table", "Value": INITIAL_BINARY_FROM_TABLE},
        {"Parameter": "Reject Rule", "Value": "Loading>=80% or V outside 0.95-1.10 pu"},
        {"Parameter": "Output File", "Value": OUTPUT_FILE},
    ])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df_gen.to_excel(writer, sheet_name="Generation History", index=False)
        df_all.to_excel(writer, sheet_name="All Individuals", index=False)
        df_elite.to_excel(writer, sheet_name="Elite History", index=False)
        df_best.to_excel(writer, sheet_name="Best Summary", index=False)
        df_cfg.to_excel(writer, sheet_name="Best Config Detail", index=False)
        df_param.to_excel(writer, sheet_name="GA Parameters", index=False)


# ── TOURNAMENT SELECTION ───────────────
def tournament_select(pop, scores):
    i, j = random.sample(range(len(pop)), 2)
    return pop[i][:] if scores[i] < scores[j] else pop[j][:]


# ── MAIN GA ────────────────────────────
def run():
    if app is None:
        raise RuntimeError("Cannot connect to PowerFactory")

    if ELITE_SIZE < 0 or ELITE_SIZE >= POP_SIZE:
        raise ValueError("ELITE_SIZE must be >= 0 and < POP_SIZE")

    ds_map = build_ds_map()
    bc = get_bc()
    target_sub = get_target_substation()

    if bc is None:
        raise RuntimeError("BC Waru tidak ditemukan")

    if target_sub is None:
        raise RuntimeError(f"Target substation '{TARGET_SUB_NAME}' tidak ditemukan")

    target_buses = get_target_buses(target_sub)
    if not target_buses:
        raise RuntimeError(f"Tidak ada bus yang valid di GI '{TARGET_SUB_NAME}'")

    all_lines = app.GetCalcRelevantObjects("*.ElmLne")
    original_state = save_original_state(ds_map)
    pop = init_pop(len(BAY_LIST))

    best = None
    best_score = 1e9
    best_p = None
    best_q = None
    best_feasible = False
    best_reject_reason = None
    best_max_loading = None
    best_min_v = None
    best_max_v = None

    gen_rows = []
    all_individual_rows = []
    elite_rows = []

    try:
        for gen in range(N_GEN):
            evaluated = []

            for idx, ind in enumerate(pop, start=1):
                s, p, q, feasible, reject_reason, max_loading, min_v, max_v = fitness(
                    ds_map, bc, target_buses, all_lines, ind
                )

                evaluated.append((ind[:], s, p, q, feasible, reject_reason, max_loading, min_v, max_v))

                all_individual_rows.append({
                    "Generation": gen + 1,
                    "Individual No": idx,
                    "Config List": str(ind),
                    "Config Binary": format_binary(ind),
                    "S": round(s, 6),
                    "P BC (MW)": round(p, 6) if p is not None else None,
                    "Q BC (Mvar)": round(q, 6) if q is not None else None,
                    "Feasible": feasible,
                    "Reject Reason": reject_reason,
                    "Max Loading (%)": round(max_loading, 6) if max_loading is not None else None,
                    "Min Vpu": round(min_v, 6) if min_v is not None else None,
                    "Max Vpu": round(max_v, 6) if max_v is not None else None
                })

            ranked = sorted(evaluated, key=lambda x: x[1])

            gen_best_ind = ranked[0][0][:]
            gen_best_score = ranked[0][1]
            gen_best_p = ranked[0][2]
            gen_best_q = ranked[0][3]
            gen_best_feasible = ranked[0][4]
            gen_best_reject_reason = ranked[0][5]
            gen_best_loading = ranked[0][6]
            gen_best_min_v = ranked[0][7]
            gen_best_max_v = ranked[0][8]

            if gen_best_score < best_score:
                best_score = gen_best_score
                best = gen_best_ind[:]
                best_p = gen_best_p
                best_q = gen_best_q
                best_feasible = gen_best_feasible
                best_reject_reason = gen_best_reject_reason
                best_max_loading = gen_best_loading
                best_min_v = gen_best_min_v
                best_max_v = gen_best_max_v

            feasible_count = sum(1 for item in evaluated if item[4])
            reject_count = len(evaluated) - feasible_count

            gen_rows.append({
                "Generation": gen + 1,
                "Generation Best S": round(gen_best_score, 6),
                "Generation Best Binary": format_binary(gen_best_ind),
                "Generation Best P BC (MW)": round(gen_best_p, 6) if gen_best_p is not None else None,
                "Generation Best Q BC (Mvar)": round(gen_best_q, 6) if gen_best_q is not None else None,
                "Generation Best Feasible": gen_best_feasible,
                "Generation Best Reject Reason": gen_best_reject_reason,
                "Generation Max Loading (%)": round(gen_best_loading, 6) if gen_best_loading is not None else None,
                "Generation Min Vpu": round(gen_best_min_v, 6) if gen_best_min_v is not None else None,
                "Generation Max Vpu": round(gen_best_max_v, 6) if gen_best_max_v is not None else None,
                "Feasible Count": feasible_count,
                "Reject Count": reject_count,
                "Global Best S": round(best_score, 6),
                "Global Best Binary": format_binary(best),
                "Global Best Feasible": best_feasible,
                "Global Best Reject Reason": best_reject_reason
            })

            for elite_rank, item in enumerate(ranked[:ELITE_SIZE], start=1):
                elite_ind, elite_s, elite_p, elite_q, elite_feasible, elite_reject_reason, elite_loading, elite_min_v, elite_max_v = item

                elite_rows.append({
                    "Generation": gen + 1,
                    "Elite Rank": elite_rank,
                    "Config List": str(elite_ind),
                    "Config Binary": format_binary(elite_ind),
                    "S": round(elite_s, 6),
                    "P BC (MW)": round(elite_p, 6) if elite_p is not None else None,
                    "Q BC (Mvar)": round(elite_q, 6) if elite_q is not None else None,
                    "Feasible": elite_feasible,
                    "Reject Reason": elite_reject_reason,
                    "Max Loading (%)": round(elite_loading, 6) if elite_loading is not None else None,
                    "Min Vpu": round(elite_min_v, 6) if elite_min_v is not None else None,
                    "Max Vpu": round(elite_max_v, 6) if elite_max_v is not None else None
                })

            print(
                f"Gen {gen+1:>2} | "
                f"Gen Best = {gen_best_score:.4f} | "
                f"Global Best = {best_score:.4f} | "
                f"Feasible = {feasible_count}/{POP_SIZE} | "
                f"Binary = {format_binary(best)}"
            )

            scores = [item[1] for item in evaluated]
            pop_only = [item[0][:] for item in evaluated]

            # ELITISM
            new_pop = [ind[:] for ind, _, _, _, _, _, _, _, _ in ranked[:ELITE_SIZE]]

            # Fill rest with selection + crossover + mutation
            while len(new_pop) < POP_SIZE:
                parent1 = tournament_select(pop_only, scores)
                parent2 = tournament_select(pop_only, scores)

                child1, child2 = crossover(parent1, parent2)

                child1 = mutate(child1)
                new_pop.append(child1)

                if len(new_pop) < POP_SIZE:
                    child2 = mutate(child2)
                    new_pop.append(child2)

            pop = new_pop[:POP_SIZE]

        print("\n=== HASIL TERBAIK ===")
        print("Config List :", best)
        print("Config Bin  :", format_binary(best))
        print("S           :", round(best_score, 4))
        print("P BC        :", round(best_p, 4) if best_p is not None else None)
        print("Q BC        :", round(best_q, 4) if best_q is not None else None)
        print("Feasible    :", best_feasible)
        print("Reject      :", best_reject_reason)
        print("Max Loading :", round(best_max_loading, 4) if best_max_loading is not None else None)
        print("Min Vpu     :", round(best_min_v, 4) if best_min_v is not None else None)
        print("Max Vpu     :", round(best_max_v, 4) if best_max_v is not None else None)
        print("\nKeterangan  : 1 = Bus A, 0 = Bus B")

        best_summary = {
            "Best Config List": str(best),
            "Best Config Binary": format_binary(best),
            "Best S": round(best_score, 6),
            "Best P BC (MW)": round(best_p, 6) if best_p is not None else None,
            "Best Q BC (Mvar)": round(best_q, 6) if best_q is not None else None,
            "Feasible": best_feasible,
            "Reject Reason": best_reject_reason,
            "Max Loading (%)": round(best_max_loading, 6) if best_max_loading is not None else None,
            "Min Vpu": round(best_min_v, 6) if best_min_v is not None else None,
            "Max Vpu": round(best_max_v, 6) if best_max_v is not None else None,
            "BC Target": "BC Waru",
            "Target GI": TARGET_SUB_NAME,
            "Meaning": "1=Bus A, 0=Bus B"
        }

        best_config_rows = build_config_detail(best)

        save_to_excel(gen_rows, all_individual_rows, elite_rows, best_summary, best_config_rows)
        print("\nHasil berhasil disimpan ke Excel:")
        print(OUTPUT_FILE)

    finally:
        restore_original_state(ds_map, original_state)
        print("\nKondisi DS sudah dikembalikan ke kondisi awal.")


if __name__ == "__main__":
    run()
```

```python
import os
import math
import random
import powerfactory
import pandas as pd

app = powerfactory.GetApplication()

# ── CONFIG ─────────────────────────────
BAY_LIST = [
    "nwaru-rnkut 2",
    "nwaru-rnkut 1",
    "waru-rnkut 2",
    "waru-rnkut 1",
    "4rnkut5_td2",
    "4rnkut5_td3",
    "4rnkut5_td4",
    "4rnkut5_td5",
    "rnkut-hanil",
    "sbsln-rnkut 1",
    "sbsln-rnkut 2",
    "rnkut-slilo 1",
    "rnkut-slilo 2"
]

POP_SIZE = 40
N_GEN = 5
MUT_PROB = 0.02
CROSSOVER_PROB = 0.80
ELITE_SIZE = 1

# Constraint target GI
# Sesuaikan kalau nama GI di PowerFactory bukan "RNKUT5"
TARGET_SUB_NAME = "RUNGKUT5"
VMIN = 0.95
VMAX = 1.10
MAX_LINE_LOADING = 80.0   # %

OUTPUT_DIR = r"D:\Robbyo\S2 UGM\2026\Tesis\DIGSILENT DATA"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "GA_DS_Rungkut_Result.xlsx")

# 1 = Bus A  -> DSA close, DSB open
# 0 = Bus B  -> DSA open,  DSB close

# Dari tabel GI Rungkut:
# 1  NWARU-RNKUT 2   = Bus B = 0
# 2  NWARU-RNKUT 1   = Bus A = 1
# 3  WARU-RNKUT 2    = Bus B = 0
# 4  WARU-RNKUT 1    = Bus A = 1
# 5  4RNKUT5_TD2     = Bus B = 0
# 6  4RNKUT5_TD3     = Bus B = 0
# 7  4RNKUT5_TD4     = Bus B = 0
# 8  4RNKUT5_TD5     = Bus B = 0
# 9  RNKUT-HANIL     = Bus B = 0
# 10 SBSLN-RNKUT 1   = Bus A = 1
# 11 SBSLN-RNKUT 2   = Bus B = 0
# 12 RNKUT-SLILO 1   = Bus A = 1
# 13 RNKUT-SLILO 2   = Bus B = 0
INITIAL_BINARY_FROM_TABLE = "0101000001010"


# ── BUILD DS MAP ───────────────────────
def build_ds_map():
    ds_map = {}
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        ds_map[obj.loc_name.lower().strip()] = obj
    return ds_map


# ── SAVE ORIGINAL STATE ────────────────
def save_original_state(ds_map):
    original = {}
    for bay in BAY_LIST:
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa in ds_map:
            original[dsa] = ds_map[dsa].IsOpen()
        if dsb in ds_map:
            original[dsb] = ds_map[dsb].IsOpen()

    return original


# ── RESTORE ORIGINAL STATE ─────────────
def restore_original_state(ds_map, original):
    for name, is_open in original.items():
        if name not in ds_map:
            continue

        obj = ds_map[name]
        if is_open:
            obj.Open()
        else:
            obj.Close()


# ── APPLY CONFIG ───────────────────────
def apply_config(ds_map, ind):
    for i, bay in enumerate(BAY_LIST):
        dsa = "dsa_" + bay
        dsb = "dsb_" + bay

        if dsa not in ds_map or dsb not in ds_map:
            continue

        if ind[i] == 1:
            ds_map[dsa].Close()
            ds_map[dsb].Open()
        else:
            ds_map[dsa].Open()
            ds_map[dsb].Close()


# ── LOADFLOW ───────────────────────────
def run_lf():
    app.Rebuild()
    lf = app.GetFromStudyCase("ComLdf")
    if lf is None:
        return False
    return lf.Execute() == 0


# ── GET BC ─────────────────────────────
def get_bc():
    for obj in app.GetCalcRelevantObjects("*.ElmCoup"):
        if obj.loc_name.lower().strip() == "bc rungkut":
            return obj
    return None


# ── GET TARGET SUBSTATION ──────────────
def get_target_substation():
    for sub in app.GetCalcRelevantObjects("*.ElmSubstat"):
        if sub.loc_name.strip().lower() == TARGET_SUB_NAME.lower():
            return sub
    return None


# ── GET TARGET BUSES ───────────────────
def get_target_buses(target_sub):
    buses = []
    terms = target_sub.GetContents("*.ElmTerm", 1)

    for bus in terms:
        try:
            if bus.GetAttribute("iUsage") != 0:
                continue
        except Exception:
            continue
        buses.append(bus)

    return buses


# ── CHECK CONSTRAINTS ──────────────────
def check_constraints(target_buses, all_lines):
    voltage_violation = False
    loading_violation = False

    min_v = None
    max_v = None
    max_loading = 0.0

    # 1) Voltage constraint on buses in target GI
    for bus in target_buses:
        try:
            vpu = bus.GetAttribute("m:u")
        except Exception:
            continue

        if vpu is None:
            continue

        if min_v is None or vpu < min_v:
            min_v = vpu
        if max_v is None or vpu > max_v:
            max_v = vpu

        if vpu < VMIN or vpu > VMAX:
            voltage_violation = True

    # 2) Line loading constraint on lines connected to target GI
    for line in all_lines:
        try:
            bus1 = line.bus1.cterm
            bus2 = line.bus2.cterm
        except Exception:
            continue

        if bus1 in target_buses or bus2 in target_buses:
            try:
                loading = line.GetAttribute("m:loading")
            except Exception:
                continue

            if loading is None:
                continue

            if loading > max_loading:
                max_loading = loading

            if loading >= MAX_LINE_LOADING:
                loading_violation = True

    reject_reasons = []
    if voltage_violation:
        reject_reasons.append("VOLTAGE_LIMIT")
    if loading_violation:
        reject_reasons.append("LINE_LOADING_LIMIT")

    feasible = not voltage_violation and not loading_violation
    reject_reason = "OK" if feasible else ";".join(reject_reasons)

    return feasible, reject_reason, max_loading, min_v, max_v


# ── FITNESS ────────────────────────────
def fitness(ds_map, bc, target_buses, all_lines, ind):
    apply_config(ds_map, ind)

    if not run_lf():
        return 9999.0, None, None, False, "LOADFLOW_FAILED", None, None, None

    p = bc.GetAttribute("m:P:bus1")
    q = bc.GetAttribute("m:Q:bus1")

    if p is None or q is None:
        return 9999.0, None, None, False, "BC_RESULT_MISSING", None, None, None

    feasible, reject_reason, max_loading, min_v, max_v = check_constraints(target_buses, all_lines)

    if not feasible:
        return 9999.0, p, q, False, reject_reason, max_loading, min_v, max_v

    s = math.sqrt(p * p + q * q)
    return s, p, q, True, reject_reason, max_loading, min_v, max_v


# ── INIT POP ───────────────────────────
def init_pop(n):
    pop = [[random.randint(0, 1) for _ in range(n)] for _ in range(POP_SIZE)]

    # Optional: masukkan kondisi existing dari tabel sebagai salah satu individu awal
    seed = [int(x) for x in INITIAL_BINARY_FROM_TABLE]
    if len(seed) == n:
        pop[0] = seed

    return pop


# ── MUTATION ───────────────────────────
def mutate(ind):
    return [bit ^ 1 if random.random() < MUT_PROB else bit for bit in ind]


# ── CROSSOVER ──────────────────────────
def crossover(a, b):
    if len(a) <= 1:
        return a[:], b[:]

    if random.random() > CROSSOVER_PROB:
        return a[:], b[:]

    pt = random.randint(1, len(a) - 1)
    return a[:pt] + b[pt:], b[:pt] + a[pt:]


# ── FORMAT BINARY ──────────────────────
def format_binary(ind):
    return "".join(str(bit) for bit in ind)


# ── BUILD CONFIG DETAIL ────────────────
def build_config_detail(ind):
    rows = []
    for i, bay in enumerate(BAY_LIST):
        bit = ind[i]

        if bit == 1:
            bus = "Bus A"
            dsa_state = "CLOSE"
            dsb_state = "OPEN"
        else:
            bus = "Bus B"
            dsa_state = "OPEN"
            dsb_state = "CLOSE"

        rows.append({
            "No": i + 1,
            "Bay": bay,
            "Bit": bit,
            "Bus": bus,
            "DSA": dsa_state,
            "DSB": dsb_state
        })
    return rows


# ── SAVE TO EXCEL ──────────────────────
def save_to_excel(gen_rows, all_individual_rows, elite_rows, best_summary, best_config_rows):
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    df_gen = pd.DataFrame(gen_rows)
    df_all = pd.DataFrame(all_individual_rows)
    df_elite = pd.DataFrame(elite_rows)
    df_best = pd.DataFrame([best_summary])
    df_cfg = pd.DataFrame(best_config_rows)
    df_param = pd.DataFrame([
        {"Parameter": "POP_SIZE", "Value": POP_SIZE},
        {"Parameter": "N_GEN", "Value": N_GEN},
        {"Parameter": "MUT_PROB", "Value": MUT_PROB},
        {"Parameter": "CROSSOVER_PROB", "Value": CROSSOVER_PROB},
        {"Parameter": "ELITE_SIZE", "Value": ELITE_SIZE},
        {"Parameter": "TARGET_SUB_NAME", "Value": TARGET_SUB_NAME},
        {"Parameter": "VMIN", "Value": VMIN},
        {"Parameter": "VMAX", "Value": VMAX},
        {"Parameter": "MAX_LINE_LOADING", "Value": MAX_LINE_LOADING},
        {"Parameter": "N_BAY", "Value": len(BAY_LIST)},
        {"Parameter": "BC_TARGET", "Value": "BC Rungkut"},
        {"Parameter": "Binary Meaning", "Value": "1=Bus A, 0=Bus B"},
        {"Parameter": "Initial Binary From Table", "Value": INITIAL_BINARY_FROM_TABLE},
        {"Parameter": "Reject Rule", "Value": "Loading>=80% or V outside 0.95-1.10 pu"},
        {"Parameter": "Output File", "Value": OUTPUT_FILE},
    ])

    with pd.ExcelWriter(OUTPUT_FILE) as writer:
        df_gen.to_excel(writer, sheet_name="Generation History", index=False)
        df_all.to_excel(writer, sheet_name="All Individuals", index=False)
        df_elite.to_excel(writer, sheet_name="Elite History", index=False)
        df_best.to_excel(writer, sheet_name="Best Summary", index=False)
        df_cfg.to_excel(writer, sheet_name="Best Config Detail", index=False)
        df_param.to_excel(writer, sheet_name="GA Parameters", index=False)


# ── TOURNAMENT SELECTION ───────────────
def tournament_select(pop, scores):
    i, j = random.sample(range(len(pop)), 2)
    return pop[i][:] if scores[i] < scores[j] else pop[j][:]


# ── MAIN GA ────────────────────────────
def run():
    if app is None:
        raise RuntimeError("Cannot connect to PowerFactory")

    if ELITE_SIZE < 0 or ELITE_SIZE >= POP_SIZE:
        raise ValueError("ELITE_SIZE must be >= 0 and < POP_SIZE")

    ds_map = build_ds_map()
    bc = get_bc()
    target_sub = get_target_substation()

    if bc is None:
        raise RuntimeError("BC Rungkut tidak ditemukan")

    if target_sub is None:
        raise RuntimeError(f"Target substation '{TARGET_SUB_NAME}' tidak ditemukan")

    target_buses = get_target_buses(target_sub)
    if not target_buses:
        raise RuntimeError(f"Tidak ada bus yang valid di GI '{TARGET_SUB_NAME}'")

    all_lines = app.GetCalcRelevantObjects("*.ElmLne")
    original_state = save_original_state(ds_map)
    pop = init_pop(len(BAY_LIST))

    best = None
    best_score = 1e9
    best_p = None
    best_q = None
    best_feasible = False
    best_reject_reason = None
    best_max_loading = None
    best_min_v = None
    best_max_v = None

    gen_rows = []
    all_individual_rows = []
    elite_rows = []

    try:
        for gen in range(N_GEN):
            evaluated = []

            for idx, ind in enumerate(pop, start=1):
                s, p, q, feasible, reject_reason, max_loading, min_v, max_v = fitness(
                    ds_map, bc, target_buses, all_lines, ind
                )

                evaluated.append((ind[:], s, p, q, feasible, reject_reason, max_loading, min_v, max_v))

                all_individual_rows.append({
                    "Generation": gen + 1,
                    "Individual No": idx,
                    "Config List": str(ind),
                    "Config Binary": format_binary(ind),
                    "S": round(s, 6),
                    "P BC (MW)": round(p, 6) if p is not None else None,
                    "Q BC (Mvar)": round(q, 6) if q is not None else None,
                    "Feasible": feasible,
                    "Reject Reason": reject_reason,
                    "Max Loading (%)": round(max_loading, 6) if max_loading is not None else None,
                    "Min Vpu": round(min_v, 6) if min_v is not None else None,
                    "Max Vpu": round(max_v, 6) if max_v is not None else None
                })

            ranked = sorted(evaluated, key=lambda x: x[1])

            gen_best_ind = ranked[0][0][:]
            gen_best_score = ranked[0][1]
            gen_best_p = ranked[0][2]
            gen_best_q = ranked[0][3]
            gen_best_feasible = ranked[0][4]
            gen_best_reject_reason = ranked[0][5]
            gen_best_loading = ranked[0][6]
            gen_best_min_v = ranked[0][7]
            gen_best_max_v = ranked[0][8]

            if gen_best_score < best_score:
                best_score = gen_best_score
                best = gen_best_ind[:]
                best_p = gen_best_p
                best_q = gen_best_q
                best_feasible = gen_best_feasible
                best_reject_reason = gen_best_reject_reason
                best_max_loading = gen_best_loading
                best_min_v = gen_best_min_v
                best_max_v = gen_best_max_v

            feasible_count = sum(1 for item in evaluated if item[4])
            reject_count = len(evaluated) - feasible_count

            gen_rows.append({
                "Generation": gen + 1,
                "Generation Best S": round(gen_best_score, 6),
                "Generation Best Binary": format_binary(gen_best_ind),
                "Generation Best P BC (MW)": round(gen_best_p, 6) if gen_best_p is not None else None,
                "Generation Best Q BC (Mvar)": round(gen_best_q, 6) if gen_best_q is not None else None,
                "Generation Best Feasible": gen_best_feasible,
                "Generation Best Reject Reason": gen_best_reject_reason,
                "Generation Max Loading (%)": round(gen_best_loading, 6) if gen_best_loading is not None else None,
                "Generation Min Vpu": round(gen_best_min_v, 6) if gen_best_min_v is not None else None,
                "Generation Max Vpu": round(gen_best_max_v, 6) if gen_best_max_v is not None else None,
                "Feasible Count": feasible_count,
                "Reject Count": reject_count,
                "Global Best S": round(best_score, 6),
                "Global Best Binary": format_binary(best),
                "Global Best Feasible": best_feasible,
                "Global Best Reject Reason": best_reject_reason
            })

            for elite_rank, item in enumerate(ranked[:ELITE_SIZE], start=1):
                elite_ind, elite_s, elite_p, elite_q, elite_feasible, elite_reject_reason, elite_loading, elite_min_v, elite_max_v = item

                elite_rows.append({
                    "Generation": gen + 1,
                    "Elite Rank": elite_rank,
                    "Config List": str(elite_ind),
                    "Config Binary": format_binary(elite_ind),
                    "S": round(elite_s, 6),
                    "P BC (MW)": round(elite_p, 6) if elite_p is not None else None,
                    "Q BC (Mvar)": round(elite_q, 6) if elite_q is not None else None,
                    "Feasible": elite_feasible,
                    "Reject Reason": elite_reject_reason,
                    "Max Loading (%)": round(elite_loading, 6) if elite_loading is not None else None,
                    "Min Vpu": round(elite_min_v, 6) if elite_min_v is not None else None,
                    "Max Vpu": round(elite_max_v, 6) if elite_max_v is not None else None
                })

            print(
                f"Gen {gen+1:>2} | "
                f"Gen Best = {gen_best_score:.4f} | "
                f"Global Best = {best_score:.4f} | "
                f"Feasible = {feasible_count}/{POP_SIZE} | "
                f"Binary = {format_binary(best)}"
            )

            scores = [item[1] for item in evaluated]
            pop_only = [item[0][:] for item in evaluated]

            # ELITISM
            new_pop = [ind[:] for ind, _, _, _, _, _, _, _, _ in ranked[:ELITE_SIZE]]

            # Fill rest with selection + crossover + mutation
            while len(new_pop) < POP_SIZE:
                parent1 = tournament_select(pop_only, scores)
                parent2 = tournament_select(pop_only, scores)

                child1, child2 = crossover(parent1, parent2)

                child1 = mutate(child1)
                new_pop.append(child1)

                if len(new_pop) < POP_SIZE:
                    child2 = mutate(child2)
                    new_pop.append(child2)

            pop = new_pop[:POP_SIZE]

        print("\n=== HASIL TERBAIK ===")
        print("Config List :", best)
        print("Config Bin  :", format_binary(best))
        print("S           :", round(best_score, 4))
        print("P BC        :", round(best_p, 4) if best_p is not None else None)
        print("Q BC        :", round(best_q, 4) if best_q is not None else None)
        print("Feasible    :", best_feasible)
        print("Reject      :", best_reject_reason)
        print("Max Loading :", round(best_max_loading, 4) if best_max_loading is not None else None)
        print("Min Vpu     :", round(best_min_v, 4) if best_min_v is not None else None)
        print("Max Vpu     :", round(best_max_v, 4) if best_max_v is not None else None)
        print("\nKeterangan  : 1 = Bus A, 0 = Bus B")

        best_summary = {
            "Best Config List": str(best),
            "Best Config Binary": format_binary(best),
            "Best S": round(best_score, 6),
            "Best P BC (MW)": round(best_p, 6) if best_p is not None else None,
            "Best Q BC (Mvar)": round(best_q, 6) if best_q is not None else None,
            "Feasible": best_feasible,
            "Reject Reason": best_reject_reason,
            "Max Loading (%)": round(best_max_loading, 6) if best_max_loading is not None else None,
            "Min Vpu": round(best_min_v, 6) if best_min_v is not None else None,
            "Max Vpu": round(best_max_v, 6) if best_max_v is not None else None,
            "BC Target": "BC Rungkut",
            "Target GI": TARGET_SUB_NAME,
            "Meaning": "1=Bus A, 0=Bus B"
        }

        best_config_rows = build_config_detail(best)

        save_to_excel(gen_rows, all_individual_rows, elite_rows, best_summary, best_config_rows)
        print("\nHasil berhasil disimpan ke Excel:")
        print(OUTPUT_FILE)

    finally:
        restore_original_state(ds_map, original_state)
        print("\nKondisi DS sudah dikembalikan ke kondisi awal.")


if __name__ == "__main__":
    run()
```
